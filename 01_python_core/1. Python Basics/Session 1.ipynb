## Why Python

1. Design Philosophy: Easiest language to learn; readable; proper indentations.
2. Batteries included: The inbuilt features and libraries that are included with the python package.
3. General Purpose Language: Python supports a lot of functionalities in all fields and is not limited only till data science.
4. Libraries/Community: Has a number of communities that support in data science. So code for general operations used in data science will not have to be typed.

## Features of Python

1. Python offers dynamic typing. That is variables do not have to be exclusively declared during or before the naming of the variable.
2. Python offers Dynamic Binding. That is, variables do not have a statically defined data type, rather, a created value of any data type can have value of data type other than the initialised one.

```python
a = 5
a = “This is a String”
```
