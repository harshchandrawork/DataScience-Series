### Q6:- Write a program that will tell the number of dogs and chicken are there when the user will provide the value of total heads and legs.

For example:
Input:
heads -> 4
legs -> 12
<br>
Output:
dogs -> 2
chicken -> 2





### Solution for the given question:
```
heads:
1d+1c = d+c = (H)
c = H-d ------- (1)

legs:
4d+2c = (L) {since dogs have 4 legs and chicken have 2}
taking the value of c from eqn (1):
4d+2(H-d) = L

Simplifying for d:
4d+2(H-d) = L
4d+2H-2d = L
2d+2H = L
d = (L-2H)/2

Substituting the value of d in c:
c = H-d
```

````python
def dogChicken(heads, legs):
  d = (legs-(2*heads))/2
  c = heads-d
  return d, c

heads, legs = input("Enter the number of heads and legs respectively, seperated by a space").split()
heads = int(heads); legs = int(legs)
dogs, chicken = dogChicken(heads, legs)
print(f"dogs -> {int(dogs)}; chicken -> {int(chicken)}")
````

### Q7:- Write a program to find the sum of squares of first n natural numbers where n will be provided by the user.

````python
n = int(input("Enter a number for n: "))
if (n < 1): print(f"Given number {n} is not a natural number.")
sumOfNum = 0
for i in range(1, n+1): sumOfNum += (i**2)
print(f"Sum of all the natural numbers up till {n} is {sumOfNum}")
````

### Q8:- Given the first 2 terms of an Arithmetic Series.Find the Nth term of the series. Assume all inputs are provided by the user.

````python
term1 = int(input("Enter 1st integer term: "))
term2 = int(input("Enter 2nd integer term: "))
n = int(input("Enter the Nth term: "))
diff = term2 - term1
latest_term = term2
for i in range(term2, n+1): latest_term += diff
print("Nth term is ", latest_term)
````

### Q9:- Given 2 fractions, find the sum of those 2 fractions.Take the numerator and denominator values of the fractions from the user.

$$
\frac{a}{b} + \frac{c}{d} = \frac{ad}{bd} + \frac{bc}{bd} = \frac{ad + bc}{bd}
$$


````python
def fractionSum(ad, bc, bd):
  return (ad+bc)/bd

nume1, denom1 = input("Enter numerator and denominator for first number, seperated by a space: ").split()
nume2, denom2 = input("Enter numerator and denominator for second number, seperated by a space: ").split()
nume1 = int(nume1); denom1 = int(denom1); nume2 = int(nume2); denom2 = int(denom2)
fractionSum(nume1*denom2, nume2*denom1, denom1*denom2)
````
