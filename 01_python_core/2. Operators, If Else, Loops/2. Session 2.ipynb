## Guessing Game
`Problem`: The computer will pick a random number between 0 and 100. User has 3 chances to guess the right number. With each incorrect guess, the computer tells the user if the guess number is greater or smaller than the random number picked by computer.

```python
import random

print('''
Rules of Guessing Game:
1. The computer will think of a number between 1 and 100.
2. You will have 3 chances to guess the right number.
''')

random_num = random.randint(1, 100)
i = 0
guess = int(input("Guess a number between 1 and 100: "))
while i < 2:
  if guess < random_num:
    print(f"Guess is too low! Try guessing a number higher than {guess} \n")
  elif guess > random_num:
    print(f"Guess is too high! Try guessing a number lower than {guess} \n")
  elif guess == random_num:
    break
  guess = int(input("Guess a number: "))
  i += 1
if guess == random_num:
  print(f"Congratulations! You guessed the right number {random_num}")
else:
  print(f"Failed to guess the right number :( The Jackpot number is {random_num}")
```

## Population Program
Below is a program that calculates the population for 10 given years, knowing that the initial population is 10,000 people. 

```python
curr_pop = 10000
print(f"Initial Population is {curr_pop}")
for i in range(0, 11):
  curr_pop += curr_pop*0.1
  print(f"Population for the {i}st year is {curr_pop}")
```

## Population Program (decreasing order)
 The above code can be print in the reverse order using a similar logic. That is, to print the predecesive population for given past 10 years, knowing that the initial population of the latest year is 10,000 people

```python
latest_pop = 10000
print(f"Population for the latest year is {latest_pop}")
for i in range(10, 0, -1):
   prev_pop = latest_pop * 0.9
   latest_pop = latest_pop - latest_pop*0.1
   print(f"Population for the {i}th year is {prev_pop}")
```

the above code may seem to be the right code but when studied with basic maths, it is not the right code. Below is provided the right code.

```python
latest_pop = 10000
print(f"Population for the latest year is {latest_pop}")
for i in range(10, 0, -1):
  prev_pop = latest_pop / 1.1
  latest_pop = prev_pop
  print(f"Population for the {i}th year is {prev_pop}")
```
