### `Problem 6` - Find the factorial of a given number.

Write a program to use the loop to find the factorial of a given number.

The factorial (symbol: `!`) means to multiply all whole numbers from the chosen number down to 1.

For example: calculate the factorial of 5

```bash
5! = 5 × 4 × 3 × 2 × 1 = 120
```

Output:

```bash
120
```

````python
def factorial(num):
  if num == 0:
    return 0
  fact = 1
  for i in range(1, num+1):
    fact = fact*i
  return fact
num = int(input("Enter a number to find Factorial: "))
if factorial(num)==0:
  print("Entered number is not valid to find Factorial")
else:
  print(f"Factorial of {num}, is {factorial(num)}")
````

### `Problem 7` - Reverse a given integer number.

Example:

`Input:`

```bash
76542
```

`Output:`

```bash
24567
```

````python
# function to return the reverse of an integer using arithmetic function
def arithmetic_reverse(num):
  reversed_num = 0
  while num != 0:
    last_digit = num % 10
    num //= 10
    reversed_num = reversed_num * 10 + last_digit
  return reversed_num

# function to reverse an integer using recursion
def recursive_reverse(num, reversed_num = 0):
  # base condition to return the reversed integer
  if num == 0:
    return reversed_num
  # extracting the last digit and adding it to the reversed integer
  last_digit = num % 10
  reversed_num = reversed_num * 10 + last_digit
  num //= 10
  return recursive_reverse(num, reversed_num)

num = int(input("Enter an integer: "))
reversed_integer = recursive_reverse(num)
print(reversed_integer)
````

````python
# recursive function
def sum_num(n, sumTotal, i):
  if i > n or sumTotal > 300 : # base condition to stop the recursion
    return sumTotal
  if i % 5 == 0: # skip if number if divisble by 5
    return sum_num(n, sumTotal, i+1) # going to the next number using i+1
  # sumTotal += i
  return sum_num(n, sumTotal+i, i+1) # incrementing current nummber (i) and adding current number (i) to sumTotal

n = int(input("Enter an integer for n"))
sumofall = sum_num(n, sumTotal=0, i=1) # calling the recursive function and providing the initial parameters
print(f"Sum of the numbers till {n} is {sumofall}")
  
````

### `Problem 9`: Write a program that keeps on accepting a number from the user until the user enters Zero. Display the sum and average of all the numbers.

````python
def acceptingNumbers(sum, count):
  sum = count = 0
  while True:
    number = float(input("Enter a number: "))
    if number == 0:
      break
    sum += number
    count += 1
  return sum, count

# trying the above program using recursion
def acceptingNumbers_recursive(sum, count):
  number = float(input("Enter a number: "))
  if number == 0:
    return sum, count
  sum += number
  count += 1
  return acceptingNumbers(sum, count)

sum, count = acceptingNumbers(sum=0, count=0)
avg = sum / count
print(f"Sum of numbers entered is: {sum} \nAverage of numbers entered is: {avg}")
````

### `Problem 10`: Write a program, which will find all such numbers between 1000 and 3000 (both included) such that each digit of the number is an even number. The numbers obtained should be printed in a space-separated sequence on a single line.



````python
# obtaining the output using python strings
def print_usingString():
  for i in range(1000, 3001):
    # converting i to a string to traverse over it
    i_str = str(i)
    even = True  # Assume all digits are even initially
    # loop to iterate over each digit
    for j in range(len(i_str)):
      digit = int(i_str[j]) # converting j back to integer to check even or odd
      if digit % 2 != 0:
        even = False # even var used as a flag to mark False if any digit is odd
        break
    if even:
      print(i, end=', ')
      
print_usingString()
````

````python
# obtaining the output using division and modulo to check each digit for even or odd
def print_usingDivmod():
  for i in range(1000, 3001):
    even = True
    num = i
    while num != 0:
      last_digit = num % 10
      if last_digit % 2 != 0:
        even = False
        break
      num = num // 10
    if even:
      print(i, end=', ')

print_usingDivmod()
````

### `Problem 11`: A robot moves in a plane starting from the original point (0,0). The robot can move toward UP, DOWN, LEFT and RIGHT with a given steps.
The trace of robot movement is shown as the following:
```
UP 5
DOWN 3
LEFT 3
RIGHT 2
!
```
> The numbers after the direction are steps.

> `!` means robot stop there.

**Please write a program to compute the distance from current position after a sequence of movement and original point.**

*If the distance is a float, then just print the nearest integer.*

Example:

`Input`:
```
UP 5
DOWN 3
LEFT 3
RIGHT 2
!
```
`Output`:
```
2
```

````python
import math
def calcDist(up, down, left, right):
  if up >= down: y_diff = up - down
  else: y_diff = down - up

  if left >= right: x_diff = left - right
  else: x_diff = right - left
  
  H = x_diff**2 + y_diff**2
  total_dist = math.sqrt(H)
  return int(total_dist)

up, down, left, right = 5, 3, 3, 2
calcDist(up, down, left, right)
````

### `Problem 12`:Write a program to print whether a given number is a prime number or not

````python
n = int(input("Enter a number to check for prime"))
def is_prime(n):
  numIs_prime = True
  for i in range(2, n):
    if n % i == 0:
      numIs_prime = False
      break
  return numIs_prime

if is_prime(n) == False:
  print(f"The number {n} is not a prime number")
else:
  print(f"The number {n} is a prime number")
````

### `Problem 13`:Print all the Armstrong numbers in a given range.
Range will be provided by the user<br>
Armstrong number is a number that is equal to the sum of cubes of its digits. For example 0, 1, 153, 370, 371 and 407 are the Armstrong numbers.

````python
start, end = input("Enter starting and ending range, seperated by a space.").split()
start = int(start); end = int(end)
for i in range(start, end+1):
  num = i; sum = 0
  while num != 0:
    last_digit = num % 10
    sum = sum + last_digit**3
    num = num // 10
  if sum == i:
    print(i, end=', ')
````
