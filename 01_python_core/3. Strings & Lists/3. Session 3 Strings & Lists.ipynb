Below is the code for finding the population for past 10 years if the population has been increasing by 10% each year.

````python
latest_pop = 10000
print(f"Population for the latest year is {latest_pop}")
for i in range(10, 0, -1):
  prev_pop = latest_pop / 1.1
  latest_pop = prev_pop
  print(f"Population for the {i}th year is {prev_pop}")
````

## Sequence Sum
Find the sum of the provided series till the Nth term. Eg: 1/1! + 2/2! + 3/3! + ... + Nth term

````python
# code for finding the factorial
def factorial(n):
  fact = 1
  for i in range (1, n+1):
    fact = fact * i  
  return fact
n = int(input("Enter a number for iteration till Nth term: "))
sum = 0
for i in range(1, n+1):
  sum += i/factorial(i)
print(f"Sum is {sum}")
````

## Pattern 2 (using nested loops)
```

```

````python
n = int(input("Enter the number of rows: "))
for i in range(1, n+1): #loop for traversing rows
  for j in range(1, i): #loop for traversing columns of each row
    print (j,end= "")
  print()
````

The working of nested loops is the same as in python. Its rather the functioning of the print function that makes the printing work differently. For this reason, the "end" attribute is set to "", which is by default "\n", thus making the print function change the line after each time the print function is called. 
The print function in the latter is being called outside the j loop but inside the i loop at the end to change the line at the end of the row.

# Strings
Strings are sequenve of Characters. In Python specifically, strings are a sequence of Unicode Characters.

## String Creation

````python
# below are the different ways to initiate a string
s = 'hello'
s = "hello"
s = '''hello'''
s = """hello"""
s = str('hello')
print(s)
````

## Accessing Substrings from a String
Upon the creation of strings, indexes are assigned to the characters of the string automatically based on their position, starting from 0.

````python
# Positive Indexing
s = "hello world"
print(s[0])  # prints 'h'

# Negative indexing
print(s[-1]) # prints 'd'

# Slicing
print(s[1:5])  # prints 'ello'
print(s[2:])   # prints 'llo world'
print(s[:3])   # prints 'hel'

# slicing with an incremental value
print(s[0:6:2]) # means that indexes each 2nd value from 0 to 6 will be printed

# similarly reverse slicing can be done by using decremental value
print(s[6:0:-1])

# prints the reverse of the whole string
print(s[::-1])
````

## Editing and Deleting in Strings
The very first thing to note about python strings is that strings in python cannot be edited after creation. That is, strings in python are immutable. 

````python
s = "hello world"
s[0] = 'H' # will give an error because of the immutability of string.
````

'del' operation can be used to delete the string, but it cannot be used to delete a part of the string, since that will be equivalent to editing as well.

````python
s = 'hello world'
del s[-1:-5:2]
print(s) # will give an error
````

### Side note: How strings are compared
* As such in the last line of the above code, i.e. *'mumbai' > 'pune'* <br>
* The comparison is done lexicographically, i.e. character by character. <br>
* In lexicography, the characters are compared in order. That is, 'm' from 'mumbai' is compared to 'p' from 'pune'.
* Whichever one has a higher Unicode Point, will be stated as the greater character. 
* In case the first characters in order have the same unicode points, the second characters in order will be compared. 

## Common Functions on Strings
The following are some common functions that can be used on strings in Python. These are common functions as they work on all the data types, such as lists, strings, set, dictionary and tuple. These common functions are:
- `len()`: Returns the length of a string.
- `max()`: Returns the largest item in a string.
- `min()`: Returns the smallest item in a string.
- `sorted()`: Returns a list of elements of the string in a sorted order.
Note that these functions use ASCII values for the purpose of comparisons, since all the basic characters have ASCII values available. For any character that does not have an ASCII value, it's unicode value will be taken for the purpose.


## Functions specific to Strings
- `capitalize()`: Capitalises the first letter or character of each sentence or phrase.
- `title()`: Converts the first letter or character of each word to capital.
- `lower()`: Converts all characters to lowercase.
- `upper()`: Converts all characters to uppercase.
- `swapcase()`: Converts the uppercase characters to lowercase and vice-versa.

## Functions of Strings related to frequency
- `count()`: Calculates the frequency or occurence of a given paremeter in the string.
- `find()`: Returns the index of a given parameter in the string.
- `index()`: Returns the index of a given parameter in the string (similar to find()).

## Functions that return boolean values on Strings
- `endswith()`: Returns true is the string ends with the provided parameter.
- `startswith()`: Returns true if the string starts with the provided parameter.
