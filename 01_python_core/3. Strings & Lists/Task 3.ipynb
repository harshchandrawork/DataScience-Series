### `Problem 2`: Print the following pattern.

```bash
*
* *
* * *
* * * *
* * * * *
* * * *
* * *
* *
*
```

````python
n = int(input("Enter the number of rows"))
n+=1
for i in range(1, n):
  if(i <= n//2):
    print('* ' * i, end=' ') # print stars in increasing pattern till half way
  if(i > n//2):
    print('* ' * (n-i), end=' ') # print stars in decreasing pattern for last half
  print() # line break
````

### `Problem 3`:Write a program to print the following pattern

        *
      * * *
    * * * * *
   * * * * * * *
* * * * * * * * *


````python
def printPyramid (n):
  for i in range (1, n+1):
    for j in range (n-i):
      print('   ', end='')
    for k in range (i*2-1):
      print(' * ', end='')
    print()
n = int(input("Enter the number of rows"))
printPyramid(n)
````

### `Problem 7` - Find the sum of the series upto n terms.

**Example 1:**

`Input:`
```bash

```

`Output:`

```bash
2+22+222+2222+22222
```

````python
def sumTillN(n):
  sumOfSeries = 0
  currentTerm = '2' 
  for i in range(1, n+1):
    sumOfSeries += int(currentTerm)
    currentTerm = currentTerm + '2'
    if i == n:
      print(currentTerm, end='')
    else:
      print(currentTerm, end='+')
  print()
  return sumOfSeries

# using list of terms
def sumTillN_usingList(n):
  currentTerm = '2'
  terms = ['2']
  for i in range (1, n):
    currentTerm += '2'
    terms.append(currentTerm)
  return terms
  
n = int(input("Enter a value for n"))
terms = sumTillN_usingList(n)
print('+'.join(terms))
for i in range(len(terms)):
  terms[i] = int(terms[i])
print(f"Sum of above series is: {sum(terms)}")
````

### `Problem 9`: Write a program that will take a decimal number as input and prints out the binary equivalent of the number

````python
def decToBitwise_divmod(num):
  if num == 0:
    return '0'
  binary = ''
  while num != 0:
    remainder = num % 2
    num //= 2
    binary += str(remainder)
  return binary
num = int(input("Enter number to convert into binary"))
binary = decToBitwise_divmod(num)
print(binary[::-1])
print(len(binary))
````

### `Problem 10`: Write a program that will take 2 numbers as input and prints the LCM and HCF of those 2 numbers

````python
# finding lowest common multiple of two numbers
def lcm(m, n):
  for i in range(max(m, n), m*n):
    if i % m == 0 and i % n == 0:
      return i
  return m*n

m, n = input("Enter two numbers, seperated by a space,  to find their LCMM").split()
m = int(m); n = int(n)
print(f"LCM of {m} and {n} is {lcm(m, n)}")
````

````python
# finding highest common factor or greatest common divisor of two numbers
def gcd(m, n):
  for i in range(min(m, n), 0, -1):
    if m % i == 0 and n % i == 0:
      return i
  return 1

m, n = input("Enter two numbers, seperated by a space,  to find their LCMM").split()
m = int(m); n = int(n)
print(f"HCF or GCD of {m} and {n} is {gcd(m, n)}")
````

### **Problem 11:** Create Short Form from initial character
Given a string create short form ofthe string from Initial character. Short form should be capitalised.

Example:

`Input:`
```
Data science mentorship program
```

`Output:`
```

```

````python
def createShortForm(wordsList):
  shortFormList = []
  for i in range(len(wordsList)):
    shortFormList.append(wordsList[i][0])
  shortForm = ''.join(shortFormList).upper()
  return shortForm


string = str(input("Enter a string to create short form"))
wordsList = string.split()
print(createShortForm(wordsList))
````

### `Problem 12`: Append second string in the middle of first string

`Input:`
```
campusx
data
```
`Output`:

```
camdatapusx
```

````python
def appendInMiddle(string1, string2):
  half = len(string1)//2
  string3 = string1[:half] + string2 + string1[half:]
  return string3

string1 = 'campusx'
string2 = 'data'
print(appendInMiddle(string1, string2))
````

### ``Problem 13``:Given string contains a combination of the lower and upper case letters. Write a program to arrange the characters of a string so that all lowercase letters should come first.

Given:

`str1 = PyNaTive`

Expected Output:

`yaivePNT`

````python
string = 'PyNaTive'
lowercase = []
uppercase = []
for i in range(len(string)):
  if string[i].islower():
    lowercase.append(string[i])
  elif string[i].isupper():
    uppercase.append(string[i])
uppercase = ''.join(uppercase)
lowercase = ''.join(lowercase)
print(lowercase, uppercase, sep='')
````

### `Problem 18`: Find uncommon words from two Strings.

**Statement:** Given two sentences as strings **A** and **B**. The task is to return a list of all uncommon words. A word is uncommon if it appears exactly once in any one of the sentences, and does not appear in the other sentence. Note: A sentence is a string of space-separated words. Each word consists only of lowercase letters.

**Example 1:**

`Input:`

```bash
A = "apple banana mango"
B = "banana fruits mango"
```

`Output:`

```bash
['apple', 'fruits']
```

````python
A = "apple banana mango"
B = "banana fruits mango"
uncommonWords = []
a_list = A.split()
b_list = B.split()
for i in range(len(a_list)):
  if a_list.count(a_list[i]) == 1 and a_list[i] not in b_list:
    uncommonWords.append(a_list[i])
for i in range(len(b_list)):
  if b_list.count(b_list[i]) == 1 and b_list[i] not in a_list:
    uncommonWords.append(b_list[i])
print(uncommonWords)
````

### `Problem 19`: Word location in String.

**Statement:** Find a location of a word in a given sentence.

**Example 1:**

`Input:`

```bash
Sentence: We can learn data science through campusx mentorship program.

word: campusx
```

`Output:`

```bash
Location of the word is 7.
```

Note- Don't use index/find functions

````python
def findIndex(wordsList, word):
  locationIndex = 1
  for i in wordsList:
    if i == word:
      return locationIndex
    locationIndex += 1
  return 0

string = "We can learn data science through campusx mentorship program."
word = str(input("Enter a word to find its position"))
wordsList = string.split()
location = findIndex(wordsList, word)
if location != 0:
  print(f"Location of the word is {location}")
else:
  print(f"The word '{word}' does not exist in the string")
````

### `Problem 20`: Write a program that can remove all the duplicate characters from a string. User will provide the input.

````python
def removeDuplicate(string):
  for i in string:
    if string.count(i)>1:
      string=string.replace(i,'')
  return string

string = str(input("Enter a string"))
print(removeDuplicate(string))
````
