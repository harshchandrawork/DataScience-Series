## Find String Length
`Problem:` Find Length of a string provided by user, without using the `len()` function

```python
userString = input("Enter a String")
count = 0
for i in userString:
  count += 1
print(f"The length of the string is: {count}")
```

## Extract username from Email
`Problem:` Have to extract a part of string, that is, username from the Email. For example: if the email if user24@gmail.com, then the username be user24.

```python
# will iterate over the string and look for the presence of @ symbol
# then use slicing to select all the characters before @ symbol

def usingForLoop(s):
  count = 0
  for i in s:
    if i == '@':
      print(s[:count])
    count += 1

def usingIndexFunc(s):
  position = s.index('@')
  print(s[:position])

s = input("Enter an Email")
usingIndexFunc(s)
```

## Count Frequency
`Problem:` Given a string, find the occurence of a particular character in the string

```python
s = input("Enter a string")
find = input("Enter the character to find count of")
counter = 0
for i in s:
  if i == find:
    counter += 1
print(f"Count of {find} in '{s}' is {counter}")
```

## Remove Character
`Problem:` Remove a particular character from a string

```python
s = input("Enter a string")
toRemove = input("Enter the characters to remove")
result = ''
for i in s:
  if i != toRemove: # adding all the characters to result that are not to be removed
    result += i
print(result)

# or we can simpy use s.replace('toRemove')
```

## Palindrome check
`Problem:` Check if a given string is palindrome or not

```python
# using reverse string slicing to reverse the string and then check for palindrome
s = "malayalam"
reversed_s = s[::-1]
if s == reversed_s:
  print(f"The string '{s}' is a palindrome.")
else:
  print(f"The string '{s}' is not a palindrome.")
```

## Count Words
`Problem:` Count the number of words withous using the split() function

```python
def usingCounter(s):
  wordCount = 0
  for i in s:
    if i == ' ':
      wordCount += 1
  return wordCount

def usingAppendFunc(s):
  words = []
  temp = ''
  for i in s:
    if i != ' ':
      temp = temp + i 
    else:
      words.append(temp)
      temp = ' '
  words.append(temp)
  print(words)

s = input("Enter a string")
# print(f"There are {wordCount+1} words in the string '{s}'")
# usingCounter(s)
usingAppendFunc(s)
```
