# Time Complexity

Time Complexity can be measured using Techniques:
1. Measuring *time* to execute
2. *Counting* operations involved
3. Abstract *notion of order of growth*

### Measuring Time

````python
from time import time

start = time()
for i in range(1000):
  print(i)
end = time()
print(end-start)
````

The above approach measures the time taken to complete the task. But, this approach is never suitable at all because of the below mentioned problems:
1. Different time for different algorithm
2. Time varies if implementation changes.
3. Time varies if hardware changes.
4. Time varies if input size changes.
Therefore, we need a different approach to measure the performance of an algorithm.

### Counting Operations
This is a simple method, where each operation for the given program is calculated. <br>
It is assumed that all the mathematical, comparions, and assigment operators and accessing objects in memory takes *constant time*. *For example*,

````python
def mysum(x):
  total = 0 # -- 1 operation
  for i in range(x+1): # -- 1 operation, loop x times
    total += i  # -- 2 operations
  return total
````

Thus, the total number of operations is given by `1 + 3x`, where:
- 1 operation occurs outside the loop.
- 3 operations are performed within the loop, repeated `x` times, depending on the input size.

Therefore, the total number of operations is `1 + 3x`.

### Upon analyzing the pros and cons of this approach, we observe the following:

#### Cons:
- **Hardware dependence**: The method is dependent on the underlying hardware for performance.
- **Complexity**: Although the method is easy to implement, it may not be the most efficient solution.
- **Input-dependent time**: The execution time varies depending on the size of the input.

#### Pros:
- **Lack of a clear relationship**: The method does not establish a clear relationship between input size and execution time.
- **Uncertainty in operation counting**: It is difficult to definitively identify and count the exact number of operations involved.
- **Variability in execution time**: Execution time can vary significantly if the implementation or environment changes.


## What do we want while analyzing an Algorithm?
1. We want to evluate the algorithm
2. We want to evaluate *scalability*
3. We want to evaluate *in terms of input size*

It is to note that throughtout the study of Time Complexity, we will be assuming the *worst case scenarios*, even though for most of the time during the implementation we will be dealing with the average cases.

## Order of Growth
- We have to evaluate program's efficiency when input is very big.
- We have to express the growth of program's run time as input size grows, by forming a mathematical or structural relationship between the Growth of order of function and inputs.
- We have to put an *upper bound* on growth as tight as possible
- We do not have to be precise about the order of growth
- We will look at the *largest factors* in runtime, that is, which section or operation of program takes up more of the time during the execution phase. 
- *For example*, $n^2$ would take more time than n, since for each value of n, $n^2$ will always be greater.


## Exact Steps for Calculating O(n)
- Compute the number of operations being attempted.
- Worst case asymptotic complecity:
  - Ignore additive constants, such as `+1, -1, +2,` etc.
  - Ignore multiplicative constants, such as `2` in `2n`.

````python
def iterationFunc(n):
  answer = 1  # -- 1 operation
  while n > 1:  # -- 1 operation
    answer *= n  # -- 2 operation
    n -= 1  # -- 2 operation
  return answer  # -- 1 operation
````

- Thus for the above function, the number of operations being performed in the execution are `1+5n+1`, i.e. `5n+2`. <br>
- According to the steps mentioned above, we ignore the additive constant `2` and the multiplicative constant `5`. Consequently being left with `n`. <br>
- Now, this `n` does not state anything by itself, and so it will be placed along with the *Big-O notation*, i.e. `O`, which resembles the worst case. Therefore implying the time complexity of `O(n)`.

### Some examples
- `n^2 + 2n + 2`: We remove the additive constant 2 as well as the multiplicative constant 2 associate with the n. Thus the largest factorial that we have to be concerned about is the `n^2`.
- `n^2 + 100000n + 3^1000`: We will start by removing the additive and multiplicative constants $100000n$ and $3^{1000}$, because even though they may seem to be large numbers, it will still be assumed that $n$ in `n^2` is a far more larger number than the constants.
- `log(n) + n + 4`: The additive constant $4$ will be removed. Then we are left with $log(n)$ and $n$. Since, $log(n)$ will be a much smaller factor as compared to $n$, we take the time complexity to be `O(n)`.
- `0.0001*n*log(n) + 300n`: Removing the additive and multiplicative constants, we are left with $n*log(n)+n$.
  - Since $n*log(n)$ gives a very huge value for numbers greater than $10$, as compared to the linear function $n$, the greates factor we will be concerned about is `n*log(n)`.
  - Surely adding the linear n would make a slight progression in the output, but that much shall be negligible while working with huge numbers (assuming the worst case scenarios).
- `2n^30 + 3^n`: Here we can clearly see a competition between polynomial and an exponetial, i.e. between $n^{30}$ and $3^n$ respectively. Here, `3^n` will be a bigger factor for large values of $n$, as compared to $n^{30}$.

## Defining Polynomial Function and Exponential function and which will be a bigger factor.

### Polynomial Growth
A $Polynomial$ function is one where the variable $n$ is raised to a fixed power, which is 30 in this case, is a polynomial function because the exponent $(30)$ is a constant.
> Form: $f(n) = n^k$, where $k$ is a constant. <br>
> Growth Rate: Polynomial growth increases at a rate proportional to the power of $n$. The functions grows faster than $n$ or $n^2$, but slower than exponential functions like $3^n$.

### Exponential Growth
An $Exponential$  function grows much faster than anu polynomial function. In this case, $3^n$ is an exponential function because the base $(3)$ is a constant, and the exponent is the variable $n$.
> Form: $f(n) = a^n$, where $a$ is a constant (in this case, $a = 3$). <br>
> Growth Rate: Exponential growth increases at a rate proportional to the previous value of the function. In other words, the function doubles, triples, or increases by some constant factor with each increment in $n$.

### Comparison between Polynomial and Exponential function
- For small values of $n$, the growth of $n^{30}$ may seem significant, But as $n$ increases, exponential functions like $3^n$ will grow much faster than polynomial functions like $n^{30}$.
- For large $n$, exponential functions will dominate and outspace polynomial functions by an enormous margin.

## Law of Addition
The law of addition states that if $f(n)$ and $g(n)$ are two functions, then their time complexities can be added and the dominant term will be used for time complexity. <br>
- This law is used with sequential statements <br>
$O(f(n)) + O(f(n)) = O(f(n)+g(n))$, *for example*,
```
for i in range (n):
  print('a')
for j in range (n*n):
  print('b')
```
is $O(n) + O(n*n) = O(n+n^2) = O(n^2)$, *because of dominant term*

## Law of Multiplication
The law of multiplication states that if $f(n)$ and $g(n)$ are two functions, then their time complexities can be multiplied and the dominant term will be used for time complexity. <br>
- This law is used with nested statements/loops.
$O(f(n)) * O(f(n)) = O(f(n)*g(n))$, *for example*,
```
for i in range (n):
  for j in range(n):
    print('a')
```
is $O(n) * O(n) = O(n*n) = O(n^2)$, *because the outer loop goes $n$ times and the inner loop goes $n$ times for every outer loop iteration.
