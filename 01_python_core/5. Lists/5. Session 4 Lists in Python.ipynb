### What are Lists

List is a data type where you can store multiple items under 1 name. More technically, lists act like dynamic arrays which means you can add more items on the fly.<br>

*for example,* `L = [20, 'Bromon', 25.89, [30, 60, 90]]` <br>

Here, in the above example we can see that '[]' or square brackets are being used to create a list. The list is not pre-declared. As well as, the list has heterogeneous data types, i.e. variables or values of all types.

### Why Lists?
Lists are a fundamental data structure in programming. They allow you to store multiple values in a single variable.

## Array Vs Lists
Arrays and lists are both data structures that can store multiple values. However, they have some key differences in the way they work. Some of these differences are:
1. **Python and C**: Arrays exist in C or C++ language. The concept of Array does not exist in Python, rather we have this similar Data Structure known as a List.
2. **Memory Allocation**: Arrays are stored in contiguous memory locations. Lists are stored in non-contiguos memory locations. However a List takes uo more storage in the memory, whereas an Array is efficient in terms of memory usage.
3. **Dynamic Memory Allocation**: Arrays are fixed in size. Lists are dynamic in size.
4. **Nature**: Arrays are homogenous in nature, i.e. they can only store variables or values of the same datatypes. Whereas, Lists are homogenous in nature, implying that they can store variables and values of any datatype. <br>

We can also check the memory locations (address) of the data stored in a list using `print(id(L))`, where `id()` prints the memory address of the parametered object. <br>

## How Python Lists work internally?
Lists in Python are implemented using **Referential Array** or **dynamic arrays** (also called resizable arrays) in the backend. This implementaion detail is somewhat abstracted away, so users generally interact with the list simply. This is explained below in details.
1. Dynamic Arrays:
  - A Python list is a **contiguous block of memory** that stores **references** (or **pointers**) to the objects it contains. The elements of the list themselves are not stored directly; instead, the list stores memory addresses (or references) to the objects, which might be located elsewhere in memory.
  - These references point to the actual objects that are stored in memory. So when you say `a = [1, 2, 3]`, the list a itself is stored in a contiguous block, but it holds references to the objects `1`, `2`, and `3` (not the integers themselves).
2. Resizing the Array:
  - When the list grows in size, Python dynamically resizes the underlying array to accommodate more elements. This resizing often involves allocating a new, larger block of memory and copying the references from the old array to the new one. This resizing is done in a way that amortizes the cost of growing the list.
3. Array of Pointers:
  - Internally, each element in the list is a pointer (or reference) to the actual object it represents. For example, when you add an integer to a list, the list stores a reference to the integer object, not the integer value itself.
4. Memory Allocation Strategy:
  - Python lists typically over-allocate space (allocate more than needed) to optimize performance when appending new elements, so they don't have to resize the list every time an element is added. This is done to minimize the overhead of memory reallocation.

## Advantages of using Dynamic Arrays
1. **Efficient Random Access**: Since Python lists are backed by dynamic arrays, they offer **O(1)** time complexity for accessing elements by index. This is because accessing an element in a contiguous block of memory is a fast operation.
2. **Amortized Constant Time for Appends**: Adding an element to the end of a list is generally an **O(1)** operation on average, even though resizing the array when it runs out of space may take **O(n)** time. However, because the list grows exponentially (usually doubling in size), the average time per append is constant over many operations.

```python
# the mentioned facts above can be checked via this example
L = [1,2,3]

print(id(L))
print(id(L[0]))
print(id(L[1]))
print(id(L[2]))
print(id(1))
print(id(2))
print(id(3))
```

## Creating List

```python
# Empty
print([])
# 1D -> Homo
print([1,2,3,4,5])
# 2D
print([1,2,3,[4,5]])
# 3D
print([[[1,2],[3,4]],[[5,6],[7,8]]])
# Hetrogenous
print([1,True,5.6,5+6j,'Hello'])
# Using Type conversion
print(list('hello'))
```

## Accessing Items from a List

```python
# Indexing
L = [[[1,2],[3,4]],[[5,6],[7,8]]]
#positive
print(L[0][0][1])

# Slicing
L = [1,2,3,4,5,6]

print(L[::-1])
```

## Adding Items to a List

```python
# append (for adding a single item)
L = [1,2,3,4,5]
L.append(True)
print(L)
```

```python
# extend (for adding multiple items)
L = [1,2,3,4,5]
L.extend([6,7,8])
print(L)
```

```python
# using append for multiple items adds the items to the list
# in form of another list, making it a nested list. Thus adding a single item
L = [1,2,3,4,5]
L.append([6,7,8])
print(L)
```

```python
L = [1,2,3,4,5]
L.extend('delhi')
print(L)
```

```python
# insert (used for adding an item at some specific index)
# the insert() takes two parameters, first if the index value where 
# an object has to be inserted, second is the object itself
L = [1,2,3,4,5]

L.insert(1,100)
print(L)
```

# Editing items in a List
Since list is a Mutable Data Type

```python
L = [1,2,3,4,5]

# editing with indexing
L[-1] = 500

# editing with slicing
L[1:4] = [200,300,400]

print(L)
```

## Deleting items from a List
Has 4 methods to perform deletion, `del`, `remove`, `pop`, and `clear`.

```python
# del
L = [1,2,3,4,5]

# indexing
del L[-1]

# slicing
del L[1:3]
print(L)

# deleting a whole list
del L
L
```

```python
# remove
# used when the index value of the item to be remove is not known
# finds the element in the list and deletes it
L = [1,2,3,4,5]

L.remove(5)

print(L)
```

```python
# pop
# pop() has -1 as the default index, i.e., by default, it will pop the last value of the list
L = [1,2,3,4,5]

L.pop()

print(L)
```

```python
# clear
# clears all the items of the list
L = [1,2,3,4,5]

L.clear()

print(L)
```

## Operations on Lists

- Arithmetic
- Membership
- Loop

```python
# Arithmetic (+ ,*)

L1 = [1,2,3,4]
L2 = [5,6,7,8]

# Concatenation/Merge
print(L1 + L2)
```

```python
# merges for 3 times
print(L1*3)
```

```python
# membership operators
L1 = [1,2,3,4,5]
L2 = [1,2,3,4,[5,6]]

print(5 not in L1)
print([5,6] in L2)
```

```python
# Loops
L1 = [1,2,3,4,5]
L2 = [1,2,3,4,[5,6]]
L3 = [[[1,2],[3,4]],[[5,6],[7,8]]]

for i in L3:
  print(i)
```

## List Functions

```python
# len/min/max/sorted
L = [2,1,5,7,0]

print(len(L))
print(min(L))
print(max(L))
print(sorted(L,reverse=True))
```

```python
# count
L = [1,2,1,3,4,1,5]
L.count(5)
```

```python
# index
L = [1,2,1,3,4,1,5]
L.index(1)
```

```python
# reverse
L = [2,1,5,7,0]
# permanently reverses the list
L.reverse()
print(L)
```

```python
# sort (vs sorted)
# sorted temporarily sorts the list for either storing sorted list in another list
# or for printing it
# while sort permanently sorts the list
L = [2,1,5,7,0]
print(L)
print(sorted(L))
print(L)
L.sort()
print(L)
```

```python
# copy -> shallow (talking about shallow and deep copy for reference)
L = [2,1,5,7,0]
print(L)
print(id(L))
L1 = L.copy()
print(L1)
print(id(L1))
```

## List Comprehension

List Comprehension provides a concise way of creating lists.

newlist = [expression for item in iterable if condition == True]image.png

Advantages of List Comprehension
- More time-efficient and space-efficient than loops.
- Require fewer lines of code.
- Transforms iterative statement into a formula.

```python
# Add 1 to 10 numbers to a list using the traditional or simple way
L = []

for i in range(1,11):
  L.append(i)

print(L)
```

```python
L = [i for i in range(1,11)]
print(L)
```

```python
# scalar multiplication on a vector
v = [2,3,4]
s = -3
# [-6,-9,-12]

[s*i for i in v]
```

```python
# Add squares
L = [1,2,3,4,5]

[i**2 for i in L]
```

```python
# Print all numbers divisible by 5 in the range of 1 to 50

[i for i in range(1,51) if i%5 == 0]
```

```python
# find languages which start with letter p
languages = ['java','python','php','c','javascript']

[language for language in languages if language.startswith('p')]
```

```python
# Nested if with List Comprehension
basket = ['apple','guava','cherry','banana']
my_fruits = ['apple','kiwi','grapes','banana']

# add new list from my_fruits and items if the fruit exists in basket and also starts with 'a'

[fruit for fruit in my_fruits if fruit in basket if fruit.startswith('a')]
```

```python
# Print a (3,3) matrix using list comprehension -> Nested List comprehension
[[i*j for i in range(1,4)] for j in range(1,4)]
```

```python
# Print a (3,3) matrix using list comprehension -> Nested List comprehension
[[i*j for i in range(1,4)] for j in range(1,4)]
```

```python
# cartesian products -> List comprehension on 2 lists together
L1 = [1,2,3,4]
L2 = [5,6,7,8]

[i*j for i in L1 for j in L2]
```

```python
# Nested if with List Comprehension
basket = ['apple','guava','cherry','banana']
my_fruits = ['apple','kiwi','grapes','banana']

# add new list from my_fruits and items if the fruit exists in basket and also starts with 'a'

[fruit for fruit in my_fruits if fruit in basket if fruit.startswith('a')]
```

## 2 ways to traverse a list

- itemwise
- indexwise

```python
# itemwise
L = [1,2,3,4]

for i in L:
  print(i)
```

```python
# indexwise
L = [1,2,3,4]

for i in range(0,len(L)):
  print(L[i])
```

## Zip

The zip() function returns a zip object, which is an iterator of tuples where the first item in each passed iterator is paired together, and then the second item in each passed iterator are paired together.

If the passed iterators have different lengths, the iterator with the least items decides the length of the new iterator.

```python
# Write a program to add items of 2 lists indexwise

L1 = [1,2,3,4]
L2 = [-1,-2,-3,-4]

list(zip(L1,L2))

[i+j for i,j in zip(L1,L2)]
```
