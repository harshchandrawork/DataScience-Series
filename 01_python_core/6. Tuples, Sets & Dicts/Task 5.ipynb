### `Q1:` Join Tuples if similar initial element
While working with Python tuples, we can have a problem in which we need to perform concatenation of records from the similarity of initial element. This problem can have applications in data domains such as Data Science.

For eg.
```
Input  : test_list = [(5, 6), (5, 7), (5, 8), (6, 10), (7, 13)]
Output : [(5, 6, 7, 8), (6, 10), (7, 13)]
```

````python
test_list = [(5, 6), (5, 7), (5, 8), (6, 10), (7, 13)]
result_list = []
current_tuple = ()
i = 0
for row in test_list:
  if row[0] not in result_list[i]:
    current_tuple += (row[0],) # since int concatenation is not possible in string, concatenating the term as a tuple
  elif row[0] == current_tuple[0]:
    current_tuple += (row[1],) # since int concatenation is not possible in string, concatenating the term as a tuple
  print(current_tuple)
  result_list.append(current_tuple)

print(result_list)
````

### `Q2:` Multiply Adjacent elements (both side) and take sum of right and lest side multiplication result.


For eg.
```
The original tuple : (1, 5, 7, 8, 10)
Resultant tuple after multiplication :

(1*5, 1*5+5*7, 7*5 + 7*8, 8*7 + 8*10, 10*8) -> (5, 40, 91, 136, 80)

output-(5, 40, 91, 136, 80)
```

````python
original_tuple = (1, 5, 7, 8, 10)
result_tuple = ()
for i in range(len(original_tuple)):
  if i == 0:
    prod = original_tuple[i] * original_tuple[i+1]
  elif i == len(original_tuple)-1:
    prod = original_tuple[i] * original_tuple[i-1]
  else:
    prod = original_tuple[i] * original_tuple[i+1] + original_tuple[i] * original_tuple[i-1]
  result_tuple += (prod,)
print(result_tuple)
````

### `Q4`: Count no of tuples, list and set from a list
```
list1 = [{'hi', 'bye'},{'Geeks', 'forGeeks'},('a', 'b'),['hi', 'bye'],['a', 'b']]

```
`Output:`

```
List-2
Set-2
Tuples-1
```

````python
list1 = [{'hi', 'bye'},{'Geeks', 'forGeeks'},('a', 'b'),['hi', 'bye'],['a', 'b']]
tupleCount = setCount = listCount = 0
for object in list1:
  if type(object) is tuple:
    tupleCount += 1
  elif type(object) is list:
    listCount += 1
  elif type(object) is set:
    setCount += 1

print(f"List {listCount} \nSet = {setCount} \nTuples {tupleCount}")
````

### `Q1:` Write a program to find set of common elements in three lists using sets.
```
Input : ar1 = [1, 5, 10, 20, 40, 80]
        ar2 = [6, 7, 20, 80, 100]
        ar3 = [3, 4, 15, 20, 30, 70, 80, 120]

Output : [80, 20]
```

````python
ar1 = [1, 5, 10, 20, 40, 80]
ar2 = [6, 7, 20, 80, 100]
ar3 = [3, 4, 15, 20, 30, 70, 80, 120]
set1, set2, set3 = set(ar1), set(ar2), set(ar3)
union = set1.intersection(set2, set3)
print(union)
````

### `Q1`: Key with maximum unique values

Given a dictionary with values list, extract key whose value has most unique values.

**Example 1:**

Input:

```bash
test_dict = {"CampusX" : [5, 7, 9, 4, 0], "is" : [6, 7, 4, 3, 3], "Best" : [9, 9, 6, 5, 5]}
```

Output:
```bash
CampusX
```

**Example 2:**

Input:
```bash
test_dict = {"CampusX" : [5, 7, 7, 7, 7], "is" : [6, 7, 7, 7], "Best" : [9, 9, 6, 5, 5]}
```

Output:
```bash
Best
```

````python
test_dict = {"CampusX" : [5, 7, 9, 4, 0], "is" : [6, 7, 4, 3, 3], "Best" : [9, 9, 6, 5, 5]}
uniqueValList = []

# for counting the unique occurences of elements in the value lists
for key, value in test_dict.items():
  uniqueValCount = 0
  temp = []
  for char in value:
    if char not in temp:
      uniqueValCount += 1
      temp.append(char)
  # uniqueValList.append({key: uniqueValCount})
  uniqueValList.append(uniqueValCount)
print(uniqueValList)

# extracting keys from the original dictionary
keysList = list(test_dict.keys())
print(keysList)

# associating the keys with the needed values and forming a dictionary
uniqueValDict = dict(zip(keysList, uniqueValList))
print(uniqueValDict)

# searching for the key with the maximum value
keyWithMaxVal = max(uniqueValDict, key = uniqueValDict.get)
print(keyWithMaxVal)
````
