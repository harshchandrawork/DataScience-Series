## What is an Object in Python?
In Python, an object is an instance of a class. It is a collection of data (attributes) and methods (functions) that operate on that data. Everything in Python is an object, and each object belongs to a specific class.

When you create a variable or a data structure like a list, string, or dictionary, Python creates an object in memory for that variable. The class defines the blueprint or structure of the object, while the object itself holds the actual data.

For example, if you create a list:

```python
my_list = [1, 2, 3]
```

Here, `my_list` is an object of the `list` class, and the data stored in it is `[1, 2, 3]`. You can then use methods like `append()` or `remove()` to manipulate that object.

In short, objects are the core components that make up the structure of your Python program. They represent data and the operations that can be performed on it.

A **class** in Python (and in Object-Oriented Programming in general) is essentially a **blueprint** or **template** for creating objects. It defines the structure and behaviors that the objects created from it will have. 

### What is a Class?
A class is like a blueprint for objects. It defines the **attributes** (data) and **methods** (functions) that objects created from the class will possess. However, a class itself doesn't hold data—it's more like a description of what the object will look like and how it will behave when it's created. 

In addition, there are two types of classes: built-in classes and user-defined classes.

### Class as a Blueprint
Just like a blueprint for a building shows how the building will be structured (but doesn't itself represent the actual building), a class defines what properties and actions its objects will have, but it isn't the object itself.

When you create an object from a class (called an **instance**), you are essentially creating a real-world object that follows the template set by the class.

### Example:

```python
class Dog:
    # This is the class definition, which defines the blueprint.
    
    def __init__(self, name, age):
        # Constructor method to initialize the object with attributes
        self.name = name
        self.age = age
    
    def bark(self):
        # Method that represents a behavior of the Dog object
        print(f"{self.name} is barking!")

# Creating instances (objects) from the Dog class
dog1 = Dog("Buddy", 3)  # Object 1
dog2 = Dog("Rex", 5)    # Object 2

# Calling methods on the objects
dog1.bark()  # Output: Buddy is barking!
dog2.bark()  # Output: Rex is barking!
```

### Explanation:
- **Class `Dog`** is the blueprint. It describes that a dog object will have a `name` and an `age`, and that it can `bark`.
- When we create **dog1** and **dog2**, these are actual objects (or instances) based on the `Dog` class. Each dog has its own name and age, which are stored separately for each object.
- The **`__init__`** method is a special method used to initialize the object with the provided data when it is created. It's called the **constructor**.
- **Methods**, like `bark()`, represent behaviors that each dog object can perform.

### Key Points:
- **Class as a Blueprint:** Think of a class as a plan or a set of instructions for creating objects. The class defines the attributes and methods, but it doesn't hold any data until an object is instantiated from it.
- **Objects as Instances:** When you create an object from a class, you are creating an instance that follows the class's blueprint, with its own unique data.

In summary, a class is a blueprint or template for creating objects, and it encapsulates both data (attributes) and the behaviors (methods) that the objects will have.

Here is an example of a `Class`.
- `methods` are the functions defined for all the objects of the class and define the behaviour of the data that will come into the class.
- `attributes` refer to the data that is being used inside the class.

## Object
Object is an instance of the class. 

In Object-Oriented Programming (OOP), **objects** are the actual instances created from a **class**. An **object** represents a real-world entity or concept based on the structure defined by the class. In simpler terms, while a class is a blueprint, an object is the physical manifestation (or instantiation) of that blueprint.

### Objects in a Class
An object in Python is a specific instance of a class. It is created based on the class's structure and has its own set of attributes (data) and can perform behaviors defined by the methods in the class.

### Objects as Instances of a Class
When you create an object from a class, that object is an instance of the class. In other words, the object is created based on the blueprint defined by the class, and it has its own independent set of data.

### Example:

```python
class Car:
    # Class definition (blueprint for a Car)
    
    def __init__(self, make, model, year):
        # This is the constructor method to initialize the object with attributes
        self.make = make
        self.model = model
        self.year = year
        
    def display_info(self):
        # Method to display information about the car
        print(f"{self.year} {self.make} {self.model}")

# Creating objects (instances) of the Car class
car1 = Car("Toyota", "Camry", 2022)
car2 = Car("Honda", "Civic", 2020)

# Calling methods on the objects
car1.display_info()  # Output: 2022 Toyota Camry
car2.display_info()  # Output: 2020 Honda Civic
```

### Explanation:
- **Class `Car`** is the blueprint that defines the attributes (`make`, `model`, `year`) and the method (`display_info`) that all car objects will have.
- **`car1`** and **`car2`** are **objects** or **instances** of the `Car` class. Each of these objects has its own unique data, like `make`, `model`, and `year`.
- When you create `car1` and `car2`, you're instantiating two separate objects (instances) of the `Car` class, each with its own set of values for the attributes.

### Key Points:
- **Objects as Instances:** Objects are instances of a class, meaning that they are individual realizations of the blueprint provided by the class. Each object has its own state (data) and can interact with other objects using the methods defined by the class.
- **Attributes and Methods:** The attributes (data) and methods (functions) defined in the class are shared across all objects created from that class, but each object can hold its own distinct data.

````python
# writing code for making a simple ATM machine
class atm:

  # constructor (a special function inside a class)
  def __init__(self):
    self.pin = ''
    self.balance = 0
    # print("This constructor function will be executed by itsef.")
    self.menu()

  def menu(self):
    user_input = input("""
    Hello, how can i help you?
    1. Press 1 to create pin
    2. Press 2 to change pin
    3. Press 3 to check balance
    4. Press 4 to withdraw
    5. Anything else to exit
    """)
    if user_input == '1':
      # create pin
      self.create_pin()
    elif user_input == '2':
      # change pin
      self.change_pin()
      pass
    elif user_input == '3':
      # check balance
      self.check_balance()
      pass
    elif user_input == '4':
      # withdraw
      self.withdraw()
      pass
    else:
      exit()
  
  def create_pin(self):
    user_pin = input('Enter your pin')
    self.pin = user_pin
    user_balance = int(input("Enter balance"))
    self.balance = user_balance

    print("Pin created successfully")
    self.menu()
  
  def change_pin(self):
    current_pin = input("Enter current pin")

    if current_pin == self.pin:
      # let him change the pin
      new_pin = input("Enter new pin")
      self.pin = new_pin
      print("Pin has been successfully changed")
      self.menu()
    else: 
      print("Cannot let you change the pin")
      self.menu()

  def check_balance(self):
    user_pin = input("Enter your pin")
    if user_pin == self.pin:
      print(f"Your balance is {self.balance}")
    else:
      print("Incorrect pin")
    self.menu()

  def withdraw(self):
    user_pin = input("Enter the pin")
    if user_pin == self.pin:
      # allow to withdraw
      amount = int(input("Enter the amount"))
      if amount <= self.balance:
        self.balance -= amount
        print(f"Withdrawl of amount {amount} is successful; Current balance is: {self.balance}")
        self.menu()
      else:
        print("Insufficient balance")
    else:
      print("Access denied")
    self.menu()
````

### Golden Rule of OOP
- Before defining the `self` keyword, we will need to understand about a Golden Rule of OOP.
- There is a Golden Rule of OOP. An object has 2 things, Class and Objects. 
- Class has all the rules that an object follows. Rules have 2 things, **attributes**, such as `pin`, `balance` or **methods**, such as `check_balance`, `withdraw`, etc.
- The members of the class, *i.e.* **Data** and **Functions** within the class can only be accessed by an **Object** that belongs to the respective class.
- This means, that even the methods defined within a class cannot access another method that is defined within the same class.
- *For example*, in class `atm` in the code earlier, the method `withdraw` cannot call the method `check_balance`.

### `self`
The above Golden Rule is a problem when we have to call another method from inside a method. This problem is solved by the `self` parameter.

*For example,*
```python
def __init__(self):
  self.pin = ''
  self.balance = 0
  self.menu()
```
This whole program uses `self` while calling other entity, such as, `menu`. 

Similarly, using `self.balance` to access the value of the attribute/data `balance` defined in the constructor, from another method.

#### What is `self`?
Finally answering the question, the `self` is the **object itself**. This self is needed because as per the Golden Rule of OOP, a method can only be called by the Object of that class, thus the defualt parameter `self` itself becomes the object and passes itself in each method.

Thus self refers to the object itself. Note that it is not mandatory to keep the name as `self`, rather self is just a naming convention that is used in industry usually.

Important to know that the `self` while being passed as a parameter to a method, refers to the created **object** and not to the class. Thus when creating multiple objects of the same class, each object's `self` will refer to the corresponding object and not to the given class. This concept will be useful while creating a **static variable**.

Note that any variable being defined using `self` will be defined globally for the class. *For example*, any variable `x` being defined within a method in a class, will not be accessed by other methods of the same class. Though, the varible being defined like `self.x`, will be access by the other methods in the same class.
> This is the reason why the variables being defined inside the `__init__` constructor using `self` can be accessed from all the methods across the class.
> Using this feature, we will be able to access all the variables and their values defined within the class, without actually seperately passing them as parameters in the other variables. We may rather simply use.

By creating a variable/attribute followed by self, means that the created attribute will be available throughout the created object to all the methods of the object. Consider the following example for this:

## Magic Methods (a.k.a Dunder Methods)
These are special kind of methods. These methods are always of the form `__methodname__`. One such example of Magic Method is the **constructor** method that we used in the program earlier, defined as `__init__`.

Magic Methods are not needed to be called explicitly, as we saw in the case earlier. There are thousands of Magic Methods in python. Though we will focus on a handful of the Magic Methods.

Given Magic Methods: `__init__` or the **constructor**, `__str__` or the string MM, `__add__`. or the add MM.

---

### `__str__`
The `__str__` method in Python is a special (magic) method that defines how an object is represented as a string. When you use `print()` or `str()` on an object, Python will call the `__str__` method to get a human-readable string representation of that object.

#### Purpose of `__str__`:
- It provides a way to define a meaningful or friendly string representation of an object, making it easier to display or print the object in a way that makes sense to the user.

#### Explanation:
- When `print(person)` is called, Python looks for the `__str__` method in the `Person` class and calls it to get a string representation of the object.
- The `__str__` method returns the string `"Alice is 30 years old"`, which is printed to the console.

#### Key Points:
- **`__str__`** is intended to return a **readable string representation** of the object, typically for end-users or display.
- If you don’t define `__str__`, Python will use the default string representation (usually something like `<__main__.ClassName object at memory_location>`).

In short, `__str__` lets you customize how an object appears when you try to print it or convert it to a string.

---

### `__add__`
The `__add__` method in Python is another special (magic) method that defines how the addition operator (`+`) behaves for objects of a class. It allows you to define custom behavior for adding two objects together.

#### Purpose of `__add__`:
- The `__add__` method allows you to specify how two instances of your class should be added together when using the `+` operator. By overriding this method, you can define what happens when you attempt to add objects of your class.

#### Key Points:
- **`__add__`** allows you to define custom behavior for the `+` operator for your class.
- It is called when you attempt to add two objects of the class, and you can decide how those objects should be combined (e.g., adding attributes, combining values, etc.).
- If you don’t define `__add__`, Python will use the default behavior, which raises a `TypeError` if you try to add objects of the class.

In summary, `__add__` gives you control over how objects of your class behave when they are added together using the `+` operator.

# Creating Our Own Data Type: Fraction

````python
class Fraction:

  # parameterized constructor
  def __init__(self, x, y):
    self.num = x
    self.den = y

  def __str__(self):
    return '{}/{}'.format(self.num, self.den)
````

Suppose, using the above class, we try the following code,
```python
frac1 = Fraction(3, 4)
frac2 = Fraction(1, 2)
print(frac1)
print(frac1 + frac2)
```
We will get an error for the last line of the code:<br>
`TypeError: unsupported operand type(s) for +: 'Fraction' and 'Fraction'`

Why so? Because we have simply not defined the operand '`+`' for the objects of class `Fraction`. Therefore, the class `Fraction` simply does not know what to do in such a case.

````python
frac1 = Fraction(3, 4)
frac2 = Fraction(1, 2)
print(frac1)
print(frac1 + frac2)
````

````python
class Fraction:

  # parameterized constructor
  def __init__(self, x, y):
    self.num = x
    self.den = y

  def __str__(self):
    return '{}/{}'.format(self.num, self.den)

  def __add__(self, other):
    new_num = self.num * other.den + other.num * self.den
    new_den = self.den * other.den

    return '{}/{}'.format(new_num, new_den)

  def __sub__(self, other):
    new_num = self.num * other.den - other.num * self.den
    new_den = self.den * other.den

    return '{}/{}'.format(new_num, new_den)

  def __mul__(self, other):
    new_num = self.num * other.num
    new_den = self.den * other.den

    return '{}/{}'.format(new_num, new_den)

  def __truediv__(self, other):
    new_num = self.num * other.den
    new_den = self.den * other.den

    return '{}/{}'.format(new_num, new_den)

  def convert_to_decimal(self):
    return self.num/self.den
    
````

````python
frac1 = Fraction(3, 4)
frac2 = Fraction(1, 2)
print(frac1)
print(frac1 + frac2)
print(frac1 - frac2)
print(frac1 * frac2)
print(frac1 / frac2)
print(frac1.convert_to_decimal)
frac1.convert_to_decimal
````
