## üîπ Exception Handling in Python  

### **üìå What is Exception Handling?**  
Exception handling is a mechanism that allows a program to **gracefully handle errors** instead of crashing unexpectedly. Python provides a `try-except` block to **catch and handle exceptions**, ensuring the program continues running.

---

### **üîπ Two Stages Where Errors Can Occur**
There are 2 stages where error may happen in a program
- During Compilation -> Syntax Error
- During Execution -> Exceptions
#### **1Ô∏è‚É£ During Compilation ‚Üí Syntax Errors**
- **Syntax errors** occur when the Python interpreter detects incorrect syntax **before** running the program.
- Something in the program is not written according to the program grammar.
- Error is raised by the interpreter/compiler
- These errors prevent the program from even starting execution.
- Example:
  ```python
  print("Hello"   # ‚ùå SyntaxError: Missing closing parenthesis
  ```
### Other examples of syntax error

- Leaving symbols like colon,brackets
- Misspelling a keyword
- Incorrect indentation
- empty if/else/loops/class/functions

Since syntax errors are **detected before execution**, they **cannot be handled** using `try-except`.

#### **2Ô∏è‚É£ During Execution ‚Üí Exceptions**
- **Exceptions** occur **while the program is running**, even if the syntax is correct.
- These are **runtime errors** caused by operations like division by zero, invalid indexing, or type mismatches.
- Example:
  ```python
  x = 10 / 0  # ‚ùå ZeroDivisionError: division by zero
  ```
- Unlike syntax errors, **exceptions can be handled** using `try-except`.

---

### **üîπ Summary**
- **Syntax Errors** ‚Üí Occur during **compilation** (before execution) and **cannot** be handled.
- **Exceptions** ‚Üí Occur during **execution** and **can** be handled using `try-except`.
- **Use `finally`** for cleanup actions (like closing files or releasing resources).  

üõ† Exception handling makes Python programs **robust and user-friendly**, preventing abrupt crashes! üöÄ

## **üîπ Exceptions in Python**  

### **üìå What are Exceptions?**  
Exceptions occur when something goes wrong **during the execution** of a program (runtime). Unlike syntax errors, which occur before execution, exceptions arise when unexpected conditions disrupt the program‚Äôs flow.

### **üìå Key Points About Exceptions**
- **Raised by the Python runtime** when an operation fails.
- **Must be handled immediately** to prevent crashes.

---

## **üîπ Examples of Exceptions**
1. **Memory overflow** ‚Äì The system runs out of memory.
2. **Divide by zero** ‚Äì A mathematical error (`ZeroDivisionError`).
3. **Database connection failure** ‚Äì When a database connection is lost.

---

## **üîπ Understanding Stack Trace**
- The **error message displayed** when an exception occurs is called a **stack trace**.
- It provides details like **where the error happened** and what **caused** it.
- Helps **developers debug efficiently**.

---

## **üîπ Why is Exception Handling Necessary?**
### **1Ô∏è‚É£ Improves User Experience**
- If not handled, errors **crash the program**.
- Exception handling allows for **user-friendly error messages**.

### **2Ô∏è‚É£ Enhances Security**
- Unhandled exceptions can reveal **sensitive system information**.
- A controlled response prevents **exploitation**.

---

## **üîπ Handling Specific Exceptions in Python**
```python
# Catching Specific Exceptions
try:
    m = 5
    f = open('sample1.txt', 'r')  # Trying to open a non-existent file
    print(f.read())
    print(m)
    print(5 / 2)
    L = [1, 2, 3]
    L[100]  # IndexError (out of range)
except FileNotFoundError:
    print('File not found')  # Handles file-related errors
except NameError:
    print('Variable not defined')  # Handles missing variable errors
except ZeroDivisionError:
    print("Can't divide by 0")  # Handles division errors
except Exception as e:
    print(e)  # General exception (handles all other cases)
```

### **üîπ Explanation of the Code**
- **`FileNotFoundError`** ‚Üí Handles file-related issues.
- **`NameError`** ‚Üí Handles undefined variable access.
- **`ZeroDivisionError`** ‚Üí Catches division by zero.
- **`Exception` (General Exception)** ‚Üí This **catches all remaining exceptions**.

---

## **üîπ Why is the General Exception Needed?**
- Some errors may **not fall under specific categories**.
- The general `Exception` block **ensures that all errors are caught**.
- **üö® It should always be the last `except` block** to prevent it from overriding more specific exceptions.

---

## **üîπ Summary**
‚úî Exception handling **prevents crashes** and **improves user experience**.  
‚úî The **stack trace** helps debug errors efficiently.  
‚úî Always **handle specific exceptions first**, then use a **general exception** at the end.  

This ensures a **stable, secure, and user-friendly** program! üöÄ

## Some more examples of Try-Except
### **üîπ Handling Exceptions Using `try-except`**
```python
try:
    x = 10 / 0  # Code that may cause an exception
except ZeroDivisionError:
    print("Error: Cannot divide by zero.")  # Handling the exception
```
#### **üîπ How it Works?**
- The `try` block contains the code that **might raise an exception**.
- If an exception occurs, Python **jumps to the `except` block**, preventing the program from crashing.

---

### **üîπ Handling Multiple Exceptions**
```python
try:
    num = int(input("Enter a number: "))  
    result = 10 / num  
except ZeroDivisionError:
    print("Error: Cannot divide by zero.")
except ValueError:
    print("Error: Invalid input! Please enter a number.")
```
- **Handles multiple exceptions separately.**
- If a **ZeroDivisionError** occurs ‚Üí `except ZeroDivisionError` executes.
- If a **ValueError** occurs (like inputting text instead of a number) ‚Üí `except ValueError` executes.

---

### **üîπ Using `finally` for Cleanup**
The `finally` block **always executes**, even if an exception occurs.
```python
try:
    f = open("test.txt", "r")  # Opening a file
    content = f.read()
except FileNotFoundError:
    print("File not found!")
finally:
    f.close()  # Ensuring the file is closed (cleanup)
```


## **üîπ The `else` Block in Exception Handling**  

### **üìå What is the `else` Block?**  
- The `else` block **executes only if** the `try` block runs **without errors**.  
- If an exception occurs, the `else` block is **skipped**.  

---

### **üîπ How `else` Works in Try-Except**
```python
# else block in exception handling
try:
    f = open('sample1.txt', 'r')  # Attempt to open a file
except FileNotFoundError:
    print('file nai mili')  # Executes if file is not found
except Exception:
    print('kuch to gdbad hai daya')  # Catches other exceptions
else:
    print(f.read())  # Executes if no exceptions occur
```

---

### **üîπ Explanation of the Code**
1. **`try` Block:** Attempts to open `"sample1.txt"`.
2. **`except FileNotFoundError` Block:** Runs if the file is missing.
3. **`except Exception` Block:** Catches any other errors.
4. **`else` Block:** Runs **only if no exceptions** occur in `try`.

---

### **üîπ Why Use the `else` Block?**
‚úî Helps keep the **try block clean** by separating error handling from normal execution.  
‚úî Ensures that code in `else` runs **only when the try block succeeds**.  
‚úî Avoids **accidentally catching errors** with general `except` blocks.  

This makes the code **structured, readable, and error-proof**! üöÄ

## **üîπ The `finally` Block in Exception Handling**  

### **üìå What is the `finally` Block?**  
- The `finally` block **always executes**, regardless of whether an exception was raised or not.  
- It ensures that certain critical code (e.g., resource cleanup, security measures) **always runs**.  
- Even if the `try` block fails and the `except` block is executed, `finally` **still runs**.  

---

### **üîπ How `finally` Works in Try-Except**
```python
# finally block in exception handling
try:
    f = open('sample1.txt', 'r')  # Attempt to open a file
except FileNotFoundError:
    print('file nai mili')  # Executes if file is not found
except Exception:
    print('kuch to gdbad hai daya')  # Catches other exceptions
else:
    print(f.read())  # Executes if no exceptions occur
finally:
    print('mai to print hounga hi')  # This always runs
```

---

### **üîπ Explanation of the Code**
1. **`try` Block:** Attempts to open `"sample1.txt"`.
2. **`except FileNotFoundError` Block:** Runs if the file is missing.
3. **`except Exception` Block:** Catches any other errors.
4. **`else` Block:** Runs **only if no exceptions** occur.
5. **`finally` Block:** **Always runs**, whether an error occurs or not.

---

### **üîπ Why Use the `finally` Block?**
‚úî **Resource Cleanup:** Closing files, database connections, network connections.  
‚úî **Security Measures:** Ensuring sensitive operations are always completed.  
‚úî **Logging:** Recording execution status, whether it failed or succeeded.  
‚úî **Ensuring Execution:** Even if an error or `return` statement is encountered, `finally` **still runs**.

The `finally` block guarantees that crucial code is **always executed**, making programs **more reliable and secure**! üöÄ

## **üîπ `raise` Keyword in Python Exception Handling**  

### **üìå What is `raise` in Python?**  
- The `raise` keyword is used to **manually trigger** an exception in Python.  
- It is useful for handling **invalid conditions**, enforcing **rules**, or stopping **invalid execution**.  
- Unlike `except`, which is used to **catch errors**, `raise` is used to **throw errors**.  

---

### **üîπ How `raise` Works**
```python
raise ZeroDivisionError('aise hi try kar raha hu')  # Manually raising an error
```
- This will instantly stop the program and print:
  ```
  ZeroDivisionError: aise hi try kar raha hu
  ```

üí° **Comparison with Java**  
- `try` in Python = `try` in Java  
- `except` in Python = `catch` in Java  
- `raise` in Python = `throw` in Java  

---

### **üîπ Example: Using `raise` in a Bank Withdrawal System**
```python
class Bank:
    def __init__(self, balance):
        self.balance = balance  # Initialize account balance

    def withdraw(self, amount):
        if amount < 0:  
            raise Exception('amount cannot be -ve')  # Raise exception for negative withdrawal
        if self.balance < amount:  
            raise Exception('no money no honey')  # Raise exception for insufficient balance
        self.balance -= amount  # Deduct amount if valid

obj = Bank(10000)  # Creating an account with ‚Çπ10,000

try:
    obj.withdraw(15000)  # Trying to withdraw ‚Çπ15,000 (more than balance)
except Exception as e:
    print(e)  # Catch and print the exception message
else:
    print(obj.balance)  # If no exception, print updated balance
```

---

### **üîπ Explanation of the Code**
1. **`Bank` Class**: Represents a simple banking system with `withdraw()` functionality.  
2. **`withdraw()` Method**:
   - **Checks if the withdrawal amount is negative** ‚Üí If yes, raises an exception.  
   - **Checks if the balance is insufficient** ‚Üí If yes, raises an exception.  
   - **If valid**, deducts the amount from the balance.  
3. **`try-except` Block**:
   - Calls `withdraw(15000)`, but since the balance is only `10000`, it raises **Exception: "no money no honey"**.
   - The `except` block **catches** the error and prints it.  
4. **`else` Block**:
   - Runs **only if no exceptions occur** and prints the updated balance.  

---

### **üîπ Why Use `raise`?**
‚úî **Enforces Business Logic** (e.g., no negative withdrawals).  
‚úî **Prevents Unexpected Errors** (stops execution before major issues occur).  
‚úî **Ensures Better Debugging** (provides clear error messages).  
‚úî **Used in Custom Exception Handling** (raising user-defined exceptions).  

By using `raise`, we make programs more **secure, predictable, and error-resistant**! üöÄ

## **üîπ Exception Hierarchy in Python**  



### **üìå What is Exception Hierarchy?**
- In Python, **`Exception` is a class itself**, and all built-in exceptions inherit from it.  
- The hierarchy follows this structure:  
  ```
  BaseException
  ‚îú‚îÄ‚îÄ Exception
  ‚îÇ   ‚îú‚îÄ‚îÄ ArithmeticError
  ‚îÇ   ‚îÇ   ‚îú‚îÄ‚îÄ ZeroDivisionError
  ‚îÇ   ‚îú‚îÄ‚îÄ FileNotFoundError
  ‚îÇ   ‚îú‚îÄ‚îÄ ValueError
  ‚îÇ   ‚îú‚îÄ‚îÄ TypeError
  ‚îÇ   ‚îú‚îÄ‚îÄ KeyError
  ‚îÇ   ‚îú‚îÄ‚îÄ ... (other exceptions)
  ```
- **Custom exceptions** can also be created by inheriting from the `Exception` class.

---

## **üîπ Custom Exception in Python**
üí° **Python allows users to define their own exception classes**.  
This is useful when handling **application-specific errors** that don‚Äôt fit standard exceptions.

---

### **üîπ Why Create Custom Exceptions?**
‚úî **Application-Based Error Handling** ‚Üí Define specific errors for your program logic.  
‚úî **Full Control** ‚Üí Decide how errors should be displayed and handled.  
‚úî **More Readable and Meaningful Errors** ‚Üí Instead of generic exceptions, give a clear message.  

---

### **üîπ Example: Custom Exception in a Bank System**
```python
class MyException(Exception):  # Custom exception class inheriting from Exception
    def __init__(self, message):
        print(message)  # Prints the error message when an exception is raised

class Bank:
    def __init__(self, balance):
        self.balance = balance  # Initialize balance

    def withdraw(self, amount):
        if amount < 0:
            raise MyException('amount cannot be -ve')  # Raise custom exception for negative withdrawal
        if self.balance < amount:
            raise MyException('no money no honey')  # Raise custom exception for insufficient balance
        self.balance -= amount  # Deduct balance if valid

obj = Bank(10000)  # Creating an account with ‚Çπ10,000

try:
    obj.withdraw(5000)  # Trying to withdraw ‚Çπ5000
except MyException as e:
    pass  # Exception handled silently
else:
    print(obj.balance)  # If no exception, print updated balance
```

---

### **üîπ Explanation of the Code**
1. **Custom Exception `MyException`**:
   - Inherits from the `Exception` class.
   - Prints an error message when raised.
2. **Bank Class**:
   - Implements withdrawal logic with **custom error handling**.
   - Raises `MyException` if:
     - The amount is negative.
     - The balance is insufficient.
3. **Exception Handling with `try-except`**:
   - If an error occurs, it is caught in the `except` block.
   - The `else` block executes only when no exception occurs.

---

### **üîπ Key Takeaways**
‚úÖ `Exception` is a class, and all errors inherit from it.  
‚úÖ Custom exceptions help in **better error handling** and **application-specific logic**.  
‚úÖ By **creating our own exception classes**, we get more control over program flow and debugging. üöÄ

````python
# Another example of creating a custom error class
# inhereting from the Exception or the BaseException Class
class SecurityError(Exception):

	def __init__(self, message):
		print(message)

	def logout(self):
		print('logout')

class Google:

	def __init__(self, name, email, password, device):
		self.name = name
		self.email = email
		self.password = password
		self.device = device

	def login(self, email, password, device):
		if device != self.device:
			raise SecurityError('ab to tu gya beta')
		if email == self.email and password == self.password:
			print('Welcome')
		else:
			print('login error')

obj = Google('firangi', 'firangi@nicemail.com', '1234', 'android')

try:
	obj.login('firangi@nicemail.com', '1234', 'windows')
except SecurityError as e:
	e.logout()
else:
	print(obj.name)
finally:
	print('database connection closed')
````
