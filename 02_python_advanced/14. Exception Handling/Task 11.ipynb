###`Q-1`: You are given a function definition. There might be several issues on execution of this function. You are asked to do exception handling for diffrent errors that this function goes in to `without altering this function`. And print error text.



Function parameters `l -> list, s -> could be anything`

```
def function(l: list, s, **args):
    last_element = l[-1]
    
    l[int(s)]=10
    any_element = l[int(s)+10]
    l[s]=10
    
    res = sum(l)
    
    p = args['p']
    # print(p)
    return res/last_element * p + any_element

```
Check for different function calls:-

```
function([1,2,1], 12)
function([1,2,1]*9, '1-2')
function([1,'2',1]*9, 12)
function([1,'2',1]*9, 12)
function([1,2,0]*9, 12  )
function([1,2,1]*9, 12, p=None)
function([1,2,0]*9, 12, p=10)
```

````python
def function(l: list, s, **args):
    last_element = l[-1]

    l[int(s)] = 10
    any_element = l[int(s) + 10]
    l[s] = 10

    res = sum(l)

    p = args["p"]
    # print(p)
    return res / last_element * p + any_element

try:
	function([1, 2, 3], 12)
except IndexError as e:
	print(f"{e}")

try:
	function([1,2,1]*9, '1-2')
except ValueError as e:
	print(e)

try:
	# function([1, '2', 1]*9, 12)
	function([1,2,1]*9, '1-2')
except ValueError as e:
	print(e)

try:
	function([1, 2, 0]*9, 12)
except KeyError as e:
	print(e, "not found")

try:
	function([1, 2, 0]*9, 12, p = None)
except ZeroDivisionError as e:
	print(e)

try:
	function([1,2,0]*9, 12, p=10)
except ZeroDivisionError as e:
	print(e)
````

### `Q-5:` Cast vote

Write a program that validate name and age as entered by the user to determine whether the person can cast vote or not. To handle the age, create **InvalidAge** exception and for name, create **InvalidName** exception. The name will be invalid when the string will be empty or name has only one word.

Example 1:

Input:

```bash
Enter the name:               goransh singh
Enter the age: 25
```

Output:

```bash
Goransh Singh  Congratulation !!! You can vote.
```

````python
class Vote:
	def validation(self):
		name = input("Enter the name: ")
		age = int(input("Enter the age: "))
		if age < 18:
			raise InvalidAge("Age is below 18")
		elif age not in range(18, 150):
			raise InvalidAge("Age is invalid")
		elif name == '' or len(name) < 2 or not name.isalpha:
			raise InvalidName("Name is invalid")
		else:
			print(f"{name}, Congratulations! You can vote.")

class InvalidAge(Exception):
	def __init__(self, message):
		print(message)

class InvalidName(Exception):
	def __init__(self, message):
		print(message)

vote_instance = Vote()
vote_instance.validation()
````

## Problem with above code
The problem with the above code is that when we try to raise more than one errors at a time, then only the last error is raised and not the earlier one.

The reason only one error is raised at a time and not the other is that **exceptions in Python are designed to stop the execution of the current block of code** when they are raised. Once an exception is raised, the control flow is immediately transferred to the `except` block (if present) or the program terminates if there is no `except` block to catch it.

In your code, you are raising exceptions inside the `validation` method, but once an exception is raised, the method terminates, and the subsequent lines of code (including any other validations) are **never executed**.

### Example of what happens:

1. The user is prompted to enter a name and age.
2. If the age is below 18, it raises the `InvalidAge` exception.
3. As soon as `InvalidAge` is raised, the rest of the validation (such as checking the name) is **not executed** because the exception interrupts the flow.
4. If the age is valid but the name is invalid, it raises the `InvalidName` exception, and the rest of the code after it (such as printing "Validated") does not run.

### Solution:
If you want to **check all conditions** (for both age and name) and then raise exceptions accordingly, you need to **validate all conditions before raising exceptions**. You can accumulate errors and raise them at the end or after all checks are done.

Here is a corrected version of the code that checks all conditions before raising any exceptions:

### Updated Code with All Validations:

```python
class Vote:
    def validation(self):
        name = input("Enter name: ")
        age = int(input("Enter age: "))
        
        errors = []  # List to collect any errors
        
        # Check age validity
        if age < 18:
            errors.append(InvalidAge("Age is below 18"))
        elif age not in range(18, 150):
            errors.append(InvalidAge("Age is invalid"))
        
        # Check name validity
        if name == '' or len(name) < 2 or not name.isalpha():
            errors.append(InvalidName("Name is invalid"))
        
        # If there are any errors, raise them
        if errors:
            for error in errors:
                print(error)
        else:
            print("Validated")

class InvalidAge(Exception):
    def __init__(self, message):
        self.message = message
        super().__init__(self.message)

class InvalidName(Exception):
    def __init__(self, message):
        self.message = message
        super().__init__(self.message)

# Testing the validation
vote_instance = Vote()
vote_instance.validation()
```

### Explanation of Changes:
1. **Accumulate errors in a list**: We store the exceptions in the `errors` list instead of raising them immediately.
2. **Check all conditions**: We check both age and name conditions, collecting any errors.
3. **Raise all errors after validation**: If there are any errors, we print them all at once.

### Outcome:
Now, if both the age and the name are invalid, the program will print both error messages, instead of stopping after raising just one exception.

For example, if the user enters an invalid age (`15`) and an invalid name (`'J'`), it will print:

```
Age is below 18
Name is invalid
```

````python
class Vote:
    def validation(self):
        name = input("Enter name: ")
        age = int(input("Enter age: "))
        
        errors = []  # List to collect any errors
        
        # Check age validity
        if age < 18:
            errors.append(InvalidAge("Age is below 18"))
        elif age not in range(18, 150):
            errors.append(InvalidAge("Age is invalid"))
        
        # Check name validity
        if name == '' or len(name) < 2 or not name.isalpha():
            errors.append(InvalidName("Name is invalid"))
        
        # If there are any errors, raise them
        if errors:
            for error in errors:
                print(error)
        else:
            print("Validated")

class InvalidAge(Exception):
    def __init__(self, message):
        self.message = message
        super().__init__(self.message)

class InvalidName(Exception):
    def __init__(self, message):
        self.message = message
        super().__init__(self.message)

# Testing the validation
vote_instance = Vote()
vote_instance.validation()
````

### `Q-6`: Write a python function which infinitely prints natural numbers in a single line. Raise the **StopIteration** exception after displaying first 20 numnbers to exit from the program.

````python
def display(n):
	i = 0
	while True:
		try:
			n += 1
			i += 1
			if i == 21:
				raise StopIteration
		except StopIteration:
			break
		else:
			print(n, end = ' ')

display(44)
````
