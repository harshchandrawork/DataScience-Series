## **ğŸ”¹ Namespaces in Python**  

A **namespace** is a mapping between **identifiers (names)** and **objects (values)**.  
In Python, namespaces are essentially **dictionaries** that store variable names as **keys** and their corresponding objects as **values**.

---

### **ğŸ”¹ Types of Namespaces**
1ï¸âƒ£ **Built-in Namespace** â†’ Contains Pythonâ€™s built-in functions like `print()`, `max()`, `len()`, etc.  
2ï¸âƒ£ **Global Namespace** â†’ Variables declared at the **top level** of a script/module.  
3ï¸âƒ£ **Enclosing Namespace** â†’ Variables from an **outer function** (used in nested functions).  
4ï¸âƒ£ **Local Namespace** â†’ Variables declared **inside a function** (only accessible within it).  

---

## **ğŸ”¹ Scope and LEGB Rule**  

### **ğŸ“Œ What is Scope?**
- **Scope** refers to the **region** of a program where a variable is **directly accessible**.  
- Each scope has its own **namespace**, which holds variable names and values.

---

### **ğŸ”¹ LEGB Rule**
ğŸ’¡ **How does Python find a variable?**  
When you reference a variable, Python searches for it in the following order:  

ğŸ”¹ **L** â†’ Local (Inside the function)  
ğŸ”¹ **E** â†’ Enclosing (Outer function, if inside nested function)  
ğŸ”¹ **G** â†’ Global (Defined at the top level)  
ğŸ”¹ **B** â†’ Built-in (Pythonâ€™s built-in functions)  

If the variable is not found in any of these **scopes**, Python raises a **`NameError`**.

## **ğŸ”¹ Local and Global Scope**  
### **ğŸ“Œ Example 1: Local and Global Variables**

**ğŸ“Œ Explanation:**  
- `a` â†’ Global scope

---

### **ğŸ“Œ Example 2: Local and Global with Same Name**
```python

  a = 3  # local variable (does NOT modify global 'a')
  print(a)  # 3 (local scope)

```
ğŸ”¹ **The local `a` inside `temp()` does not affect the global `a`**.

---

### **ğŸ“Œ Example 3: Local Scope Accessing Global Variable**
```python

  print(a)  # 2 (accessing global variable)

```
ğŸ”¹ **A local scope can READ from the global scope but cannot modify it**.

### **ğŸ“Œ Example 4: Attempt to Modify Global Variable Without `global` Keyword**

### **ğŸ“Œ Example 5: Modifying Global Variable Using `global`**

ğŸ”¹ **Using `global` allows modification, but itâ€™s NOT a good practice as it leads to unintended side effects**.

### **ğŸ“Œ Overriding Built-in Functions**
Python allows redefining built-in functions, but this is **not recommended**.


ğŸ”¹ **Avoid using built-in function names for custom functions**.

## **ğŸ”¹ Enclosing Scope (Nested Functions)**
A variable in an **outer function** can be used inside an **inner function**.

```python
def outer():
  def inner():
    print(a)  # Accessing 'a' from the outer scope
  inner()
  print('outer function')

a = 1
outer()
print('main program')
```
ğŸ”¹ **The inner function can access variables from the enclosing function**.

---

### **ğŸ“Œ `nonlocal` Keyword**
If an inner function needs to modify a variable in the **enclosing scope**, we use `nonlocal`.

```python
def outer():
  a = 1
  def inner():
    nonlocal a  # Allows modification of 'a' from enclosing scope
    a += 1
    print('inner', a)
  inner()
  print('outer', a)

outer()
print('main program')
```
ğŸ”¹ **Without `nonlocal`, modifying `a` inside `inner()` would raise an error**.

| **Global**     | Top-level variables | Accessible throughout the program |
| **Built-in**   | Python's predefined functions | Accessible everywhere |

## **ğŸ”¹ Decorators in Python**  

A **decorator** is a function that **takes another function as input, modifies or enhances its behavior, and returns it**.  

**Python supports decorators** because functions are **first-class citizens**, meaning:
âœ… Functions can be **assigned to variables**.  
âœ… Functions can be **passed as arguments**.  
âœ… Functions can **return other functions**.

---

## **ğŸ”¹ Types of Decorators**
1ï¸âƒ£ **Built-in Decorators** â†’ `@staticmethod`, `@classmethod`, `@abstractmethod`, `@property`, etc.  
2ï¸âƒ£ **User-defined Decorators** â†’ Custom decorators created by programmers.

---

## **ğŸ”¹ First-Class Functions in Python**
Since functions are first-class citizens, they can be passed as arguments.

```python
# Python functions as first-class citizens
def modify(func, num):
  return func(num)

def square(num):
  return num**2

print(modify(square, 2))  # Output: 4
```
Here, `modify()` accepts `square` as an argument and applies it to `num`.

---

## **ğŸ”¹ Simple Decorator Example**
```python
def my_decorator(func):
  def wrapper():
    print('***********************')
    func()
    print('***********************')
  return wrapper

def hello():
  print('hello')

def display():
  print('hello son')

a = my_decorator(hello)
a()  # Adds decoration to 'hello'

b = my_decorator(display)
b()  # Adds decoration to 'display'
```
ğŸ”¹ The **inner function `wrapper()` adds extra functionality** before and after calling `func()`.

## **ğŸ”¹ Why Does the Inner Function Remember the Variable `a` Even After `outer()` Ends?**
```python

  a = 5
  def inner():
    print(a)  # Why does 'a' still exist?
  return inner

```
ğŸ”¹ **Closure Property** allows `inner()` to **remember `a` from `outer()`** even after `outer()` finishes execution.  
ğŸ’¡ **Closures retain the environment in which they were created**.

## **ğŸ”¹ Using `@` for Cleaner Syntax**
Instead of manually wrapping functions, we can use the `@decorator_name` syntax.

```python
def my_decorator(func):
  def wrapper():
    print('***********************')
    func()
    print('***********************')
  return wrapper

@my_decorator  # Equivalent to hello = my_decorator(hello)
def hello():
  print('hello')

hello()  # Decorator applied!
```
ğŸ”¹ The `@my_decorator` **automatically wraps `hello()`**, improving readability.

## **ğŸ”¹ A Meaningful Decorator: Measuring Execution Time**
A **useful** decorator is one that measures the **execution time** of a function.

```python
import time

def timer(func):
  def wrapper(*args):
    start = time.time()  # Start time
    func(*args)
    print('time taken by', func.__name__, time.time() - start, 'secs')
  return wrapper

@timer
def hello():
  print('hello world')
  time.sleep(2)  # Simulate delay

@timer
def square(num):
  time.sleep(1)
  print(num**2)

@timer
def power(a, b):
  print(a**b)

hello()       # Measures time for 'hello'
square(2)     # Measures time for 'square'
power(2, 3)   # Measures time for 'power'
```
ğŸ”¹ The `timer` decorator calculates **how long a function takes to execute**.

---

## **ğŸ”¹ Why Use `*args` in `wrapper()`?**
ğŸ’¡ `*args` allows us to pass any number of arguments to the decorated function.  
ğŸ’¡ Without `*args`, the decorator would **fail** for functions with parameters.

```python
def wrapper():  # âŒ Incorrect (no parameters)
  start = time.time()
  func()  # âŒ Error if function requires arguments!
  print('time taken by', func.__name__, time.time() - start, 'secs')
```
ğŸ”¹ The correct way is:
```python
def wrapper(*args):  # âœ… Accepts any number of arguments
  start = time.time()
  func(*args)  # âœ… Works with all function signatures
  print('time taken by', func.__name__, time.time() - start, 'secs')
```

## **ğŸ”¹ Parameterized Decorators**
A decorator can **accept arguments** to customize behavior.

```python
def sanity_check(data_type):
  def outer_wrapper(func):
    def inner_wrapper(*args):
      if type(*args) == data_type:
        func(*args)
      else:
        raise TypeError('Ye datatype nai chalega')
    return inner_wrapper
  return outer_wrapper

@sanity_check(int)
def square(num):
  print(num**2)

@sanity_check(str)
def greet(name):
  print('hello', name)

square(2)  # Works âœ…
greet('Alice')  # Works âœ…
square('Hello')  # âŒ TypeError
```
ğŸ”¹ **How it works?**
- `sanity_check(int)` returns `outer_wrapper`, which **checks the data type** before executing `func()`.
