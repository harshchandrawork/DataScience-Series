## ğŸ“˜ **1. `np.sort()`**
**Purpose**: Returns a **sorted copy** of an array.

### ğŸ“Œ Syntax:
```python
np.sort(array, axis=-1, kind='quicksort')
```

### âœ… Parameters:
- `array`: Input array to be sorted.
- `axis`: Axis along which to sort. Default is `-1` (last axis).
- `kind`: Sorting algorithm. Options: `'quicksort'`, `'mergesort'`, `'heapsort'`, `'stable'`.

### ğŸ§  Notes:
- Does **not** modify the original array.
- Can sort both 1D and 2D arrays.
- Reversing the result using slicing (`[::-1]`) gives descending order.

### ğŸ§ª Example:

```python
a = np.random.randint(1, 100, 15)
print(np.sort(a))           # ascending
print(np.sort(a)[::-1])     # descending

b = np.random.randint(1, 100, 24).reshape(6, 4)
print(np.sort(b, axis=0))   # column-wise sort
print(np.sort(b, axis=1))   # row-wise sort
```

## ğŸ“˜ **3. `np.concatenate()`**
**Purpose**: Joins a sequence of arrays along an existing axis.

### ğŸ“Œ Syntax:
```python
np.concatenate((array1, array2), axis=0)
```

### âœ… Parameters:
- Arrays must have the same shape except in the concatenation axis.

### ğŸ§  Notes:
- Can be used as a substitute for `vstack()` and `hstack()`.
- Unlike `append`, `concatenate` requires shape consistency along non-concatenation axes.

### ğŸ§ª Example:
```python
c = np.arange(6).reshape(2, 3)
d = np.arange(6, 12).reshape(2, 3)

np.concatenate((c, d), axis=0)  # vertical stack
np.concatenate((c, d), axis=1)  # horizontal stack
```

## ğŸ“˜ **6. `np.where()`**
**Purpose**: Conditional filtering or selection.

### ğŸ“Œ Syntax:
```python
np.where(condition, [x, y])
```

### âœ… Parameters:
- `condition`: Boolean expression.
- `x`, `y`: Optional. Elements from `x` are selected where condition is True, and from `y` otherwise.

### ğŸ§  Use Cases:
- Extracting indices.
- Replacing values conditionally (vectorized operations).

### ğŸ§ª Examples:
```python

np.where(a > 50, 0, a)        # Replace all > 50 with 0
np.where(a % 2 == 0, 0, a)    # Replace even values with 0
```

## ğŸ“ˆ **`np.percentile()`**
**Purpose:** Computes the nth **percentile** of data.

### ğŸ“Œ Syntax:
```python
np.percentile(a, q)
```

### ğŸ“ Parameters:
- `a`: input array.
- `q`: percentile value between 0 and 100.

### ğŸ§  What is Percentile?
It indicates the value **below which a given percentage** of observations fall.

### ğŸ“Š Use Case:
Used in outlier detection, descriptive statistics, and normalization.

### âœ… Example:
```python

```

### ğŸ§® Percentile Formula:
If `n` is total number of observations and `P` is the percentile rank:

```
Percentile = (P / 100) * n-th value in sorted data
```

## ğŸ”— **`np.corrcoef()`**
**Purpose:** Returns the **Pearson correlation coefficient matrix**.

### ğŸ“Œ Syntax:
```python
np.corrcoef(x, y)
```

### ğŸ§  What is Correlation?
A measure of linear relationship between two datasets:
- `+1`: strong positive correlation
- `0`: no correlation
- `-1`: strong negative correlation

### âœ… Example:
```python
salary = np.array([20000, 40000, 25000, 35000, 60000])
experience = np.array([1, 3, 2, 4, 2])

np.corrcoef(salary, experience)
```

### ğŸ“ Output:
A 2x2 matrix:
- `corrcoef[0, 1]` or `corrcoef[1, 0]` gives the **correlation coefficient** between salary and experience.

### âœ… `np.isin()`
**Purpose:** Checks whether each element of one array is present in another array.

**Syntax:**
```python
np.isin(element, test_elements)
```

**Parameters:**
- `element`: array-like â€” Input array.
- `test_elements`: array-like â€” Values to test against.
- Returns: Boolean array of the same shape as `element`.

**Use Cases in Data Science:**
- Filtering specific values (e.g., valid categories, outliers, missing value placeholders).

**Example:**
```python
a = np.array([10, 25, 30, 45, 50])
items = [10, 20, 30, 40, 50]
print(np.isin(a, items))  # [ True False  True False  True ]
print(a[np.isin(a, items)])  # [10 30 50]
```

### ğŸ” `np.flip()`
**Purpose:** Reverses the order of elements in an array along a specified axis.

**Syntax:**
```python
np.flip(m, axis=None)
```

**Parameters:**
- `m`: array-like â€” Input array.
- `axis`: int, optional â€” Axis to flip. Default is `None`, which reverses all axes.

**Use Cases:**
- Time series reversal.
- Image processing (mirror flipping).
- Data augmentation in machine learning.

**Example:**
```python
a = np.array([1, 2, 3])

b = np.array([[1, 2], [3, 4]])
np.flip(b, axis=0)  # Flips rows
np.flip(b, axis=1)  # Flips columns
```

## ğŸ§® Set Operations in NumPy

| Function           | Description |
|--------------------|-------------|
| `np.union1d()`     | Union of two arrays. |
| `np.intersect1d()` | Intersection of two arrays. |
| `np.setdiff1d()`   | Elements in one array but not the other. |
| `np.setxor1d()`    | Elements in either array, but not both. |
| `np.in1d()`        | Check whether elements of one array are in another (returns boolean array). |

**Example:**
```python
m = np.array([1, 2, 3, 4, 5])
n = np.array([3, 4, 5, 6, 7])

np.union1d(m, n)       # [1 2 3 4 5 6 7]
np.intersect1d(m, n)   # [3 4 5]
np.setdiff1d(n, m)     # [6 7]
np.setxor1d(m, n)      # [1 2 6 7]
m[np.in1d(m, [1])]     # [1]
```

**Use Cases:**
- Set-based filtering
- Feature comparison
- Removing duplicates or finding unique overlaps
