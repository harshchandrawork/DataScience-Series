# Session on Open Source Contribution

## Introduction to Open Source Contribution

Open Source Software (OSS) refers to software whose source code is publicly available and can be viewed, modified, and distributed by anyone. Contributing to open source projects allows developers to collaborate globally, improve existing tools, and gain practical development experience. It also helps build a public portfolio that can be viewed by recruiters and technical professionals.

In this session, we will understand how to professionally contribute to open source projects while utilizing **Git** and **GitHub** as the primary collaboration tools.

---

## Basic Workflow of Open Source Contribution

The typical workflow for contributing to an open source project follows three major stages:

1. Identification of a project to contribute to
2. Identification of a feature to implement or an issue to fix
3. Making and submitting the changes

Most open source projects are hosted on GitHub as repositories maintained by one or more maintainers. Contributors interact with these repositories using a structured workflow.

---

## Repository Forking and Contribution Process

When contributing to a project hosted on GitHub, contributors generally do not directly modify the original repository. Instead, they create their own copy and propose changes through a controlled process.

The typical process is as follows:

1. The contributor **forks** the repository.

   * Forking creates a personal copy of the original repository within the contributor's GitHub profile.

2. The forked repository is **cloned** to the contributor's local system.

   * This allows the contributor to work on the project locally.

3. Changes are made locally to the cloned repository.

4. The contributor **pushes the changes** to their forked repository using:

```bash
git push
```

5. The contributor then creates a **Pull Request (PR)** to the original repository.

A **Pull Request** is a request sent to the maintainers of the original repository asking them to review and potentially merge the proposed changes into the main codebase.

GitHub automatically detects the differences between the forked repository and the original repository and allows maintainers to review the modifications.

---

## Roles in Open Source Contribution

Both contributors and maintainers have defined responsibilities in the open source workflow.

### Responsibilities of a Contributor

1. Create a fork of the repository
2. Clone the forked repository to the local machine
3. Create a new branch and make the required changes
4. Push the changes to the forked repository
5. Create a Pull Request to the original repository

### Responsibilities of a Maintainer

1. Review the submitted changes
2. Approve and merge the Pull Request if the contribution meets project standards

---

## Contribution Workflow in Small vs Large Projects

In small open source projects, the contribution workflow is usually straightforward and easy to understand.

However, when working with large or mature projects, the process can become more complex due to the scale of the codebase and the number of contributors involved.

The stages of contributing to a large project are typically:

1. **Deciding what to contribute** (the most complex stage)
2. **Coding and submitting the contribution** (comparatively simpler)

Since the coding and submission workflow has already been discussed, the next important step is learning how to identify what to contribute. For this purpose, GitHub provides a very useful feature called **GitHub Issues**.

---

## GitHub Issues

GitHub Issues provide a structured interface for discussing problems, improvements, and new features within a project. They act as a communication channel between contributors and maintainers.

Issues allow contributors to ask questions, propose changes, or discuss potential improvements before starting development.

It is important to note that the term **"issue"** does not necessarily mean a bug. An issue may represent:

* A bug or defect
* A feature request
* A documentation improvement
* A design discussion
* Any change that needs to be addressed in the project

Using issues before starting development helps ensure that the work being done is useful and aligns with the maintainers’ expectations.

---

## Features of GitHub Issues

Each GitHub issue can have several attributes that help organize project work:

* **Labels**

  * Issues can have one or multiple labels applied to them to categorize the type of work (e.g., bug, enhancement, documentation).

* **Assignees**

  * Issues can be assigned to specific contributors responsible for resolving them.

* **Milestones**

  * Issues can be associated with milestones indicating the release version in which they should be resolved.

Another important feature of GitHub Issues is that each issue supports **comment threads**. This allows contributors and maintainers to discuss the problem, propose solutions, and provide feedback.

Additional capabilities of issues include:

* Subscribing to issues to receive notifications about updates or discussions
* Communicating directly with maintainers regarding proposed changes

GitHub Issues also support **Markdown**, allowing contributors to write structured descriptions with formatting, code snippets, and links.

---

## Creating and Managing Issues

Contributors can also create their own issues when they identify improvements or problems within a project.

When creating issues, it is important to:

* Write clear and descriptive issue titles
* Provide sufficient context about the problem or proposed feature
* Include relevant examples or code snippets if necessary

Before creating a new issue, it is recommended to:

* Review existing issues to check if a similar topic already exists
* Continue discussion on an existing issue if it is related
* Create a new issue only if no similar discussion is present

Once the issue is created, contributors should communicate their intention to work on the change with the project maintainers.

---

## Best Practices While Developing a Contribution

When working on a contribution, it is recommended to follow certain development practices.

### Work on a Topic Branch

All development should be done on a dedicated branch.

Best practices include:

* Do not work directly on the `master` or `main` branch
* Create a new **topic branch** for each feature or fix
* Use clear and descriptive branch names

---

### Best Practices for Git Commits

Maintaining high-quality commit history is important for collaborative development.

Recommended practices include:

* Make **frequent, smaller commits** rather than large monolithic commits
* Write **clear and descriptive commit messages**
* Ensure that commit messages explain the purpose of the change
* Update the `README.md` file when documentation needs to be modified

---

## Project Contribution Guidelines

Before contributing to any project, contributors must follow the guidelines provided by the project maintainers.

These guidelines are typically documented in the following files within the repository:

* `CONTRIBUTING.md`
* `CODE_OF_CONDUCT.md`

These documents describe:

* Contribution procedures
* Coding standards
* Communication expectations
* Community behavior guidelines

Adhering to these documents is essential for maintaining professionalism and ensuring that contributions are accepted smoothly.

## Useful Git Commands for Understanding Repository History

When working on an open source repository, it is often useful to examine the history of contributions. Git provides several commands that help analyze the repository’s development history.

### Viewing a Summary of Contributors

```bash
git shortlog
```

This command provides a summary of contributors along with their commit messages.

---

### Viewing Commit Logs Sorted by Number

```bash
git shortlog -n
```

This lists commit messages sorted by the number of commits made by each contributor.

---

### Viewing Contributor Contribution Frequency

```bash
git shortlog -n -s
```

This command displays the names of contributors along with the number of commits they have made.

---

### Viewing Commits by a Specific Author

```bash
git log --author="Author's Name"
```

This command lists all commits made by a specific author.

---

### Searching Commits by Keyword

```bash
git log --grep="concept name"
```

This command searches the commit history and displays commits whose messages contain the specified keyword.

These commands are particularly useful when exploring large repositories, understanding previous work, and identifying relevant parts of the codebase.

## Platforms to Find Open Source Projects

After understanding the process of contributing to open source software, the next important step is identifying **where to find projects** to contribute to. There are several platforms that help developers discover repositories that are open to contributions, especially for beginners.

### GitHub Explore

One of the most direct ways to find open source projects is through the **Explore** section on GitHub.

The Explore page recommends repositories based on a user's browsing activity and interests. Contributors can also manually browse repositories using different sections such as:

* **Topics** – Browse repositories categorized by specific technologies or domains.
* **Trending** – Displays repositories that are gaining popularity within the developer community.
* **Events** – Highlights activity happening across the platform.

Using these sections allows contributors to discover active projects and communities that welcome external contributions.

---

### CodeTriage

CodeTriage is a platform that helps developers discover open issues in GitHub repositories.

Users can:

* Select a programming language of interest
* Receive notifications about open issues from projects written in that language
* Start working on issues that need contributions

This platform helps developers identify tasks that are suitable for contribution and gradually become familiar with real-world codebases.

---

### First Contributions

First Contributions is a project specifically designed for beginners who want to learn the workflow of contributing to open source.

It provides:

* Step-by-step guidance for making the first contribution
* A curated list of beginner-friendly repositories
* Simple tasks that can often be completed within a few hours

This platform is particularly useful for developers who want to understand the contribution process without being overwhelmed by large codebases.

---

### First Timers Only

First Timers Only is another initiative focused on helping developers make their first open source contribution.

Repositories that follow this initiative mark certain issues as **first-timers-only**, which means:

* The issue is intentionally kept simple
* Maintainers provide additional guidance
* The issue is reserved for new contributors

This helps reduce the barrier to entry for beginners.

---

### 24 Pull Requests

24 Pull Requests is a community initiative that encourages developers to submit pull requests to open source projects during December.

The platform:

* Suggests projects that welcome contributions
* Encourages consistent contribution over multiple days
* Promotes collaboration within the open source community

---

## Open Source Contribution Events and Competitions

Several events and competitions encourage developers to participate in open source contribution. These events often provide mentorship, recognition, and networking opportunities.

### Google Summer of Code

Google Summer of Code is one of the most well-known open source programs. In this program:

* Organizations publish open source projects and tasks
* Students and developers apply to work on these projects
* Selected participants contribute under the guidance of mentors

The program aims to encourage new developers to participate in open source development while gaining practical experience.

---

### Hacktoberfest

Hacktoberfest is a global open source event where developers are encouraged to contribute to open source repositories during a specific month of the year.

Participants typically:

* Submit pull requests to open source repositories
* Collaborate with maintainers and contributors
* Receive recognition for their participation

Events like these help developers build confidence and gain experience in collaborative software development.

## Learning Curve in Open Source Contribution

It is completely normal for beginners to initially struggle with open source contribution. Large repositories can be complex, and understanding the structure of the codebase often takes time.

Developing the ability to contribute effectively requires:

* Familiarity with Git and GitHub workflows
* Understanding of project architecture
* Experience in reading and navigating large codebases

With time and consistent practice, contributors gradually become more comfortable with the open source ecosystem.

## Professional Tips for Open Source Contribution

The following best practices can help developers approach open source contribution in a structured and professional manner.

### Start with Smaller Projects

Beginners should start with smaller repositories rather than immediately attempting to contribute to large or complex projects.

Smaller projects typically:

* Have simpler codebases
* Provide quicker feedback
* Allow contributors to understand the workflow more easily

---

### Make Small Contributions

Instead of attempting large feature implementations, contributors should initially focus on smaller tasks such as:

* Fixing minor bugs
* Improving documentation
* Adding small enhancements

Small contributions help build familiarity with the repository and the contribution process.

---

### Use the Project Yourself

One of the most effective ways to contribute to an open source project is to **use the software actively**.

By becoming a regular user of the software, contributors can:

* Identify bugs and usability issues
* Suggest improvements
* Develop a deeper understanding of the project's functionality

This perspective helps contributors act as both **users and testers**, making their contributions more meaningful.

---

### Handling Rejections

Rejections are a normal part of open source contribution.

Maintainers may reject pull requests for various reasons such as:

* Misalignment with project goals
* Incomplete implementation
* Coding standard violations

Additionally, there may be situations where maintainers do not respond immediately due to limited time or project workload.

Developers should treat such situations as learning opportunities rather than discouragement.

---

### Be Patient and Consistent

Becoming an effective open source contributor takes time.

Most experienced contributors develop their expertise through:

* Continuous learning
* Consistent contributions
* Gradual involvement in larger projects

Meaningful contributions to large libraries or frameworks often require months or even years of familiarity with the codebase.

---

### Open Source Contribution is Not Mandatory

There is often a perception—especially on social media—that contributing to open source is mandatory for every developer.

In reality, many professional developers do not actively contribute to open source projects.

The primary purpose of open source contribution should be **learning and skill development**. As a secondary benefit, contributors may also gain:

* Professional visibility
* Networking opportunities
* Portfolio development
* Community recognition

Therefore, developers should approach open source contribution as a **learning journey rather than a competitive requirement**.
