## Introduction to Seaborn

**Seaborn** is a high-level data visualization library built on top of **Matplotlib**, designed to create attractive and informative statistical graphics with minimal code. It integrates seamlessly with **NumPy** and **Pandas**, making it especially powerful for exploratory data analysis (EDA) and statistical modeling workflows.

Seaborn abstracts many low-level Matplotlib operations and provides built-in themes, color palettes, and statistical plotting functions that allow data professionals to focus more on analysis and interpretation rather than plot customization.

From a professional standpoint, Seaborn is widely adopted in data science workflows for:

* Exploratory Data Analysis (EDA)
* Statistical visualization
* Feature understanding
* Quick prototyping of insights
* Presentation-ready plots with minimal effort

---

## Should I Use Seaborn or Plotly Professionally?

A common question among beginners is whether to use Seaborn or Plotly for professional work.

Both are powerful libraries. **Plotly** excels in interactive dashboards, web-based visualization, and business intelligence tools. However, in most professional data science workflows, **Seaborn is generally preferred** for the following reasons:

1. **Industry Adoption in Data Science Pipelines**
   Seaborn is deeply integrated into the Python data ecosystem and is widely used in research environments, ML workflows, and enterprise analytics teams.

2. **Strong Statistical Foundation**
   Seaborn is built specifically for statistical data visualization. Many of its plots include built-in statistical aggregations (confidence intervals, regression fits, KDE estimation).

3. **Cleaner Syntax for EDA**
   For quick exploratory analysis in Jupyter notebooks, Seaborn is often more concise and readable.

4. **Better Integration with Pandas DataFrames**
   Seaborn is designed to work directly with DataFrames, supporting column-based plotting without manual array extraction.

5. **Enterprise Familiarity**
   In recent years, Seaborn has gained popularity among enterprises because:

   * It aligns well with machine learning workflows.
   * It integrates smoothly with Matplotlib for customization.
   * It avoids heavy browser-based rendering.
   * It is stable and widely supported in production ML pipelines.

This does not mean Plotly is inferior. Plotly is highly recommended for:

* Interactive dashboards
* Business reporting
* Web deployment
* Stakeholder presentations requiring dynamic interaction

However, for core data science tasks, Seaborn is often the default choice.

## Seaborn Roadmap

### A Common Beginner Confusion

One common confusion when starting Seaborn is that:

* There exist similar functions that create similar types of graphs.
* Two differently named functions sometimes generate the same category of visualization.
* The same graph can be created using different function types.

To understand this properly, we must recall how Matplotlib works.

---

### Understanding the Matplotlib Hierarchy

In Matplotlib, every plot follows an object hierarchy:

* **Figure (`fig`)**
  The entire plotting area.

* **Axes (`ax`)**
  The actual plotting region inside the figure.

* **Plot Elements**
  Lines, ticks, markers, labels, legends — all are objects inside the axes.

There can be:

* One figure containing multiple axes.
* Each axis containing multiple graphical objects.

This establishes a hierarchy:

```
Figure
  └── Axes
        ├── Lines
        ├── Ticks
        ├── Labels
        └── Other graphical objects
```

Understanding this hierarchy is essential because Seaborn builds on top of it.

---

## Types of Functions in Seaborn

Seaborn defines two major types of plotting functions:

### 1. Figure-Level Functions

* Create their own figure.
* Manage subplots internally.
* Often return a grid-like object.
* Suitable for multi-plot visualizations.

Example: `relplot()`, `displot()`, `catplot()`

Advantages:

* Automatically handles faceting.
* Ideal for multi-dimensional visualization.

Disadvantages:

* Less flexible for fine-grained customization.

---

### 2. Axes-Level Functions

* Plot directly onto a specific Matplotlib axis.
* Require manual subplot management.
* Return a Matplotlib Axes object.

Example: `scatterplot()`, `lineplot()`, `histplot()`

Advantages:

* More control.
* Easier to integrate into custom subplot layouts.

Disadvantages:

* Requires manual handling of figure and axes.

Important Note:
The same type of graph can often be created using both figure-level and axes-level functions. The choice depends on the use case and required customization.

## Limitations of Seaborn

Despite its strengths, Seaborn has certain limitations:

* No direct support for Pie Charts.

* Very fine-grained customization still depends on Matplotlib.

Therefore, even when using Seaborn, **Matplotlib knowledge remains essential**. In professional environments, Seaborn and Matplotlib are often used together.

### 1. Relational Plot

Relational plots are used to visualize the **statistical relationship between two or more variables**.

This category primarily falls under **bivariate analysis**, though it can easily extend to multivariate analysis using additional semantic encodings such as `hue`, `style`, and `size`.

Plots under this section:

* Scatter Plot

## Scatter Plot

A scatter plot is used to visualize the relationship between two numerical variables. It helps identify:

* Correlation (positive, negative, none)
* Clusters
* Outliers
* Distribution patterns
* Potential trends

---

### Creating Scatter Plot Using Axes-Level Function

```python
# scatter plot -> axes level function
sns.scatterplot(data=tips, x='total_bill', y='tip')
```

* Returns a Matplotlib `Axes` object.
* Requires manual subplot handling if combining multiple plots.
* Offers fine-grained control.

---

### Creating Scatter Plot Using Figure-Level Function

```python
# relplot -> figure level function
sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter')
```

* `relplot()` is a **figure-level function**.
* Internally manages the `Figure`.
* Returns a `FacetGrid` object.
* Designed for multi-plot visualizations and faceting.

---

### Why Seaborn Recommends Figure-Level Functions

Seaborn documentation generally encourages the use of **figure-level functions** because:

1. They follow a more structured and object-oriented design.
2. They internally manage figure creation.
3. They are more scalable for multi-dimensional visualization.
4. They easily support faceting.
5. They use a generalized syntax.

For example:

* Scatter plot → `kind='scatter'`
* Line plot → `kind='line'`

The overall structure of the code remains the same. Only the `kind` parameter changes. This makes figure-level functions more generalized and reusable in larger analytical pipelines.

However, axes-level functions remain important for:

* Custom subplot grids
* Fine-tuned control
* Integration into existing Matplotlib layouts

---

### Customizing Scatter Plot (Multivariate Representation)

```python
# scatter plot -> axes level function
sns.scatterplot(data=tips, x='total_bill', y='tip',
                hue='sex', style='time', size='size')

# relplot -> figure level function
sns.relplot(data=tips, x='total_bill', y='tip',
            kind='scatter',
            hue='sex', style='time', size='size')
```

Here, we are encoding:

* `x` → total_bill
* `y` → tip
* `hue` → sex
* `style` → time
* `size` → size

This means we represented **5 out of 7 columns** from the dataset with minimal code.

This highlights one of Seaborn’s strongest advantages:

* Efficient multivariate visualization
* Clean syntax
* Professional visual output

## Line Plot

A line plot is closely related to a scatter plot.

The key difference:

* Scatter plot → shows individual data points.
* Line plot → connects data points, typically used when there is a **time component or ordered sequence**.

Line plots are commonly used in:

* Time-series analysis
* Trend visualization
* Performance tracking
* Forecast visualization

---

### Simple Line Plot (Axes-Level)

```python
# axes level function
sns.lineplot(data=india_df, x='year', y='lifeExp')
```

---

### Line Plot Using Figure-Level Function

```python
# figure level function relplot
sns.relplot(data=india_df, x='year', y='lifeExp', kind='line')
```

Again, notice that the only change from scatter plot is:

```python
kind='line'
```

This demonstrates the generalized structure of figure-level functions.

---

### Preparing Data for Multi-Line Plot

### Multi-Line Plot (Figure-Level)

```python
             size='continent')
```

Here we are encoding additional variables:

* `hue` → country
* `style` → continent
* `size` → continent

This allows multivariate time-series analysis in a single plot.

## Facet Plot

Facet plots allow us to create **multiple subplots based on categorical splits**.

They are extremely useful when:

* Comparing subgroups
* Performing segmented analysis
* Visualizing interactions between categories

Important:
Facet functionality is available only in **figure-level functions**, because:

* Figure-level functions control the entire figure.
* Axes-level functions only control individual axes.

---

### Facet Plot Using Columns

```python
# facet plot on cols
sns.relplot(data=tips, x='total_bill', y='tip',
            kind='scatter', col='sex')
```

Here:

* Each column represents a different value of `sex`.

---

### Facet Plot Using Rows

```python
# facet plot on rows
sns.relplot(data=tips, x='total_bill', y='tip',
            kind='scatter', row='sex')
```

Each row represents a category of `sex`.

---

### Facet Grid Using Rows and Columns

```python
# facet plot on both rows and cols
sns.relplot(data=tips, x='total_bill', y='tip',
            kind='scatter',
            col='smoker', row='sex')
```

This creates a grid layout:

* Rows → sex
* Columns → smoker

Facet plots help visualize **an additional column of information** by distributing data across multiple subplots within the same figure.

---

### Using `col_wrap` for Large Facet Grids

When the number of facets becomes large, `col_wrap` helps limit the number of columns per row.

```python
# col wrap
sns.relplot(data=gap, x='lifeExp', y='gdpPercap',
            kind='scatter',
            col='year', col_wrap=3)
```

* `col='year'` creates one subplot per year.
* `col_wrap=3` ensures that only 3 plots appear per row.
* Remaining plots automatically move to the next row.

This keeps the visualization structured and readable.
