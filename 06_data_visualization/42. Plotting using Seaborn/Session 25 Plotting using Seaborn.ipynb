## Seaborn Roadmap

### A Common Beginner Confusion

One common confusion when starting Seaborn is that:

* There exist similar functions that create similar types of graphs.
* Two differently named functions sometimes generate the same category of visualization.
* The same graph can be created using different function types.

To understand this properly, we must recall how Matplotlib works.

---

### Understanding the Matplotlib Hierarchy

In Matplotlib, every plot follows an object hierarchy:

* **Figure (`fig`)**
  The entire plotting area.

* **Axes (`ax`)**
  The actual plotting region inside the figure.

* **Plot Elements**
  Lines, ticks, markers, labels, legends — all are objects inside the axes.

There can be:

* One figure containing multiple axes.
* Each axis containing multiple graphical objects.

This establishes a hierarchy:

```
Figure
  └── Axes
        ├── Lines
        ├── Ticks
        ├── Labels
        └── Other graphical objects
```

Understanding this hierarchy is essential because Seaborn builds on top of it.

---

## Types of Functions in Seaborn

Seaborn defines two major types of plotting functions:

### 1. Figure-Level Functions

* Create their own figure.
* Manage subplots internally.
* Often return a grid-like object.
* Suitable for multi-plot visualizations.

Example: `relplot()`, `displot()`, `catplot()`

Advantages:

* Automatically handles faceting.
* Ideal for multi-dimensional visualization.

Disadvantages:

* Less flexible for fine-grained customization.

---

### 2. Axes-Level Functions

* Plot directly onto a specific Matplotlib axis.
* Require manual subplot management.
* Return a Matplotlib Axes object.

Example: `scatterplot()`, `lineplot()`, `histplot()`

Advantages:

* More control.
* Easier to integrate into custom subplot layouts.

Disadvantages:

* Requires manual handling of figure and axes.

Important Note:
The same type of graph can often be created using both figure-level and axes-level functions. The choice depends on the use case and required customization.

## 1. Relational Plot

Relational plots are used to visualize the **statistical relationship between two or more variables**.

This category primarily falls under **bivariate analysis**, though it can easily extend to multivariate analysis using additional semantic encodings such as `hue`, `style`, and `size`.

Plots under this section:

* Scatter Plot
* Line Plot

---

### Scatter Plot

A scatter plot is used to visualize the relationship between two numerical variables. It helps identify:

* Correlation (positive, negative, none)
* Clusters
* Outliers
* Distribution patterns
* Potential trends

---

#### Creating Scatter Plot Using Axes-Level Function

```python
# scatter plot -> axes level function
sns.scatterplot(data=tips, x='total_bill', y='tip')
```

* Returns a Matplotlib `Axes` object.
* Requires manual subplot handling if combining multiple plots.
* Offers fine-grained control.

---

#### Creating Scatter Plot Using Figure-Level Function

```python
# relplot -> figure level function
sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter')
```

* `relplot()` is a **figure-level function**.
* Internally manages the `Figure`.
* Returns a `FacetGrid` object.
* Designed for multi-plot visualizations and faceting.

---

#### Why Seaborn Recommends Figure-Level Functions

Seaborn documentation generally encourages the use of **figure-level functions** because:

1. They follow a more structured and object-oriented design.
2. They internally manage figure creation.
3. They are more scalable for multi-dimensional visualization.
4. They easily support faceting.
5. They use a generalized syntax.

For example:

* Scatter plot → `kind='scatter'`
* Line plot → `kind='line'`

The overall structure of the code remains the same. Only the `kind` parameter changes. This makes figure-level functions more generalized and reusable in larger analytical pipelines.

However, axes-level functions remain important for:

* Custom subplot grids
* Fine-tuned control
* Integration into existing Matplotlib layouts

---

#### Customizing Scatter Plot (Multivariate Representation)

```python
# scatter plot -> axes level function
sns.scatterplot(data=tips, x='total_bill', y='tip',
                hue='sex', style='time', size='size')

# relplot -> figure level function
sns.relplot(data=tips, x='total_bill', y='tip',
            kind='scatter',
            hue='sex', style='time', size='size')
```

Here, we are encoding:

* `x` → total_bill
* `y` → tip
* `hue` → sex
* `style` → time
* `size` → size

This means we represented **5 out of 7 columns** from the dataset with minimal code.

This highlights one of Seaborn’s strongest advantages:

* Efficient multivariate visualization
* Clean syntax
* Professional visual output

---

### Facet Plot

Facet plots allow us to create **multiple subplots based on categorical splits**.

They are extremely useful when:

* Comparing subgroups
* Performing segmented analysis
* Visualizing interactions between categories

Important:
Facet functionality is available only in **figure-level functions**, because:

* Figure-level functions control the entire figure.
* Axes-level functions only control individual axes.

---

#### Facet Plot Using Columns

```python
# facet plot on cols
sns.relplot(data=tips, x='total_bill', y='tip',
            kind='scatter', col='sex')
```

Here:

* Each column represents a different value of `sex`.

---

#### Facet Plot Using Rows

```python
# facet plot on rows
sns.relplot(data=tips, x='total_bill', y='tip',
            kind='scatter', row='sex')
```

Each row represents a category of `sex`.

---

#### Facet Grid Using Rows and Columns

```python
# facet plot on both rows and cols
sns.relplot(data=tips, x='total_bill', y='tip',
            kind='scatter',
            col='smoker', row='sex')
```

This creates a grid layout:

* Rows → sex
* Columns → smoker

Facet plots help visualize **an additional column of information** by distributing data across multiple subplots within the same figure.

---

#### Using `col_wrap` for Large Facet Grids

When the number of facets becomes large, `col_wrap` helps limit the number of columns per row.

```python

            kind='scatter',
            col='year', col_wrap=3)
```

* `col='year'` creates one subplot per year.
* `col_wrap=3` ensures that only 3 plots appear per row.
* Remaining plots automatically move to the next row.

This keeps the visualization structured and readable.

## Histplot (Histogram)

A histogram:

* Divides the x-axis into **bins of equal width**
* Counts how many observations fall into each bin
* Uses the count (or density) to determine bar height

In Seaborn:

* **Axes-level function** → `histplot()`
* **Figure-level function** → `displot(kind='hist')`

## KDE Plot (Kernel Density Estimation)

### What is a KDE Plot?

A KDE (Kernel Density Estimation) plot is a **smoothed version of a histogram**.

Instead of discrete bins, it:

* Places a smooth curve over the data
* Uses a kernel function to estimate probability density
* Produces a continuous distribution curve

---

### Meaning of the Name

* **Kernel** → A mathematical function used to smooth data
* **Density** → Refers to probability density
* **Estimation** → Because it estimates the true underlying distribution

---

### What is a Gaussian Kernel?

A Gaussian Kernel is based on the **normal distribution curve**.

For each data point:

* A small bell-shaped curve (Gaussian) is placed.
* All curves are summed together.
* The result is a smooth continuous density curve.

---

### What Does the Y-Axis Represent?

The Y-axis represents **probability density**, not raw counts.

Important clarification:

* Values on Y-axis are not probabilities.
* The **area under the curve equals 1**.
* It represents density per unit on the x-axis.

---

### KDE Plot Using Axes-Level Function

```python
# KDE Plot -> axes level function
sns.kdeplot(data=tips, x='total_bill')
```

---

### KDE Plot Using Figure-Level Function

```python
# KDE Plot -> figure level function
sns.displot(data=tips, x='total_bill', kind='kde')
```

---

### KDE Plot with Hue and Fill

```python
# KDE Plot -> figure level function
sns.displot(data=tips, x='total_bill', kind='kde',
            hue='sex', fill=True)
```

* `fill=True` fills the area under the curve.
* Purely aesthetic preference.

---

### KDE vs Histogram

#### Advantages of Histogram

* Shows actual counts
* Can show empty bins
* Less smoothing distortion
* Easier to interpret for discrete datasets

#### Disadvantages of Histogram

* Sensitive to bin size
* Can appear noisy

---

#### Advantages of KDE

* Smooth representation
* Better for detecting distribution shape
* Clearly shows modality (single peak, bimodal, multimodal)

#### Disadvantages of KDE

* Can oversmooth data
* Can hide gaps in data
* Sensitive to bandwidth selection
* Not ideal for small datasets

---

### Key Concept: Continuity

* KDE plots are **continuous**.
* Histograms are **discrete**.
* Histograms can show empty bins.
* KDE will always show smooth movement across values.

---

## Rug Plot

A Rug Plot:

* Displays individual observations as small ticks.
* Plots ticks along the axis.
* Represents marginal distribution.
* Is usually used alongside another plot (e.g., KDE or Histogram).

It provides granular insight without cluttering the visualization.

---

### Rug Plot with KDE Plot

```python

sns.rugplot(data=tips, x='total_bill')
```

The ticks in the rug plot:

* Are drawn perpendicular to the x-axis.
* Represent individual data points.
* Align under peaks of the KDE curve where data is dense.

This combination helps validate the smooth KDE estimate using raw data positions.

---

## Bivariate Histogram

* Divides the 2D plane into rectangular bins.
* Counts observations in each rectangle.
* Uses color intensity to represent count.

Interpretation:

* Darker or more saturated regions indicate higher density.
* Similar in concept to a heatmap.
* Useful for identifying concentration zones.

---

## Bivariate KDE Plot

A bivariate KDE plot:

* Smooths (x, y) observations using a **2D Gaussian kernel**.
* Produces contour lines representing density levels.

```python

```

---

### How to Interpret a Bivariate KDE Plot

* Contour lines represent levels of equal density.
* Inner contours → Higher density.
* Outer contours → Lower density.
* The center of the densest contour shows where most data points are concentrated.

It helps:

* Identify correlation patterns.
* Detect clusters.
* Visualize joint distribution.
* Observe non-linear relationships.

## 3. Matrix Plots

Matrix plots are used when data is naturally structured in a **grid or tabular format**, where two categorical dimensions intersect and values are represented inside cells.

These plots are especially useful for:

* Correlation analysis
* Time-series comparisons across entities
* Feature similarity
* Clustering and segmentation
* High-density structured data

---

### Heatmap

A heatmap plots **rectangular data as a color-encoded matrix**.

* Each cell represents a value.
* Color intensity encodes magnitude.
* Rows and columns form a grid.

#### Understanding the Grid Structure

When creating a heatmap:

* A grid of `rows × columns` is formed.
* Each intersection represents one value.

For example:

If we have:

* 200 countries
* 10 years

Then the grid will be:

```
200 rows × 10 columns
```

* Rows → Countries
* Columns → Years
* Cell values → Life Expectancy

This structure makes heatmaps extremely powerful for structured comparisons.

---

#### Preparing Data for Heatmap

Since heatmaps require a matrix format, we must reshape data using `pivot()`.

```python
# convert gap wide data to gap long data
gapLong_df = gap.pivot(index='country', columns='year', values='lifeExp')
gapLong_df
```

Here:

* `index='country'` → rows
* `columns='year'` → columns
* `values='lifeExp'` → cell values

---

#### Basic Heatmap

```python
# plotting heatmap and configuring via plt
plt.figure(figsize=(12, 12))
sns.heatmap(data=gapLong_df)
```

Important:

* `heatmap()` is an **axes-level function**.
* It does **not** have a figure-level equivalent.
* Since it is axes-level, customization is done using `plt` (Matplotlib).

---

#### Customizing Heatmap

```python
asia_df = gap[gap['continent'] == 'Asia'] \
    .pivot(index='country', columns='year', values='lifeExp')

plt.figure(figsize=(12, 12))
sns.heatmap(asia_df,
            annot=True,
            linewidths=0.5,
            cmap='summer',
            linecolor='green')
```

Explanation of parameters:

* `annot=True` → Displays numerical values inside cells
* `linewidths=0.5` → Adds spacing between cells
* `cmap='summer'` → Sets color palette
* `linecolor='green'` → Sets border color

Heatmaps are widely used for:

* Correlation matrices
* Feature importance grids
* Performance tracking
* Cross-tabulated summaries

---

### Clustermap

A clustermap is an advanced version of a heatmap that:

* Performs hierarchical clustering.
* Reorders rows and columns based on similarity.
* Displays dendrograms alongside the heatmap.

It helps identify:

* Natural groupings
* Similar patterns
* Hidden structure in high-dimensional data

---

#### Example: Clustermap

```python
# clustermap
iris = px.data.iris()
sns.clustermap(iris.iloc[:, [0, 1, 2, 3]])
```

What happens here:

* The numeric features of the dataset are selected.
* Hierarchical clustering is applied.
* Rows and columns are rearranged based on similarity.
* Dendrograms appear on the top and left.

---

#### How to Interpret Clustermap

* Similar rows cluster together.
* Similar columns cluster together.
* The dendrogram shows hierarchical relationships.
* Shorter branches indicate higher similarity.

Clustermap is particularly useful in:

* Genomics
* Customer segmentation
* Feature similarity analysis
* Pattern recognition tasks

---

### Heatmap vs Clustermap

| Feature                    | Heatmap | Clustermap                       |
| -------------------------- | ------- | -------------------------------- |
| Basic grid visualization   | Yes     | Yes                              |
| Hierarchical clustering    | No      | Yes                              |
| Reordering of rows/columns | No      | Yes                              |
| Dendrograms                | No      | Yes                              |
| Axes-level                 | Yes     | Special (creates its own figure) |

Note:

* `heatmap()` is purely axes-level.
* `clustermap()` internally creates its own figure and cannot be easily embedded inside an existing Matplotlib subplot layout.

---

### Key Takeaways from Matrix Plots

1. Heatmaps represent structured grid data using color encoding.
2. Data must be reshaped into matrix form using `pivot()`.
3. Heatmaps are axes-level functions and require Matplotlib for figure control.
4. Clustermaps add hierarchical clustering and dendrogram visualization.
5. Matrix plots are essential for correlation analysis and pattern detection.

Mastery of matrix plots is crucial for understanding structured relationships in high-dimensional datasets.
