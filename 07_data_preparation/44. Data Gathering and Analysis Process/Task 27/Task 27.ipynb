```python
# simple logic to get 50 cities, while still following the restrictions
import requests, time, os
from dotenv import load_dotenv
from pathlib import Path

# load .env file for API Key
current_dir = Path(__file__).parent
parent_dir = current_dir.parent
file_path = parent_dir / ".env"

load_dotenv(file_path)

API_KEY = os.getenv("GeoDB_API_KEY")

url = "https://wft-geo-db.p.rapidapi.com/v1/geo/cities"

# querystring = {"limit":"50"}

headers = {
	"x-rapidapi-key": API_KEY,
	"x-rapidapi-host": "wft-geo-db.p.rapidapi.com",
	"Content-Type": "application/json"
}

all_cities = []

for offset in range(0, 50, 10):
  querystring = {
    "limit": "10",
    "offset": str(offset)
  }
  response = requests.get(url, headers=headers, params=querystring)
  data = response.json()

  all_cities.extend(data["data"])

  time.sleep(1)

print(len(all_cities))
```

### Final change:
- Statically add country codes of 5 countries, whose code is known to exist in the GeoDB's response. Fetch 500 cities from each of these country, making the row count 500(cities)*5(countries)=2500 rows.

```python
import requests, time, random, pycountry, os
from dotenv import load_dotenv
from pathlib import Path

# load .env file for API Key
current_dir = Path(__file__).parent
parent_dir = current_dir.parent
file_path = parent_dir / ".env"

load_dotenv(file_path)

API_KEY = os.getenv("GeoDB_API_KEY")

url = "https://wft-geo-db.p.rapidapi.com/v1/geo/cities"

# querystring = {"limit":"50"}

API_KEY

headers = {
	"x-rapidapi-key": API_KEY,
	"x-rapidapi-host": "wft-geo-db.p.rapidapi.com",
	"Content-Type": "application/json"
}

# country_codes = [country.alpha_2 for country in pycountry.countries]
country_codes = ['IN', 'US', 'CN', 'RU', 'PK']

all_cities = []

for code in country_codes:
    try:
        # Get total cities in this country
        params = {
            "countryIds": code,
            "limit": "1"
        }

        response = requests.get(url, headers=headers, params=params)
        data = response.json()
        time.sleep(1)

        # Validate metadata
        if "metadata" not in data:
            print(f"[{code}] Skipped (no metadata): {data}")
            continue

        total = data["metadata"].get("totalCount", 0)
        target = min(500, total)

        if total < 500:
            print(f"[{code}] Skipped (not enough cities: {total})")
            continue

        # Fetch up to 500 cities in batches of 10
        for offset in range(0, target, 10):
            params = {
                "countryIds": code,
                "limit": "10",
                "offset": str(offset)
            }

            response = requests.get(url, headers=headers, params=params)
            data = response.json()

            if isinstance(data.get("data"), list):
                all_cities.extend(data["data"])
                print(f"[{code}] Added 10 cities (offset {offset})")
            else:
                print(f"[{code}] Invalid data format: {data}")

            time.sleep(1)  # avoid rate limits

    except Exception as e:
        print(f"[{code}] Error: {e}")
        continue

print(f"\n✅ Total cities collected: {len(all_cities)}")
```
