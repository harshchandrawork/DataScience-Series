## Data Analysis Process (DAP)

The Data Analysis Process (DAP) refers to a structured sequence of steps followed to extract meaningful insights from raw data. Broadly, it consists of three core stages:

* Data Gathering
* Data Cleaning
* Exploratory Data Analysis (EDA)

This session primarily focuses on **Data Gathering**, which forms the foundation of any data-driven workflow.

---

## Data Gathering

Data Gathering involves collecting data from various sources and formats to prepare it for analysis. In real-world scenarios, data is rarely available in a single clean format and may need to be acquired from multiple sources.

### Key Data Sources and Methods

* **File Formats**

  * CSV (Comma-Separated Values)
  * Excel Files
  * Text Files
  * JSON (JavaScript Object Notation)

* **Databases**

  * Relational databases (e.g., MySQL, PostgreSQL)
  * NoSQL databases

* **APIs (Application Programming Interfaces)**

  * Fetching real-time or dynamic data from external services

* **Web Scraping**

  * Extracting data from websites (basic level in this context)

### Data Exporting

Data can also be exported into various formats for storage, sharing, or further processing:

* CSV
* Excel
* JSON
* Databases
* HTML

---

## What is Data Analysis?

Data Analysis is defined as:

> The process of inspecting, cleansing, transforming, and modeling data with the objective of discovering useful information, informing conclusions, and supporting decision-making.

---

## Data Analysis Workflow

A typical data analysis workflow can be represented as:

**Asking Questions → Data Wrangling → Exploratory Data Analysis → Drawing Conclusions → Communicating Results**

---

## 1. Asking Questions

The first step in any analysis is to clearly define the problem through relevant and structured questions. The quality of analysis is directly dependent on the quality of questions asked.

### Examples of Analytical Questions

* Which features are relevant to the analysis?
* Which features can be excluded?
* Are there strong correlations between variables?
* Is data preprocessing required?
* What type of feature engineering is necessary?

There is no fixed set of questions. The ability to frame meaningful questions is a skill developed over time.

### Improving Question Formulation

* **Domain Knowledge**
  Understanding the subject matter helps in identifying impactful questions.

* **Experience**
  Repeated exposure to datasets builds intuition for identifying patterns and gaps.

---

## 2. Data Wrangling (Data Munging)

Data Wrangling is the process of transforming raw data into a structured and usable format for analysis. It consists of three major steps:

### 2.1 Gathering Data

Data may be collected from:

* Pre-existing datasets (e.g., CSV files)
* APIs
* Web scraping
* Databases

In practical scenarios, data gathering often requires combining multiple sources.

---

### 2.2 Assessing Data

This step focuses on understanding the structure and quality of the dataset.

#### Common Assessment Checks

* Dataset dimensions (number of rows and columns)
* Data types of each column (`info()`)
* Presence of missing values
* Duplicate entries (`is_unique`, `duplicated()`)
* Memory usage
* Statistical summary (`describe()`)

---

### 2.3 Cleaning Data

Data Cleaning ensures the dataset is accurate, consistent, and ready for analysis.

#### Common Cleaning Operations

* Handling missing values

  * Mean / Median imputation
  * Forward fill (`ffill`) / Backward fill (`bfill`)

* Removing duplicates

  * `drop_duplicates()`

* Correcting data types

  * `astype()`

---

## 3. Exploratory Data Analysis (EDA)

EDA is a critical stage where analysts explore the dataset to uncover patterns, relationships, and anomalies.

It consists of two key activities:

### 3.1 Exploring Data

* Correlation and covariance analysis
* Univariate and multivariate analysis
* Data visualization (graphs and plots)

---

### 3.2 Augmenting Data

Data is modified or enhanced to improve analysis outcomes.

#### Common Augmentation Techniques

* Removing outliers
* Merging datasets
* Creating new features (columns)

---

## 4. Drawing Conclusions

At this stage, insights are derived from the analysis. Conclusions may be:

* Descriptive (based on observed data)
* Predictive (using Machine Learning or Inferential Statistics)

For a Data Analyst role, conclusions are often derived using **Descriptive Statistics**.

### Example Analytical Questions

* Is a player more effective under specific conditions?
* Does a demographic factor influence outcomes?
* Which location has the highest cost trends?

---

## 5. Communicating Results (Data Storytelling)

The final step involves presenting insights in a clear and impactful manner.

### Common Communication Formats

* Reports
* Presentations (PPT/slide decks)
* Blog posts
* Verbal presentations

Effective communication ensures that insights are understandable and actionable for stakeholders.

---

## Key Note

The Data Analysis Process is **not strictly linear**. Steps may:

* Occur in different orders
* Be repeated multiple times
* Evolve based on findings during analysis

This iterative nature makes data analysis both a technical process and a problem-solving discipline.

## Data Importing

Data Importing is the first practical step in the data analysis pipeline, where raw data is loaded into a working environment (typically using libraries such as Pandas). Efficient data importing ensures that datasets are correctly structured and optimized for downstream processing.

---

## Importing CSV and Related File Formats

### 1. Loading from a Local CSV File

The `pd.read_csv()` function is the most commonly used method to load structured data. It supports both local file paths and remote URLs.

```python
pd.read_csv('local_dataset.csv')
```

---

### 2. Loading a CSV File from a URL

Data can also be fetched from remote servers. In some cases, additional headers are required to simulate a browser request.

```python
import requests
from io import StringIO

url = "https://raw.githubusercontent.com/cs109/2014_data/master/countries.csv"
```

**Note:** This approach is useful when direct URL-based loading is restricted.

---

### 3. Loading from a TSV File

TSV (Tab-Separated Values) files use tabs instead of commas as delimiters. This can be handled using the `sep` parameter.

```python
pd.read_csv('movies_titles_metadata.tsv', sep='\t')
```

---

## Important Parameters of `pd.read_csv()`

Understanding these parameters is essential for handling real-world datasets effectively.

---

### `names`

Used to assign column names when the dataset does not include headers.

```python
pd.read_csv(
    'movies_titles_metadata.tsv',
    sep='\t',
    names=['sno', 'name', 'release_year', 'rating', 'votes', 'genres']
)
```

---

### `index_col`

Sets a specific column as the DataFrame index.

```python
pd.read_csv('aug_train.csv', index_col='enrollee_id')
```

---

### `header`

Specifies the row number to be used as column headers.

```python
pd.read_csv('test.csv', header=1)
```

---

### `usecols`

Loads only selected columns, improving performance and memory efficiency.

```python
pd.read_csv(
    'aug_train.csv',
    usecols=['enrollee_id', 'gender', 'education_level']
)
```

---

### `squeeze` (Deprecated in newer versions)

Used to return a Series instead of a DataFrame when reading a single column.

```python
pd.read_csv('aug_train.csv', usecols=['gender'], squeeze=True)
```

---

### `skiprows`

Skips specified rows during data loading.

```python
pd.read_csv('aug_train.csv', skiprows=[2, 5])
```

A function can also be passed to skip rows based on conditions (e.g., every 5th row).

---

### `nrows`

Limits the number of rows read from the dataset.

```python
pd.read_csv('aug_train.csv', nrows=100)
```

---

### `encoding`

Specifies the character encoding of the dataset. This is critical when working with non-UTF-8 data.

```python
pd.read_csv('zomato.csv', encoding='latin-1')
```

---

### Handling Bad Lines

Inconsistent rows (e.g., incorrect number of columns) can cause parsing errors.

```python
pd.read_csv(
    'BX-Books.csv',
    sep=';',
    encoding='latin-1',
    on_bad_lines='skip'  # updated parameter name
)
```

---

### `dtype`

Defines data types at the time of import, improving memory efficiency and performance.

```python
pd.read_csv('aug_train.csv', dtype={'target': int})
```

---

### Handling Dates (`parse_dates`)

Converts date columns from string format to datetime objects.

```python
pd.read_csv('IPL Matches 2008-2020.csv', parse_dates=['date'])
```

#### Combining Multiple Columns into a Date

```python
pd.read_csv('data.csv', parse_dates=[[1, 3]])
```

#### Naming Parsed Date Columns

```python
pd.read_csv('data.csv', parse_dates={'combined_date': [1, 3]})
```

---

### `converters`

Applies custom transformation functions to column values during import.

```python
def rename(name):
    return "RCB" if name == "Royal Challengers Bangalore" else name

pd.read_csv(
    'IPL Matches 2008-2020.csv',
    converters={'team1': rename}
)
```

---

### `na_values`

Specifies additional representations of missing values.

```python
pd.read_csv('aug_train.csv', na_values=['-'])
```

---

### Loading Large Datasets in Chunks

For very large datasets, loading data in chunks helps manage memory usage efficiently.

```python
dfs = pd.read_csv('aug_train.csv', chunksize=5000)

for chunk in dfs:
    print(chunk.shape)
```

## Scalable Data Collection using Pagination

To retrieve the complete dataset, pagination must be handled programmatically.

---

## Optimized Implementation with Retry Logic

The improved implementation uses a **session-based approach with retry handling**, which is more robust and production-ready.

### Key Components Explained

---

### 1. Session with Retry Strategy

```python
from requests.adapters import HTTPAdapter
from urllib3.util.retry import Retry

session = requests.Session()

retries = Retry(
    total=5,
    backoff_factor=0.3,
    status_forcelist=[500, 502, 503, 504]
)

adapter = HTTPAdapter(max_retries=retries)
session.mount("https://", adapter)
```

#### Purpose

* **Session**: Reuses TCP connections, improving performance
* **Retry Strategy**:

  * Automatically retries failed requests
  * Handles server-side errors (e.g., 500, 502)
  * `backoff_factor` introduces delay between retries

---

### 2. Base Configuration

```python
BASE_URL = "https://api.themoviedb.org/3/movie/top_rated"

base_params = {
    "api_key": API_KEY,
    "language": "en-US"
}
```

#### Purpose

* Centralizes API endpoint and parameters
* Avoids redundancy in loop execution

---

### 3. Initial Request to Determine Total Pages

```python
response = session.get(BASE_URL, params={**base_params, "page": 1})
response.raise_for_status()

data = response.json()
total_pages = data.get("total_pages", 1)
```

#### Purpose

* Identifies how many pages need to be fetched
* Ensures robust error handling using `raise_for_status()`

---

### 4. Efficient Data Storage Strategy

```python
all_results = []
```

#### Why This Matters

* Appending to a list is significantly faster than repeatedly appending to a DataFrame
* DataFrame is created only once at the end (memory-efficient)

---

### 5. Pagination Loop

```python
for page in range(1, total_pages + 1):
    params = {**base_params, "page": page}

    try:
        response = session.get(BASE_URL, params=params)
        response.raise_for_status()

        results = response.json().get("results", [])

        for movie in results:
            all_results.append({
                "id": movie.get("id"),
                "title": movie.get("title"),
                "release_date": movie.get("release_date"),
                "vote_average": movie.get("vote_average"),
                "vote_count": movie.get("vote_count"),
                "popularity": movie.get("popularity"),
                "overview": movie.get("overview")
            })

        print(f"Page {page} read")

    except requests.exceptions.RequestException as e:
        print(f"Error on page {page}: {e}")
```

#### Key Highlights

* Iterates through all pages dynamically
* Extracts only relevant fields from each movie object
* Uses `.get()` to safely handle missing keys
* Includes exception handling to prevent pipeline failure

---

### 6. Final DataFrame Creation

```python
full_df = pd.DataFrame(all_results)
```

#### Result

* A consolidated dataset with approximately **10,000 rows and 7 columns**

---

## Important Limitation

Although the API indicates ~10,712 results, in practice:

* Only ~10,000 rows are reliably retrievable
* Beyond this threshold, some pages may not return valid data

This is a known limitation of the TMDb API pagination system.
However, a dataset of ~10,000 records is sufficiently large for most analytical tasks.

````python
import requests
import os
import time
import pandas as pd
from dotenv import load_dotenv

from requests.adapters import HTTPAdapter
from urllib3.util.retry import Retry

load_dotenv()

API_KEY = os.getenv("TMDB_API_KEY")

BASE_URL = "https://api.themoviedb.org/3/movie/top_rated"

# -------------------------------
# Session with Retry Strategy
# -------------------------------
session = requests.Session()

retries = Retry(
    total=5,
    backoff_factor=0.3,
    status_forcelist=[500, 502, 503, 504]
)

adapter = HTTPAdapter(max_retries=retries)
session.mount("https://", adapter)

headers = {
    "Content-Type": "application/json;charset=utf-8"
}

base_params = {
    "api_key": API_KEY,
    "language": "en-US"
}

# -------------------------------
# First request (to get total pages)
# -------------------------------
response = session.get(BASE_URL, headers=headers, params={**base_params, "page": 1})
response.raise_for_status()

data = response.json()
total_pages = data.get("total_pages", 1)

# -------------------------------
# Store data efficiently
# -------------------------------
all_results = []

# -------------------------------
# Loop through pages
# -------------------------------
for page in range(1, total_pages + 1):
    params = {**base_params, "page": page}

    try:
        response = session.get(BASE_URL, headers=headers, params=params)
        response.raise_for_status()

        results = response.json().get("results", [])

        for movie in results:
            all_results.append({
                "id": movie.get("id"),
                "title": movie.get("title"),
                "release_date": movie.get("release_date"),
                "vote_average": movie.get("vote_average"),
                "vote_count": movie.get("vote_count"),
                "popularity": movie.get("popularity"),
                "overview": movie.get("overview")
            })

        print(f"Page {page} read")

        # Optional: respect rate limits
        # time.sleep(0.25)

    except requests.exceptions.RequestException as e:
        print(f"Error on page {page}: {e}")

# -------------------------------
# Convert to DataFrame (once!)
# -------------------------------
full_df = pd.DataFrame(all_results)

print(f"\nReading {total_pages} pages complete!")
````

## Data Extraction using Web Scraping

In many real-world scenarios, data is not readily available through APIs or downloadable files. In such cases, **Web Scraping** becomes an essential technique for data gathering.

Web scraping involves extracting data directly from the HTML structure of web pages.

---

## Real-World Use Case: Company Data Extraction

A relevant example is AmbitionBox, a platform that provides detailed insights about companies, including:

* Company reviews
* Ratings
* Salary insights
* Interview experiences

This platform is widely used by job seekers to research companies before applying. Since it does not provide a public API for structured data access, web scraping becomes the only viable method for data extraction.

---

## Target Data

From each company listing, the goal is to extract the following attributes:

* Name
* Rating
* Review count
* Organization type (Public/Private)
* Headquarters location
* Years of operation
* Number of employees
* Tags
* Company description (“About”)

---

## Initial Request and Access Denied Issue

```python
from bs4 import BeautifulSoup
import requests

requests.get("https://www.ambitionbox.com/list-of-companies?page=1").text
```

### Result

* The request returns an **"Access Denied"** response

### Reason

Websites often implement mechanisms to block automated scraping. One such mechanism involves controlling crawler behavior using a configuration file.

---

## Understanding `robots.txt`

### Definition

A `robots.txt` file is a configuration file placed in a website’s root directory that instructs web crawlers on which parts of the website can or cannot be accessed.

### Key Points

* Used to manage crawler access and server load
* Helps prevent indexing of restricted or duplicate content
* Not a security mechanism, but a guideline for bots

### Common Use Cases

* Blocking directories (e.g., admin panels)
* Restricting access to specific files
* Providing sitemap references
* Controlling AI and search engine crawlers

---

## Bypassing Basic Access Restrictions

To overcome access denial, requests can be modified to mimic a real browser by adding headers.

```python
headers = {
    "User-Agent": "Mozilla/5.0 (Macintosh, Intel Mac OS X 10.14; rv:66.0) Gecko/20100101 Firefox/66.0"
}

requests.get(
    "https://www.ambitionbox.com/list-of-companies?page=1",
    headers=headers
).text
```

### Outcome

* The server now returns the actual HTML source code of the webpage

---

## Introduction to BeautifulSoup

**BeautifulSoup** is a powerful Python library used for parsing HTML and XML documents.

### Key Characteristics

* Converts raw HTML into a structured, navigable tree
* Enables easy extraction of specific elements using tags, classes, and attributes
* Highly flexible and widely used in web scraping workflows

Due to its extensive functionality, it is not practical to memorize all features. Instead, referring to documentation as needed is the recommended approach.

---

## Parsing HTML Content

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(webpage, 'html.parser')

```

### Notes

* `prettify()` formats the HTML for better readability

## Optimized Scraping with Retry Strategy

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
from requests.adapters import HTTPAdapter
from urllib3.util.retry import Retry

BASE_URL = "https://www.ambitionbox.com/list-of-companies?page={}"

session = requests.Session()

retries = Retry(
    total=5,
    backoff_factor=0.3,
    status_forcelist=[500, 502, 503, 504]
)

session.mount("https://", HTTPAdapter(max_retries=retries))

headers = {
    "User-Agent": "Mozilla/5.0",
}

all_results = []

for page in range(1, 500):
    url = BASE_URL.format(page)

    try:
        response = session.get(url, headers=headers)
        response.raise_for_status()

        soup = BeautifulSoup(response.text, "html.parser")
        companies = soup.find_all("div", class_="companyCardWrapper")

        for company in companies:
            name = company.find("h2")
            rating = company.find("div", class_="rating_text rating_text--md")
            rating_count = company.find("span", class_="companyCardWrapper__companyRatingCount")

            all_results.append({
                "name": name.text.strip() if name else None,
                "rating": rating.text.strip() if rating else None,
                "rating_count": rating_count.text.strip() if rating_count else None
            })

        print(f"Page {page} scraped")

    except requests.exceptions.RequestException as e:
        print(f"Error on page {page}: {e}")

final_df = pd.DataFrame(all_results)
```
