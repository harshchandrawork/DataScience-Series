### Types of Unclean Data

Unclean data can broadly be categorized into two types:

---

### 1. Dirty Data (Quality Issues)

Dirty data refers to **content-related problems** that affect the accuracy and reliability of the dataset.

**Common characteristics:**

* Duplicate records
* Missing values
* Corrupted entries
* Inaccurate or inconsistent data

These issues directly impact the correctness of analysis and model performance.

---

### 2. Messy Data (Tidiness Issues)

Messy data refers to **structural problems** that make data difficult to analyze, even if the content itself is correct.

A dataset is considered **tidy** when it satisfies the following principles:

* Each variable forms a column
* Each observation forms a row
* Each observational unit forms a table

---

### Example: Messy vs Tidy Data

**Messy Data Format:**

| Year | China | India | Pakistan |
| ---- | ----- | ----- | -------- |
| 1950 | value | value | value    |
| 1970 | value | value | value    |
| 1980 | value | value | value    |

**Issue:**

* Multiple variables (countries) are stored as columns instead of values in a single column.
* Violates the principle: *Each variable should form a column.*

---

**Tidy Data Format:**

| Year | Country  | Population |
| ---- | -------- | ---------- |
| 1950 | China    | value      |
| 1950 | India    | value      |
| 1950 | Pakistan | value      |

**Why this is better:**

* Each variable (Year, Country, Population) is clearly defined as a column
* Each row represents a single observation
* Enables easier filtering, grouping, and analysis

---

### Key Insight

Data assessing is not about fixing issues—it is about **identifying and understanding them**. A well-assessed dataset makes the cleaning process systematic and efficient.

## 1. Dataset Summary

This dataset represents a **clinical trial involving diabetic patients** aimed at evaluating the effectiveness of an oral insulin drug.

* Total patients: **500**
* Patients enrolled in trial: **350**
* Initial condition:

  * None were using *Novodra* (injectable insulin) or *Auralin* (oral insulin under study) as their primary insulin source
  * All patients exhibited **elevated HbA1c levels**

### Trial Design

1. **Baseline Phase (4 weeks):**

   * All 350 patients were administered *Novodra*
   * Objective: Establish baseline HbA1c levels and insulin dosage

2. **Treatment Phase (24 weeks):**

   * 175 patients switched to *Auralin*
   * 175 patients continued with *Novodra*

3. **Additional Data Collected:**

   * Adverse reactions reported by patients

### Key Consideration

* The initial 4-week period is **insufficient to capture meaningful HbA1c changes**, as HbA1c reflects average blood glucose over approximately 3 months.
* The primary evaluation is based on the **24-week treatment period**.

## Types of Data Assessment

Data assessment can be broadly categorized into two complementary approaches:

### Manual Assessment

Manual assessment involves **visually inspecting the dataset** using tools such as spreadsheets (e.g., Excel, Google Sheets).

* Analysts scroll through the dataset to identify anomalies, inconsistencies, and patterns
* Particularly effective for spotting:

  * Formatting inconsistencies
  * Obvious outliers
  * Human-readable errors (e.g., typos)

**Industry Insight:**
Despite the availability of advanced tools, experienced analysts often rely heavily on manual inspection to build intuition about the data.

---

### Programmatic Assessment

Programmatic assessment uses **code-based techniques**, typically leveraging libraries such as Pandas.

Common functions include:

* `info()` → Understand structure, data types, and missing values
* `describe()` → Generate statistical summaries
* `sample()` → Randomly inspect subsets of data

This approach is essential for:

* Scaling analysis to large datasets
* Systematically identifying hidden issues
* Validating assumptions discovered during manual assessment

---

## Data Quality Dimensions

To systematically evaluate data issues, four key quality dimensions are used:

* **Completeness**
  Determines whether required data is missing

* **Validity**
  Ensures data conforms to defined rules and constraints

  * Examples:

    * Negative height
    * Duplicate `patient_id`

* **Accuracy**
  Checks whether data correctly represents real-world values

  * Example:

    * Weight = 1 kg for an adult (possible, but highly unrealistic)

* **Consistency**
  Ensures uniform representation of data across the dataset

  * Example:

    * "New York" vs "NY"

**Key Distinction:**

* *Invalid data* is impossible
* *Inaccurate data* is possible but implausible

---

## Purpose of Labelling Issues

Assigning each issue a **data quality label** helps in:

* Prioritizing fixes
* Understanding the nature of the problem
* Designing an efficient cleaning strategy

> Labelling is applied only to **Dirty Data (quality issues)** and not to **Messy Data (tidiness issues)**.

---

## Order of Severity

The **Order of Severity** defines the relative impact of different data quality issues:

```
Completeness < Validity < Accuracy < Consistency
```

### Interpretation

* **Completeness (Most Severe)**
  Missing data can completely block analysis or introduce bias

* **Validity**
  Invalid data breaks assumptions and may cause system or computation failures

* **Accuracy**
  Inaccurate data leads to misleading insights, even if technically usable

* **Consistency (Least Severe)**
  Primarily affects readability and processing, but is usually easier to fix

**Key Insight:**
This ordering helps prioritize which issues should be addressed first to minimize downstream errors.

---

## Data Cleaning Order

Based on the severity and dependencies between issues, the recommended sequence for cleaning is:

1. **Quality → Completeness**

   * Handle missing values first

2. **Tidiness (Structural Issues)**

   * Reshape and organize data into a proper format

3. **Quality → Validity**

   * Fix invalid entries (e.g., incorrect formats, impossible values)

4. **Quality → Accuracy**

   * Correct logically inconsistent or suspicious values

5. **Quality → Consistency**

   * Standardize formats and representations

### Why This Order Matters

* Cleaning **out of order** can:

  * Introduce new errors
  * Make transformations harder
  * Lead to inefficient workflows

* For example:

  * Fixing consistency before handling missing values may require repeating work
  * Tidiness must come early because structural changes affect how data is processed

---

````python
patients_df = patients.copy()
treatments_df = treatments.copy()
treatments_cut_df = treatments_cut.copy()
adverse_reactions_df = adverse_reactions.copy()
````

### Define
- replace all missing values of patients_df with "No data"
- subtract `hba1c_start` from `hba1c_end` to get all the `hba1c_change` values.

````python
# code
patients_df = patients_df.fillna("No data")

# test
patients_df.isna().sum()
````

````python
# code
treatments_df['hba1c_change'] = treatments_df['hba1c_start'] - treatments_df['hba1c_end']
treatments_cut_df['hba1c_change'] = treatments_cut_df['hba1c_start'] - treatments_cut_df['hba1c_end']
````

````python
treatments_df.info()
````

````python
treatments_cut_df.info()
````

````python
treatments_df = pd.concat([treatments_df, treatments_cut_df])
treatments_df
````

````python
treatments_df = treatments_df.melt(id_vars=['given_name', 'surname', 'hba1c_start', 'hba1c_end', 'hba1c_change'],
                   var_name='type', value_name='dosage_range')
treatments_df
````

````python
treatments_df = treatments_df[treatments_df['dosage_range'] != '-']
treatments_df
````

````python
treatments_df['dosage_start'] = treatments_df['dosage_range'].str.split('-').str.get(0)
treatments_df['dosage_end'] = treatments_df['dosage_range'].str.split('-').str.get(1)
treatments_df = treatments_df.drop(columns='dosage_range')
treatments_df
````

````python
treatments_df['dosage_start'] = treatments_df['dosage_start'].str.replace('u', '')
treatments_df['dosage_end'] = treatments_df['dosage_end'].str.replace('u', '')
treatments_df
````

````python
treatments_df['dosage_start'] = treatments_df['dosage_start'].astype('int')
treatments_df['dosage_end'] = treatments_df['dosage_end'].astype('int')
````

````python
treatments_df.info()
````

````python
treatments_df
````

````python
treatments_df = treatments_df.merge(adverse_reactions_df, how='left', on=['given_name', 'surname'])
treatments_df
````
