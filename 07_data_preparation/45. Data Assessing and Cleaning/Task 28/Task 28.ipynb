## Dataset Summary

## Issues Identified
### Quality issues
- `name` contains symbols like '#', ' " ', etc
- `author` has prefix "**Writtenby:**" attached to every value
- `author` has name written in Pascal Case
- `author` has names of authors seperated by commas
- `narrator` has prefix "**Narratedby:**" attached to every value
- `narrator` has name written in Pascal Case
- `narrator` has names of narrators seperated by commas
- `time` is not a `datetimeobject`
- `releasedate` is not a `datetimeobject`
- `language` must be a categorical type
- `stars` contains quantitative data stored as a non-numeric type
- `stars` also contains the review count
- `price` is an object instead of float

### **Consistency** (formatting, representation, standardization issues)

These problems arise when the same type of data is not represented uniformly:

* `name` contains symbols like `#`, `"`, etc
* `author` has prefix "**Writtenby:**" attached
* `author` has name written in Pascal Case
* `author` has names separated by commas
* `narrator` has prefix "**Narratedby:**" attached
* `narrator` has name written in Pascal Case
* `narrator` has names separated by commas

These are all about inconsistent formatting or representation rather than correctness.

---

### **Validity** (data type / rule violations)

These issues break predefined rules, schemas, or expected data types:

* `time` is not a datetime object
* `releasedate` is not a datetime object
* `language` must be a categorical type
* `stars` contains quantitative data stored as a non-numeric type
* `price` is an object instead of float

These violate structural or schema expectations.

---

### **Accuracy** (incorrect or misleading representation of real-world meaning)

These issues involve values that don’t correctly reflect what they’re supposed to represent:

* `stars` also contains the review count

This mixes two different real-world concepts (rating vs. count), making the data misleading.

---

### **Completeness**

* No issues from the list fall under completeness (nothing explicitly missing).

---

### Quick Summary

| Dimension    | Issues Count | Core Problem Type             |
| ------------ | ------------ | ----------------------------- |
| Consistency  | 7            | Formatting / representation   |
| Validity     | 5            | Data type / schema violations |
| Accuracy     | 1            | Misrepresentation of meaning  |
| Completeness | 0            | Missing data                  |

Then the order of severity is:
`Completeness > Validity > Accuracy > Consistency`

```python
# remove whitespaces prefix and whitespaces from col 'author' values
audible_df['author'] = audible_df['author'].str.replace('Writtenby:', '', regex=False).str.replace(' ', '', regex=False)

# ensure it works
audible_df['author'].head().to_list()
```

```python
# turn pasal case of col 'author' into title case
# Insert a space before capital letters (ignoring the start of the string)
audible_df['author'] = audible_df['author'].str.replace(r'(?<!^)(?=[A-Z])', ' ', regex=True)

# ensure it works
audible_df['author'].head().to_list()
```

```python
# remove whitespaces prefix and whitespaces from col 'narrator' values
audible_df['narrator'] = audible_df['narrator'].str.replace('Narratedby:', '', regex=False).str.replace(' ', '', regex=False)

# ensure it works
audible_df['narrator'].head().to_list()
```

```python
# turn pasal case of col 'author' into title case
# Insert a space before capital letters (ignoring the start of the string)
audible_df['narrator'] = audible_df['narrator'].str.replace(r'(?<!^)(?=[A-Z])', ' ', regex=True)

# ensure it works
audible_df['narrator'].head().to_list()
```

```python
# convert 'time' to Timedelta object
def parse_time(x):
    if pd.isna(x):
        return pd.NaT
    
    # Handle special case
    if 'less than' in x.lower():
        return pd.Timedelta(minutes=1)
    
    hrs = 0
    mins = 0
    
    hr_match = re.search(r'(\d+)\s*hr', x)
    min_match = re.search(r'(\d+)\s*min', x)
    
    if hr_match:
        hrs = int(hr_match.group(1))
    if min_match:
        mins = int(min_match.group(1))
    
    return pd.Timedelta(hours=hrs, minutes=mins)

audible_df['time'] = audible_df['time'].apply(parse_time)
audible_df['time']
```

```python
# convert 'releasedate' to datetime object
audible_df['releasedate'] = pd.to_datetime(audible_df['releasedate'], format='%d-%m-%y')
audible_df['releasedate']
```

```python
# convert 'language' to 'category' type
audible_df['language'] = audible_df['language'].astype('category')
```

```python
# Extract rating (e.g., 5, 4.5)
audible_df['star_rating'] = audible_df['stars'].str.extract(r'(\d+\.?\d*)').astype(float)

# Extract number of ratings (e.g., 34)
audible_df['rating_count'] = audible_df['stars'].str.extract(r'(\d+)\s*ratings').fillna(0).astype('int')

# Drop original column stars, no longer needed
audible_df = audible_df.drop(columns=['stars'])
```

```python
# convert 'price' to float
audible_df['price'] = pd.to_numeric(audible_df['price'].str.replace(',', ''), errors='coerce')
audible_df['price']
```
