## What is Data?

Before studying databases, it is important to revisit the concept of data itself.

## What is a Database?

A Database is a shared collection of logically related data, along with the description of that data, designed to meet the information needs of an organization.

A database allows organizations to systematically store, manage, retrieve, and manipulate large volumes of data efficiently.

---

## Major Usages of Databases

### Data Storage

Databases are used to store large volumes of structured data in an organized manner, making the data easily accessible, searchable, and retrievable.

### Data Analysis

Databases support complex analytical operations, report generation, and insight extraction, which are essential for business intelligence and data analytics.

### Record Keeping

Organizations use databases to maintain important records such as:

* Financial transactions
* Customer information
* Employee details
* Inventory records

### Web Applications

Modern web applications heavily depend on databases for:

* User authentication
* Dynamic content generation
* Transaction management
* Data persistence

---

>
> * Create
> * Retrieve
> * Update
> * Delete

---

## Properties of an Ideal Database

### Integrity

Integrity refers to the accuracy and consistency of data.

* Accuracy ensures that stored values are correct.
* Consistency ensures that the same information remains uniform across the entire system at all times.

### Availability

A database should remain accessible whenever required. Ideally, downtime should be minimized to ensure uninterrupted business operations.

### Security

Databases must protect sensitive information from unauthorized access, misuse, or cyber threats.

### Independence from Applications

Multiple applications should be able to interact with a centralized database system.

For example, a company may have:

* A website
* An Android application
* A desktop application

All these platforms should ideally interact with the same centralized database.

### Concurrency

A database should support simultaneous access by multiple users or applications efficiently rather than serving requests strictly sequentially.

## Types of Databases

## 1. Relational Databases (RDBMS)

Relational Databases, also known as SQL Databases, organize data into tables consisting of rows and columns.

Each table in a relational database is referred to as a relation.

Different tables can also be connected with one another using relationships.

### Example

Consider two tables:

* `sportsperson`
* `cars`

The `sportsperson` table stores information about athletes, while the `cars` table stores information about the cars owned by them.

These tables can be related using concepts such as:

* Primary Keys
* Foreign Keys

These concepts establish relationships between tables and will be studied later in detail.

### Examples of Relational Databases

* MySQL
* PostgreSQL
* Oracle Database

---

> Relational Databases are row-oriented databases, meaning that each row stores a complete record along with all its associated attributes.

---

## 2. NoSQL Databases

NoSQL Databases are designed to handle large volumes of unstructured or semi-structured data.

These databases are especially useful when the structure of data is highly dynamic or unpredictable.

### Examples of NoSQL Databases

* MongoDB

### Typical Use Cases

A social media post is a good example of unstructured data because a post may contain:

* Text captions
* Images
* Videos
* Comments
* Reactions
* Variable metadata

Since the structure can vary significantly between records, NoSQL databases become more suitable than rigid relational systems.

These databases are often extremely large due to the scale and diversity of the data they handle.

---

## 3. Column Databases

Column-oriented databases store data column-wise instead of row-wise.

These databases are highly optimized for:

* Data Warehousing
* Analytical workloads
* Aggregation operations

### Why Column Databases are Efficient

Consider three separate tables:

| Table   | Attributes      |
| ------- | --------------- |
| Table 1 | `SId`, `Name`   |
| Table 2 | `SId`, `Branch` |
| Table 3 | `SId`, `CGPA`   |

Suppose an analyst wants to calculate:

* Mean CGPA
* Most common branch (Mode)

In a column-oriented structure:

* Only the required columns need to be scanned.
* Data values are stored contiguously in memory.
* Unnecessary data retrieval is avoided.

This significantly improves:

* Query speed
* Memory efficiency
* Analytical performance

Column databases are widely used in analytics and data warehouse environments.

---

## 4. Graph Databases

Graph Databases are designed to store and query graph-structured data.

These databases are particularly useful when relationships between entities are highly interconnected.

### Common Use Cases

* Social networks
* Recommendation systems
* Fraud detection
* Network analysis

### Examples of Graph Databases

* Neo4j
* Amazon Neptune

Graph databases are generally used in highly specialized scenarios.

### Industry Usage

Large social media platforms use graph-based systems for applications such as:

* Friend recommendations
* Relationship mapping
* Content recommendation engines

---

## 5. Key-Value Databases

Key-Value Databases store information in the form of key-value pairs.

These databases are designed for extremely fast retrieval operations and are commonly used for:

* Caching
* Session storage
* Temporary high-speed data access

### Examples

* Redis
* Amazon DynamoDB

### Example Scenario

Consider a social networking platform displaying:

* Follower count
* User statistics
* Frequently accessed profile data

Instead of repeatedly traversing large relational datasets, frequently required information can be stored as key-value pairs for near-instant retrieval.

This improves performance significantly, although such systems may have limitations in terms of storage persistence or structure flexibility.

## Relational Databases

Relational Databases, also known as SQL Databases, use the relational model to organize data into structured tables consisting of rows and columns.

These databases are among the most widely used database systems due to their:

* Structured organization
* Data consistency
* Reliability
* Support for relationships between data entities

---

## Fundamental Terminologies in Relational Databases

### Relation

A table in a relational database is known as a relation.

For example:

```sql
Employee(ID, ENAME, SALARY, BONUS, DEPT)
```

Here, `Employee` represents a relation.

---

### Attributes or Fields

Columns in a relation are called attributes or fields.

In the above example:

* `ID`
* `ENAME`
* `SALARY`
* `BONUS`
* `DEPT`

are attributes of the `Employee` relation.

Attributes define the properties or characteristics of the stored entity.

---

### Tuples or Records

Rows in a relation are called tuples or records.

Each tuple represents one complete instance of the entity being stored.

For example, one row in the `Employee` table may represent information about a single employee.

---

### Cardinality of a Relation

The total number of rows (tuples/records) in a relation is known as the cardinality of the relation.

Example:

* If an `Employee` table contains 500 rows, then the cardinality of the relation is 500.

---

### Degree of a Relation

The total number of columns (attributes/fields) in a relation is known as the degree of the relation.

Example:

```sql
Employee(ID, ENAME, SALARY, BONUS, DEPT)
```

This relation contains 5 attributes, therefore:

* Degree = 5

---

### NULL Values

Empty or missing values in a database are represented using `NULL`.

A `NULL` value does not necessarily mean zero or an empty string. Instead, it indicates that:

* The value is unknown
* The value is missing
* The value is not applicable

---

### Domain

A domain refers to the set or type of values that can be stored in a particular column.

Examples:

| Attribute | Possible Domain    |
| --------- | ------------------ |
| `ID`      | Integer values     |
| `ENAME`   | Text/String values |
| `SALARY`  | Numeric values     |
| `DEPT`    | Department names   |

Domains help maintain data consistency and validity within a database.

---

## Representation of a Relation

A relation is commonly represented using the following notation:

```sql
Employee(ID, ENAME, SALARY, BONUS, DEPT)
```

Where:

* `Employee` is the relation name
* The values inside parentheses are the attributes of the relation

---

## DBMS (Database Management System)

A Database Management System (DBMS) is a software system that provides the tools and interface required to store, organize, manage, and retrieve data from databases.

A DBMS acts as an intermediary layer between:

* Users or applications
* The actual database system

It simplifies database interactions and abstracts the complexities of low-level data handling.

---

## Hierarchy of the DBMS Ecosystem

The overall database ecosystem can be represented as:

```text
Hardware → Operating System → Database → DBMS → Users
```

### Explanation

* Hardware provides the physical infrastructure.
* The Operating System manages system-level operations.
* The Database stores the actual data.
* The DBMS manages and controls database operations.
* Users interact with the DBMS to access or manipulate data.

---

## Flow of a Database Query

Whenever a user performs any database operation such as retrieval or storage, the request flows through multiple layers:

```text
Users → DBMS → Database → Operating System → Hardware
```

Similarly, the response travels back through the same chain to the user.

This layered architecture helps provide:

* Security
* Abstraction
* Data consistency
* Efficient resource management

---

## Functions of a DBMS

## 1. Data Management

The DBMS handles:

* Data storage
* Data retrieval
* Data modification

It ensures that information is efficiently managed and accessible when required.

---

## 2. Integrity Management

A DBMS maintains the accuracy and consistency of data throughout the database system.

It prevents invalid or inconsistent data from being stored.

---

## 3. Concurrency Control

Multiple users may attempt to access or modify the same data simultaneously.

The DBMS manages concurrent access to ensure:

* Data consistency
* Correct execution of operations
* Prevention of conflicts

---

## 4. Transaction Management

A transaction is a logical unit of work performed on a database.

A DBMS ensures that a transaction:

* Either completes successfully in its entirety
* Or does not occur at all

This principle helps maintain database reliability and consistency.

---

## 5. Security Management

The DBMS restricts database access to authorized users only.

Security mechanisms may include:

* Authentication
* Authorization
* Access control
* Data encryption

---

## 6. Utility Services

A DBMS also provides several utility functions such as:

* Data import and export
* User management
* Backup and recovery
* Activity logging
* Monitoring and maintenance tools

These utilities help ensure efficient database administration and operational reliability.

## Database Keys

A key in a database is an attribute (column) or a set of attributes that uniquely identifies a tuple (row) in a table.

Keys are extremely important in relational databases because they help in:

* Maintaining data integrity
* Preventing duplicate records
* Establishing relationships between tables
* Improving query efficiency
* Ensuring data consistency

---

## 1. Super Key

A Super Key is a set of one or more attributes that can uniquely identify each tuple (row) in a relation.

A Super Key may contain extra or redundant attributes in addition to the minimum required attributes.

### Example

Consider the following `Employee` table:

| EmpID | Email                                 | Name | Department |
| ----- | ------------------------------------- | ---- | ---------- |
| 101   | [a@company.com](mailto:a@company.com) | Amit | HR         |
| 102   | [b@company.com](mailto:b@company.com) | Ravi | IT         |

Possible Super Keys:

* `EmpID`
* `Email`
* `(EmpID, Name)`
* `(EmpID, Department)`
* `(Email, Name)`

All these combinations can uniquely identify a row.

However, combinations like `(EmpID, Name)` contain unnecessary attributes because `EmpID` alone is already sufficient.

---

## 2. Candidate Key

A Candidate Key is a minimal Super Key.

This means:

* It uniquely identifies each tuple.
* It does not contain any redundant attributes.

In simple terms, it is the smallest possible set of attributes capable of uniquely identifying a row.

### Example

Using the same `Employee` table:

| EmpID | Email                                 | Name | Department |
| ----- | ------------------------------------- | ---- | ---------- |
| 101   | [a@company.com](mailto:a@company.com) | Amit | HR         |
| 102   | [b@company.com](mailto:b@company.com) | Ravi | IT         |

Possible Candidate Keys:

* `EmpID`
* `Email`

Why?

* Both uniquely identify every employee.
* Neither contains unnecessary attributes.

However:

* `(EmpID, Name)` is not a Candidate Key because `Name` is redundant.

---

## 3. Primary Key

A Primary Key is the selected Candidate Key used as the main identifier for tuples in a table.

A table can have only one Primary Key.

The Primary Key:

* Must uniquely identify every tuple
* Cannot contain `NULL` values
* Cannot contain duplicate values

The Primary Key is chosen from the available Candidate Keys.

---

### Mandatory Criteria for a Primary Key

#### It Cannot Contain NULL Values

Every row must contain a valid Primary Key value.

#### It Must Not Contain Duplicate Values

Each tuple must have a unique identity.

---

### Preferable Characteristics of a Good Primary Key

#### Numerical Values are Preferred

Numeric keys generally improve indexing and performance.

#### Small in Length

Smaller keys improve storage efficiency and query performance.

#### Stable and Constant

The key should not change frequently because updates to Primary Keys can create maintenance complexity.

---

## Example of a Primary Key

| EmpID | Email                                 | Name | Department |
| ----- | ------------------------------------- | ---- | ---------- |
| 101   | [a@company.com](mailto:a@company.com) | Amit | HR         |
| 102   | [b@company.com](mailto:b@company.com) | Ravi | IT         |

Here:

* `EmpID` can be selected as the Primary Key.

Reason:

* Unique
* Non-null
* Numeric
* Small in size
* Stable

Although `Email` is also a Candidate Key, `EmpID` is generally preferred as the Primary Key.

---

## 4. Alternate Key

An Alternate Key is a Candidate Key that is not selected as the Primary Key.

In simple terms:

```text id="p1m3fx"
Alternate Key = Candidate Key − Primary Key
```

### Example

Consider the table:

| EmpID | Email                                 | Name |
| ----- | ------------------------------------- | ---- |
| 101   | [a@company.com](mailto:a@company.com) | Amit |
| 102   | [b@company.com](mailto:b@company.com) | Ravi |

Candidate Keys:

* `EmpID`
* `Email`

Suppose `EmpID` is selected as the Primary Key.

Then:

* `Email` becomes the Alternate Key.

---

## 5. Composite Key

A Composite Key is a Primary Key formed using two or more attributes together.

Composite Keys are used when a single attribute alone is insufficient to uniquely identify tuples.

---

## Example of a Composite Key

Consider a `Student_Course` table:

| StudentID | CourseID | Semester |
| --------- | -------- | -------- |
| 101       | CS101    | 1        |
| 101       | MA102    | 1        |
| 102       | CS101    | 1        |

Here:

* `StudentID` alone is not unique.
* `CourseID` alone is not unique.

But the combination:

```text id="ykjz1x"
(StudentID, CourseID)
```

uniquely identifies each tuple.

Therefore:

* `(StudentID, CourseID)` becomes the Composite Key.

---

## 6. Surrogate Key

A Surrogate Key is an artificially created key used when no existing attribute or combination of attributes can uniquely identify tuples.

These keys are manually generated solely for identification purposes.

Surrogate Keys are commonly:

* Auto-incremented integers
* System-generated IDs
* Artificial identifiers

---

## Example of a Surrogate Key

Consider the following `Student` table:

| Name | Branch | CGPA |
| ---- | ------ | ---- |
| Amit | CSE    | 8.5  |
| Amit | CSE    | 8.5  |
| Ravi | ECE    | 7.9  |

None of the attributes or combinations reliably guarantee uniqueness.

Therefore, a new attribute can be introduced:

| StudentID | Name | Branch | CGPA |
| --------- | ---- | ------ | ---- |
| 1         | Amit | CSE    | 8.5  |
| 2         | Amit | CSE    | 8.5  |
| 3         | Ravi | ECE    | 7.9  |

Here:

* `StudentID` is a Surrogate Key.

It is artificially created solely to uniquely identify tuples.

---

## 7. Foreign Key

A Foreign Key is an attribute in one table that refers to the Primary Key of another table.

Foreign Keys are used to establish relationships between tables.

They help maintain Referential Integrity within a database.

---

## Example of a Foreign Key

### Student Table

| StudentID | Name |
| --------- | ---- |
| 101       | Amit |
| 102       | Ravi |

Here:

* `StudentID` is the Primary Key.

---

### Course Table

| CourseID | StudentID | CourseName       |
| -------- | --------- | ---------------- |
| C1       | 101       | DBMS             |
| C2       | 102       | Machine Learning |

In the `Course` table:

* `StudentID` is a Foreign Key.

This is because it references:

```text id="j9z4fh"
Student.StudentID
```

from the `Student` table.

---

## Purpose of Foreign Keys

Foreign Keys help in:

* Connecting related tables
* Preventing invalid references
* Maintaining referential integrity
* Supporting relational database design

---

## Summary of Database Keys

| Key Type      | Purpose                                                   |
| ------------- | --------------------------------------------------------- |
| Super Key     | Any attribute combination that uniquely identifies tuples |
| Candidate Key | Minimal Super Key                                         |
| Primary Key   | Selected Candidate Key used as main identifier            |
| Alternate Key | Candidate Key not chosen as Primary Key                   |
| Composite Key | Primary Key formed using multiple attributes              |
| Surrogate Key | Artificially created identifier                           |
| Foreign Key   | Establishes relationship between tables                   |


## Cardinality of Relationships

Cardinality in database relationships refers to the number of occurrences of one entity that can be associated with another entity in a relationship.

In simple terms, cardinality defines how many instances of one entity can relate to instances of another entity.

Understanding cardinality is essential for designing efficient and logically correct relational databases.

---



---

## Types of Relationships in Databases

There are three primary types of relationships in relational databases:

1. One-to-One Relationship
2. One-to-Many Relationship
3. Many-to-Many Relationship

---

> A Many-to-One relationship is simply the reverse perspective of a One-to-Many relationship.

---

## 1. One-to-One Relationship

A One-to-One (1:1) relationship occurs when:

* One entity is associated with only one entity from another table
* And vice versa

This means each record in Table A corresponds to exactly one record in Table B.

---

## Characteristics of One-to-One Relationships

* Each entity has only one matching entity.
* Usually implemented within a single table.
* Commonly used when data separation is required for:

  * Security
  * Modularity
  * Optional attributes

---

## Example

Consider a system storing employee information and passport information.

| EmployeeID | EmployeeName | PassportNumber |
| ---------- | ------------ | -------------- |
| 101        | Amit         | P12345         |
| 102        | Ravi         | P67890         |

Here:

* One employee has one passport.
* One passport belongs to one employee.

This forms a One-to-One relationship.

---

## Important Observation

In most practical scenarios, a One-to-One relationship can often be handled using a single table.

---

## 2. One-to-Many Relationship

A One-to-Many (1:M) relationship occurs when:

* One entity can be associated with multiple entities
* But each of those entities belongs to only one parent entity

This is one of the most common relationship types in relational databases.

---

## Example

### Table: `branch`

| branch_id | name | HOD   |
| --------- | ---- | ----- |
| 1         | CSE  | HOD 1 |
| 2         | ECE  | HOD 2 |

---

### Table: `student`

| student_id | name   | branch_id |
| ---------- | ------ | --------- |
| 1          | Stud 1 | 1         |
| 2          | Stud 2 | 1         |
| 3          | Stud 3 | 2         |
| 4          | Stud 4 | 2         |

---

## Relationship Analysis

From the above tables:

* One branch can contain many students.
* Each student belongs to only one branch.

Therefore:

```text id="kz7x2n"
branch → student = One-to-Many Relationship
```

---

## Role of Foreign Key

Notice the attribute:

```text id="p4v8rq"
branch_id
```

inside the `student` table.

Here:

* `branch.branch_id` is the Primary Key
* `student.branch_id` acts as a Foreign Key

The Foreign Key establishes the relationship between the two tables.

---

## Important Observation

A One-to-Many relationship generally requires two tables:

1. Parent table
2. Child table

The child table contains the Foreign Key referencing the parent table.

---

## 3. Many-to-Many Relationship

A Many-to-Many (M:N) relationship occurs when:

* One entity can be associated with multiple entities from another table
* And vice versa

This means both sides can have multiple corresponding records.

---

## Example Scenario

Consider:

* Students enrolling in courses

A student can enroll in multiple courses.

Similarly:

* A course can contain multiple students.

This creates a Many-to-Many relationship.

---

## Why a Third Table is Required

A Many-to-Many relationship cannot be implemented directly using only two tables in a relational database.

Instead, a third table called a Junction Table or Bridge Table is required.

This intermediate table stores the relationships between the two entities.

---

## Example Structure

### Table: `student`

| student_id | name |
| ---------- | ---- |
| 1          | Amit |
| 2          | Ravi |

---

### Table: `course`

| course_id | course           |
| --------- | ---------------- |
| C101      | DBMS             |
| C102      | Machine Learning |

---

### Table: `student_course`

| student_id | course_id | enrollment_date |
| ---------- | --------- | --------------- |
| 1          | C101      | 2026-01-10      |
| 1          | C102      | 2026-01-12      |
| 2          | C101      | 2026-01-15      |

---

## Relationship Analysis

From the above example:

* Student `1` is enrolled in multiple courses.
* Course `C101` contains multiple students.

Thus:

```text id="j2f8vc"
student ↔ course = Many-to-Many Relationship
```

---

## Role of the Junction Table

The table:

```text id="y5r1qt"
student_course
```

acts as the bridge between:

* `student`
* `course`

It contains:

* Foreign Key referencing `student`
* Foreign Key referencing `course`

This table is responsible for maintaining the Many-to-Many relationship.

---

## Important Observation

A Many-to-Many relationship generally requires:

1. First entity table
2. Second entity table
3. Junction/Bridge table

Without the third table, proper representation of the relationship becomes difficult in relational databases.

---

## Summary of Relationship Cardinalities

| Relationship Type | Description                                | Typical Number of Tables |
| ----------------- | ------------------------------------------ | ------------------------ |
| One-to-One        | One entity maps to one entity              | Usually 1                |
| One-to-Many       | One entity maps to multiple entities       | Usually 2                |
| Many-to-Many      | Multiple entities map to multiple entities | Usually 3                |

---

> The table count mentioned above should be understood as a practical rule of thumb commonly followed in relational database design. In real-world systems, implementations may vary depending on normalization, optimization strategies, or specific application requirements.

## Drawbacks of Databases

Although databases provide powerful mechanisms for storing, managing, and analyzing data, they also come with certain limitations and challenges. These drawbacks become more significant as systems grow in complexity and scale.

---

### 1. Complexity

Setting up, configuring, and maintaining a database system can be complex and time-consuming, especially for large-scale applications.

Database administration involves multiple responsibilities such as:

* Schema design
* Performance optimization
* Backup management
* Security configuration
* Query optimization
* Concurrency management

As the system size increases, managing the database infrastructure becomes increasingly challenging.

---

### 2. Cost

The overall cost of maintaining a database system can be high.

The expenses may include:

* Hardware infrastructure
* Licensed database software
* Cloud hosting services
* Maintenance tools
* Skilled database administrators (DBAs)
* Backup and recovery systems

Enterprise-level database systems often require significant financial investment.

---

### 3. Scalability Challenges

As the volume of stored data increases, database systems may experience:

* Slower query performance
* Increased storage requirements
* Higher computational overhead
* Network bottlenecks

Managing extremely large datasets efficiently requires advanced optimization techniques such as:

* Indexing
* Partitioning
* Sharding
* Distributed database architectures

Without proper optimization, scalability becomes a major concern.

---

### 4. Data Integrity Challenges

Maintaining data accuracy and consistency can become difficult, particularly in environments where multiple users simultaneously access and update data.

Common issues include:

* Duplicate records
* Inconsistent updates
* Transaction conflicts
* Concurrency-related anomalies

Ensuring strong data integrity often requires complex transaction and locking mechanisms.

---

### 5. Security Risks

Databases frequently store highly sensitive information such as:

* Financial records
* Personal user data
* Business intelligence
* Authentication credentials

Protecting databases against unauthorized access and cyber attacks is a critical challenge.

Potential security threats include:

* Data breaches
* SQL Injection attacks
* Insider threats
* Unauthorized privilege escalation
* Ransomware attacks

Proper database security requires continuous monitoring and strict access control mechanisms.

---

### 6. Data Migration Complexity

Migrating data from one database system to another can be difficult and time-consuming.

Data migration may become necessary due to:

* System upgrades
* Technology changes
* Cloud migration
* Database vendor transitions

Migration challenges may include:

* Schema incompatibility
* Data transformation requirements
* Downtime risks
* Data loss possibilities
* Performance disruptions

Careful planning and validation are essential during migration processes.

---

### 7. Limited Flexibility

Traditional relational databases often follow rigid predefined schemas.

This rigidity can create difficulties when:

* Business requirements change frequently
* New data formats need to be introduced
* Unstructured or semi-structured data must be handled

Modifying an existing schema in large production systems can become expensive and operationally risky.

This limitation is one of the reasons why NoSQL databases gained popularity for highly dynamic and flexible data environments.

---

### Conclusion

Despite these challenges, databases remain one of the foundational technologies of modern computing systems. Most drawbacks can be mitigated through:

* Proper database design
* Efficient architecture planning
* Security best practices
* Performance optimization
* Scalable infrastructure management

Understanding these limitations is important for selecting the appropriate database system based on organizational and application requirements.
