## Database Ecosystem and Hierarchy

Before learning SQL and SQL DDL, it is important to understand the hierarchy of how database systems are organised.

```text
Database Server
    └── DBMS (Database Management System)
            └── Databases
                    └── Tables / Relations
```

### 1. Database Server

A Database Server is the machine or system where database software runs and stores data.
It can be:

* A local computer
* A cloud server
* A dedicated enterprise server

The database server is responsible for:

* Storing data
* Processing database requests
* Managing users and connections
* Running database services continuously

---

### 2. DBMS (Database Management System)

A DBMS is the software that allows users to create, manage, and interact with databases.

Examples:

* PostgreSQL
* MySQL
* Oracle Database
* Microsoft SQL Server

A DBMS provides:

* Data storage
* Security
* Query processing
* Transaction management
* Backup and recovery
* Multi-user access

In this course, the DBMS being used is PostgreSQL.

---

### 3. Databases

A database is an organised collection of related data.

Examples:

* College Management Database
* Banking Database
* E-Commerce Database

A single DBMS can contain multiple databases.

For example, inside PostgreSQL, you may create:

* `college_db`
* `library_db`
* `employee_db`

Each database stores its own separate tables and data.

---

### 4. Tables / Relations

Tables are the actual structures where data is stored.

In relational databases, tables are also called Relations.

Example Table: `students`

| student_id | name  | course |
| ---------- | ----- | ------ |
| 101        | Harsh | MCA    |
| 102        | Rahul | BCA    |

A table consists of:

* Rows (Records/Tuples)
* Columns (Attributes/Fields)

## What is SQL?

Structured Query Language (SQL) is a programming language used for managing and manipulating data in relational databases. It allows you to insert, update, retrieve, and delete data in a database. It is widely used for data management in many applications, websites, and businesses.

In simple terms, SQL is used to communicate with and control databases.

---

## Types of SQL Commands

* SQL: Structured Query Language

### 1. DDL — Data Definition Language

DDL commands are used to define and manage database structures.

* `CREATE`

  * Creates new database objects

* `ALTER`

  * Modifies existing database objects

* `DROP`

  * Deletes existing database objects permanently

* `TRUNCATE`

  * Removes all rows from a table while keeping the table structure intact

---

### 2. DML — Data Manipulation Language

DML commands are used to manipulate the data stored inside tables.

* `INSERT`

  * Adds new rows into tables

* `UPDATE`

  * Modifies existing data in tables

* `DELETE`

  * Removes rows from tables

* `SELECT`

  * Retrieves data from tables

> The `SELECT` command is sometimes categorised under DQL (Data Query Language) because it is primarily used for querying data.

---

### 3. DCL — Data Control Language

DCL commands are used for access control and permissions.

* `GRANT`

  * Provides access rights on database objects

* `REVOKE`

  * Removes access rights from users

---

### 4. TCL — Transaction Control Language

TCL commands are used to manage transactions in databases.

* `COMMIT`

  * Saves changes permanently

* `ROLLBACK`

  * Undoes changes that are not committed

## Understanding Why We Use the `postgres` Database

A very important PostgreSQL concept is:

> A connection in PostgreSQL is always made to one specific database.

* You cannot execute queries without being connected somewhere first.

---

## Important Concept: `CREATE DATABASE` is a Cluster-Level Operation

The command:

```sql
CREATE DATABASE firstdb;
```

does not create an object *inside* the current database.

Instead, it creates an entirely new database in the PostgreSQL cluster.

Because of this:

* PostgreSQL requires you to already be connected to an existing database.
* Usually, the administrative database:

  ```sql
  postgres
  ```

  is used for this purpose.

---

## Why You Cannot Connect to a Database Before It Exists

This is impossible:

```sql
-- invalid logic
CONNECT firstdb;
CREATE DATABASE firstdb;
```

because:

* `firstdb` does not exist yet
* PostgreSQL cannot establish a connection to a non-existent database

Therefore, the correct workflow is:

```sql
-- connected to postgres database
CREATE DATABASE firstdb;
```

Then afterwards:

* connect to `firstdb`
* and begin creating tables and other objects inside it

---

## Internal Working of `CREATE DATABASE`

Internally, PostgreSQL creates a new database by copying a template database.

By default, PostgreSQL uses:

```text
template1
```

as the base template.

This process happens at the server/cluster level rather than inside the current database session.

---

## Additional Technical Details

### 1. `CREATE DATABASE` Cannot Run Inside a Transaction Block

PostgreSQL restricts this command because database creation is a system-level operation.

Example:

```sql
BEGIN;

CREATE DATABASE testdb; -- ERROR

COMMIT;
```

This produces an error because:

* `CREATE DATABASE` must execute independently

---

### 2. Query Tabs in pgAdmin4 are Database-Specific

Each Query Tool tab in pgAdmin4 is attached to:

* one active database connection

So:

* a query window opened from `postgres`
  operates inside `postgres`
* a query window opened from `firstdb`
  operates inside `firstdb`

## Difference Between PostgreSQL and MySQL

This behaviour differs from systems like MySQL.

In MySQL:

* databases/schemas are handled somewhat differently within the same server connection

In PostgreSQL:

* database isolation is stronger
* every session is attached to exactly one database
* cluster-level operations are treated separately from database-level operations

This design improves:

* stability
* isolation
* administration
* security
* transaction handling in enterprise systems.

## PostgreSQL vs MySQL: `IF NOT EXISTS` with `CREATE DATABASE`

In systems like MySQL, developers commonly write safer and more robust queries such as:

```sql id="4im87y"
CREATE DATABASE IF NOT EXISTS firstdb;
```

Similarly, while deleting a database:

```sql id="9v77kq"
DROP DATABASE IF EXISTS firstdb;
```

instead of:

```sql id="v74t4r"
DROP DATABASE firstdb;
```

The purpose of these clauses is:

* preventing unnecessary errors
* making scripts safer
* improving automation and deployment workflows

---

## PostgreSQL Limitation with `CREATE DATABASE IF NOT EXISTS`

In PostgreSQL, the following syntax is **not supported**:

```sql id="vvsh0s"
CREATE DATABASE IF NOT EXISTS learningdb;
```

If the database already exists, PostgreSQL throws an error such as:

```text id="qj3m79"
ERROR: database "learningdb" already exists
```

Therefore, while working with PostgreSQL:

* you must manually check whether a database already exists
* usually through the pgAdmin4 Object Explorer

---

## Why PostgreSQL Does Not Support It

`CREATE DATABASE` in PostgreSQL is a cluster-level administrative operation.

Since databases are isolated objects managed separately from ordinary SQL objects, PostgreSQL intentionally restricts some convenience features for this command.

This behaviour differs from MySQL, where schemas/databases are handled differently internally.

---

## Objects That Support `IF NOT EXISTS` in PostgreSQL

Although PostgreSQL does not support:

```sql id="nlt0sn"
CREATE DATABASE IF NOT EXISTS ...
```

many other objects do support it.

Examples include:

```sql id="kl7q8d"
CREATE TABLE IF NOT EXISTS ...
```

```sql id="c8mn7m"
CREATE SCHEMA IF NOT EXISTS ...
```

```sql id="8gjk6r"
CREATE INDEX IF NOT EXISTS ...
```

```sql id="7dhuwr"
CREATE EXTENSION IF NOT EXISTS ...
```

This feature is extremely useful when writing:

* reusable SQL scripts
* deployment scripts
* migration files
* automation pipelines

# TRUNCATE TABLE

## What is `TRUNCATE`?

The `TRUNCATE` command removes **all rows** from a table very quickly.

Example:

```sql id="p8f8cm"
TRUNCATE TABLE users;
```

---

## Important Characteristics of `TRUNCATE`

### 1. Deletes All Rows

After execution:

| Before TRUNCATE | After TRUNCATE      |
| --------------- | ------------------- |
| Data exists     | Table becomes empty |

The table structure still remains.

---

### 2. Table Structure is Preserved

`TRUNCATE` does **not** remove:

* table name
* columns
* constraints
* primary keys

Only the data is removed.

---

### 3. Faster Than `DELETE`

`TRUNCATE` is usually much faster than:

```sql id="u4zctg"
DELETE FROM users;
```

because:

* PostgreSQL removes table data internally in bulk
* row-by-row deletion is avoided

---

### 4. Must Be Handled Carefully

`TRUNCATE` is a dangerous operation because:

* it removes all records instantly
* accidental execution can wipe important data

Therefore:

* always verify the table name carefully
* avoid using it carelessly in production databases

---

## Example Workflow

Before:

| user_id | name  |
| ------- | ----- |
| 1       | Harsh |
| 2       | Rahul |

Execute:

```sql id="ij12vi"
TRUNCATE TABLE users;
```

After:

| user_id   | name      |
| --------- | --------- |
| *(empty)* | *(empty)* |

---

# DROP TABLE

The `DROP` command permanently removes a table from the database.

Example:

```sql id="g3jlwm"
DROP TABLE users;
```

---

## What Happens After `DROP TABLE`

The following are permanently deleted:

* table structure
* rows/data
* indexes
* constraints
* associated metadata

The table no longer exists in the database.

---

## Difference Between `TRUNCATE` and `DROP`

| Feature                 | TRUNCATE | DROP |
| ----------------------- | -------- | ---- |
| Removes data            | Yes      | Yes  |
| Removes table structure | No       | Yes  |
| Table remains usable    | Yes      | No   |
| Removes columns         | No       | Yes  |
| Removes constraints     | No       | Yes  |

---

## Safer Version of DROP TABLE

PostgreSQL supports:

```sql id="lrjry0"
DROP TABLE IF EXISTS users;
```

This prevents errors if the table does not exist already.

This is preferred in:

* reusable SQL scripts
* migration files
* automated setups
* deployment workflows.

## Data Integrity

Data Integrity refers to the:

* accuracy
* consistency
* reliability
* completeness

of data stored inside a database.

It ensures that:

* invalid data is not inserted
* relationships between tables remain correct
* accidental corruption is prevented
* unauthorized or inconsistent modifications do not occur

Data integrity is one of the most important goals of relational database systems such as PostgreSQL.

---

## Methods Used to Ensure Data Integrity

There are several mechanisms used to maintain integrity in databases.

### 1. Constraints

Constraints are rules enforced by the DBMS on table columns or tables.

They ensure that:

* only valid data is inserted
* invalid updates are prevented
* relationships between tables remain consistent

Constraints are generally defined using DDL commands.

---

### 2. Transactions

A Transaction is a sequence of database operations treated as a single logical unit of work.

Transactions help ensure:

* consistency
* rollback capability
* atomicity of operations

Example:

* transferring money between bank accounts

If one operation fails:

* the entire transaction can be rolled back.

---

### 3. Normalization

Normalization is a database design technique used to:

* reduce redundancy
* organize data efficiently
* improve consistency

It divides large tables into smaller related tables.

---

# Constraints

Constraints are rules applied to database columns or tables to maintain data integrity.

They help prevent:

* invalid data
* duplicate entries
* broken relationships
* inconsistent values

---

# Types of Constraints in PostgreSQL

The major constraints and related features are:

1. `NOT NULL`
2. `UNIQUE`
3. `PRIMARY KEY`
4. `SERIAL`
5. `CHECK`
6. `DEFAULT`
7. `FOREIGN KEY`

---

# 1. NOT NULL Constraint

The `NOT NULL` constraint prevents a column from storing `NULL` values.

Example:

```sql id="0x2mtv"
CREATE TABLE users (
    user_id INTEGER NOT NULL,
    name VARCHAR(255) NOT NULL,
    email VARCHAR(255),
    password VARCHAR(255)
);
```

---

## Important PostgreSQL Clarification

The following statement is **incorrect in PostgreSQL**:

> "If values are omitted, PostgreSQL automatically inserts `0` for INTEGER and empty string for VARCHAR."

PostgreSQL does **not** automatically generate:

* `0`
* empty strings
* default placeholder values

Instead:

* omitted columns receive `NULL`
* unless a `DEFAULT` value exists

Therefore, this insert:

```sql id="i7e86d"
INSERT INTO users (email)
VALUES ('abc@email.com');
```

would produce an error because:

* `user_id`
* `name`

are marked as `NOT NULL`.

PostgreSQL error example:

```text id="6hiv6r"
ERROR: null value in column "user_id" violates not-null constraint
```

---

# 2. UNIQUE Constraint

The `UNIQUE` constraint prevents duplicate values in a column.

Example:

```sql id="ukto9z"
CREATE TABLE users (
    user_id INTEGER NOT NULL,
    name VARCHAR(255) NOT NULL,
    email VARCHAR(255) NOT NULL UNIQUE,
    password VARCHAR(255) NOT NULL
);
```

Now:

* duplicate email addresses are not allowed.

---

## PostgreSQL Error Example

If duplicate values are inserted:

```text id="6e9khm"
duplicate key value violates unique constraint "users_email_key"
Key (email)=(moneyk@email.com) already exists.
```

This ensures uniqueness of email addresses.

---

# Alternate Method for Defining Constraints

Constraints can also be defined separately using the `CONSTRAINT` keyword.

Example:

```sql id="wqv1bd"
CREATE TABLE users (
    user_id INTEGER NOT NULL,
    name VARCHAR(255) NOT NULL,
    email VARCHAR(255) NOT NULL,
    password VARCHAR(255) NOT NULL,

    CONSTRAINT users_unique_credentials
        UNIQUE (name, email, password)
);
```

---

## Advantages of Named Constraints

### 1. Better Readability

The constraint receives a meaningful name:

```sql id="mjlwmv"
users_unique_credentials
```

This improves:

* debugging
* maintenance
* schema readability

---

### 2. Easier Error Identification

PostgreSQL error messages display the constraint name directly.

---

### 3. Composite Constraints

This example creates a composite unique constraint:

```sql id="s71u03"
UNIQUE (name, email, password)
```

Meaning:

* the *combination* must remain unique
* individual columns may still contain duplicates

For example:

| name  | email                             | password | Allowed? |
| ----- | --------------------------------- | -------- | -------- |
| Harsh | [a@gmail.com](mailto:a@gmail.com) | 123      | Yes      |
| Harsh | [a@gmail.com](mailto:a@gmail.com) | 123      | No       |
| Harsh | [a@gmail.com](mailto:a@gmail.com) | xyz      | Yes      |

---

## PostgreSQL Object Explorer Behaviour

In pgAdmin 4:

* the constraint appears under:

  ```text
  users → Constraints
  ```

---

## Best Practice in PostgreSQL

For production-grade PostgreSQL schemas:

* named constraints are preferred
* especially in large projects

because:

* they improve maintainability
* simplify migrations
* make debugging easier

---

# 3. PRIMARY KEY Constraint

A `PRIMARY KEY`:

* uniquely identifies each row
* automatically becomes:

  * `NOT NULL`
  * `UNIQUE`

---

## Method 1 — Inline Primary Key

```sql id="9tnh1n"
CREATE TABLE users (
    user_id INTEGER PRIMARY KEY,
    name VARCHAR(255) NOT NULL,
    email VARCHAR(255) NOT NULL,
    password VARCHAR(255) NOT NULL,

    CONSTRAINT users_email_unique
        UNIQUE (name, email, password)
);
```

---

## Method 2 — Named Primary Key Constraint

```sql id="0m2vfo"
CREATE TABLE users (
    user_id INTEGER,
    name VARCHAR(255) NOT NULL,
    email VARCHAR(255) NOT NULL,
    password VARCHAR(255) NOT NULL,

    CONSTRAINT users_email_unique
        UNIQUE (name, email, password),

    CONSTRAINT users_pk
        PRIMARY KEY (user_id)
);
```

---

## Composite Primary Keys

This statement is correct:

```sql id="xwzgl4"
PRIMARY KEY (user_id, email)
```

This creates a Composite Primary Key.

Important clarification:

* A table can only have **one PRIMARY KEY constraint**
* but that single primary key may contain multiple columns

So technically:

* there are not multiple primary keys
* there is one composite primary key

---

## Method 3 — Using ALTER TABLE

```sql id="g8lsxx"
CREATE TABLE users (
    user_id INTEGER NOT NULL,
    name VARCHAR(255) NOT NULL,
    email VARCHAR(255) NOT NULL,
    password VARCHAR(255) NOT NULL,

    CONSTRAINT users_email_unique
        UNIQUE (name, email, password)
);

ALTER TABLE users
ADD PRIMARY KEY (user_id);
```

---

## Important PostgreSQL Behaviour

Correct statement:

> A PRIMARY KEY automatically implies:

* `NOT NULL`
* `UNIQUE`

Therefore:

* explicitly writing them again is unnecessary.

---

# 4. SERIAL

In PostgreSQL, `SERIAL` is used for auto-incrementing integer columns.

Example:

```sql id="mhcyfj"
CREATE TABLE users (
    user_id SERIAL PRIMARY KEY,
    name VARCHAR(255) NOT NULL,
    email VARCHAR(255) NOT NULL UNIQUE,
    password VARCHAR(255) NOT NULL
);
```

---

## How SERIAL Works Internally

`SERIAL` is not technically a constraint.

It is a PostgreSQL pseudo-type that automatically:

1. creates a sequence object
2. assigns auto-incrementing values
3. links the sequence to the column

---

## GUI Behaviour

When rows are inserted without specifying `user_id`:

```text id="2q58y0"
1  moneyk  moneyk@email.com  moneyk
2  ankit   ankit@email.com   ankit
```

the IDs are automatically generated.

---

## PostgreSQL Modern Recommendation

In modern PostgreSQL versions, the preferred approach is:

```sql id="ojf8u8"
GENERATED ALWAYS AS IDENTITY
```

Example:

```sql id="56tvxk"
user_id INTEGER GENERATED ALWAYS AS IDENTITY PRIMARY KEY
```

because:

* it follows SQL standards more closely
* is more explicit
* improves portability

However:

* `SERIAL` is still extremely common and widely used.

---

# 5. CHECK Constraint

The `CHECK` constraint validates data using logical conditions.

Example:

```sql id="plmr4m"
CREATE TABLE students (
    student_id SERIAL PRIMARY KEY,
    name VARCHAR(50) NOT NULL,
    age INTEGER CHECK (age > 6 AND age < 25)
);
```

---

## Better PostgreSQL Style

Named constraints are preferred:

```sql id="yrj2x7"
CREATE TABLE students (
    student_id SERIAL PRIMARY KEY,
    name VARCHAR(50) NOT NULL,
    age INTEGER,

    CONSTRAINT students_age_check
        CHECK (age > 6 AND age < 25)
);
```

---

## Constraint Violation Example

```text id="g1jgfw"
new row for relation "students" violates check constraint "students_age_check"
Failing row contains (1, moneyk, 45).
```

---

# 6. DEFAULT Constraint

The `DEFAULT` clause automatically inserts a predefined value if no value is supplied.

Example:

```sql id="3a2iyd"
CREATE TABLE ticket (
    ticket_id INTEGER PRIMARY KEY,
    name VARCHAR(255) NOT NULL,
    travel_date DATE DEFAULT CURRENT_DATE
);
```

---

## PostgreSQL Behaviour

If `travel_date` is omitted:

* PostgreSQL automatically inserts:

  ```sql
  CURRENT_DATE
  ```

---

# 7. FOREIGN KEY Constraint

A `FOREIGN KEY` creates a relationship between two tables.

Example:

```sql id="t2h6hk"
CREATE TABLE customers (
    cid SERIAL PRIMARY KEY,
    name VARCHAR(255) NOT NULL,
    email VARCHAR(255) NOT NULL UNIQUE
);

CREATE TABLE orders (
    order_id INTEGER PRIMARY KEY,
    cid INTEGER,
    order_date TIMESTAMP NOT NULL DEFAULT CURRENT_TIMESTAMP,

    CONSTRAINT orders_fk
        FOREIGN KEY (cid)
        REFERENCES customers(cid)
);
```

---

## Purpose of Foreign Keys

Foreign keys ensure:

* referential integrity
* valid relationships between tables

Meaning:

* `orders.cid`
  must reference an existing:

  ```sql
  customers.cid
  ```

---

## TRUNCATE Restriction

The following PostgreSQL error is expected:

```text id="5jupc0"
ERROR: cannot truncate a table referenced in a foreign key constraint
```

because:

* removing parent table data would break references in child tables

---

## DROP TABLE Restriction

Similarly:

```text id="08xyg6"
cannot drop table customers because other objects depend on it
```

This protects database consistency.

# Referential Actions

Referential Actions define what happens to related rows when:

* parent rows are updated
* parent rows are deleted

---

# Types of Referential Actions

## 1. RESTRICT (Default Behaviour)

Prevents:

* deletion
* updates

if dependent rows exist.

---

## 2. CASCADE

Automatically propagates changes.

Example:

```sql id="36gdma"
CREATE TABLE customers (
    cid SERIAL PRIMARY KEY,
    name VARCHAR(255) NOT NULL,
    email VARCHAR(255) NOT NULL UNIQUE
);

CREATE TABLE orders (
    order_id INTEGER PRIMARY KEY,
    cid INTEGER,
    order_date TIMESTAMP NOT NULL DEFAULT CURRENT_TIMESTAMP,

    CONSTRAINT orders_fk
        FOREIGN KEY (cid)
        REFERENCES customers(cid)
        ON DELETE CASCADE
        ON UPDATE CASCADE
);
```

---

## PostgreSQL Clarification About CASCADE

This statement needs correction:

> "Whatever changes I make in one table gets shared across both tables."

Actual behaviour:

* only referenced key changes propagate
* ordinary column changes do not automatically synchronize tables

Example:

* deleting a customer deletes related orders
* updating `cid` updates related foreign keys

But:

* updating `customers.name`
  does not update anything in `orders`

---

## 3. SET NULL

When parent rows are deleted or updated:

* related foreign key values become `NULL`

Requirement:

* the foreign key column must allow `NULL`.

---

## 4. SET DEFAULT

When parent rows are deleted or updated:

* foreign key values are set to their column default value.

Example:

```sql id="z1mdso"
cid INTEGER DEFAULT 0
```

## ALTER TABLE Command

The `ALTER TABLE` statement in SQL is used to modify the structure of an existing table.

In PostgreSQL, `ALTER TABLE` is one of the most commonly used DDL commands because database structures often evolve over time.

Using `ALTER TABLE`, we can:

1. Add columns
2. Remove columns
3. Modify columns
4. Add constraints
5. Remove constraints
6. Rename columns
7. Rename tables

---

## ADD COLUMN

The `ADD COLUMN` clause is used to add new columns to an existing table.

Example:

```sql id="lz9wo6"
ALTER TABLE customers
ADD COLUMN password VARCHAR(255) NOT NULL;
```

This query:

* modifies the existing `customers` table
* adds a new column:

  ```sql
  password
  ```
* stores variable-length text up to 255 characters
* prevents `NULL` values

---

### PostgreSQL vs MySQL: Column Positioning

In MySQL, developers can specify the position of newly added columns using clauses like:

* `AFTER`
* `FIRST`

Example (MySQL-only syntax):

```sql id="8t9h4p"
ALTER TABLE customers
ADD COLUMN password VARCHAR(255) NOT NULL AFTER name;
```

However, PostgreSQL does **not** support:

* `AFTER`
* `BEFORE`
* `FIRST`

Therefore, the following statements produce errors in PostgreSQL:

```sql id="8zsv4x"
ALTER TABLE customers
ADD COLUMN password VARCHAR(255) NOT NULL AFTER name;
```

```sql id="m6y6lr"
ALTER TABLE customers
ADD COLUMN password VARCHAR(255) NOT NULL BEFORE password;
```

---

### Why PostgreSQL Does Not Support Column Reordering

PostgreSQL considers column ordering primarily:

* cosmetic
* presentation-related

Internally:

* column order does not affect relational design
* query functionality remains unchanged

Therefore:

* PostgreSQL intentionally avoids unnecessary complexity related to physical column positioning.

---

### Important PostgreSQL Behaviour

In PostgreSQL:

* newly added columns always appear at the end of the table definition.

---

### Adding Multiple Columns

Multiple columns can be added in a single query.

Example:

```sql id="c2t4i5"
ALTER TABLE customers
ADD COLUMN pan_number VARCHAR(255),
ADD COLUMN joining_date TIMESTAMP NOT NULL DEFAULT CURRENT_TIMESTAMP;
```

This query adds:

1. `pan_number`
2. `joining_date`

simultaneously.

---

## DROP COLUMN

The `DROP COLUMN` clause removes columns from a table.

Example:

```sql id="wdh8ub"
ALTER TABLE customers
DROP COLUMN pan_number,
DROP COLUMN joining_date;
```

---

### What Happens Internally

When a column is dropped:

* its data is permanently deleted
* associated indexes and constraints may also be removed
* dependent objects may break

Therefore:

* `DROP COLUMN` should be used carefully.

---

### PostgreSQL Safety Feature

PostgreSQL may prevent column deletion if:

* other objects depend on the column

Example:

* views
* constraints
* indexes
* triggers

In such cases:

* PostgreSQL throws dependency errors.

---

## MODIFY COLUMN in PostgreSQL

Unlike MySQL, PostgreSQL does **not** support:

```sql id="m0xzh8"
MODIFY COLUMN
```

or:

```sql id="xmgc4y"
MODIFY CONSTRAINT
```

syntax.

Instead, PostgreSQL uses:

```sql
ALTER COLUMN
```

inside the `ALTER TABLE` statement.

---

## ALTER COLUMN Examples

### Example 1 — Change Data Type

```sql id="9r2gt0"
ALTER TABLE customers
ALTER COLUMN name TYPE VARCHAR(500);
```

This changes:

* `name`
  from:

```sql
VARCHAR(255)
```

to:

```sql
VARCHAR(500)
```

---

### Example 2 — Add NOT NULL Constraint

```sql id="o97fjr"
ALTER TABLE customers
ALTER COLUMN email SET NOT NULL;
```

---

### Example 3 — Remove NOT NULL Constraint

```sql id="z5q0ji"
ALTER TABLE customers
ALTER COLUMN email DROP NOT NULL;
```

---

### Example 4 — Set Default Value

```sql id="q8i1rj"
ALTER TABLE customers
ALTER COLUMN joining_date
SET DEFAULT CURRENT_TIMESTAMP;
```

---

### Example 5 — Remove Default Value

```sql id="rxps5w"
ALTER TABLE customers
ALTER COLUMN joining_date
DROP DEFAULT;
```

---

## ALTER TABLE for Constraints

Another important use of `ALTER TABLE` is:

* adding constraints after table creation.

This is extremely common in real-world database design.

---

## ADD CONSTRAINT

Example:

```sql id="phu2sl"
ALTER TABLE customers
ADD CONSTRAINT customer_age_check
CHECK (age > 13);
```

This:

* creates a named constraint:

  ```sql
  customer_age_check
  ```
* ensures:

  ```sql
  age > 13
  ```

---

### Why Named Constraints Are Preferred

Named constraints improve:

* debugging
* maintenance
* readability
* migration management

Example PostgreSQL error:

```text id="91x2u8"
violates check constraint "customer_age_check"
```

The constraint name immediately identifies the issue.

---

## DROP CONSTRAINT

Constraints can be removed using:

```sql id="rjlwm2"
ALTER TABLE customers
DROP CONSTRAINT customer_age_check;
```

---

## Important PostgreSQL Clarification About Constraints

This statement requires slight correction:

> "We cannot use ALTER CONSTRAINT."

In PostgreSQL:

* some limited constraint alterations are possible
* but most structural changes require:

  1. dropping the constraint
  2. recreating it

This is especially true for:

* `CHECK`
* `PRIMARY KEY`
* `UNIQUE`
* `FOREIGN KEY`

constraints.

Therefore, in practice, developers usually:

1. drop the old constraint
2. create the updated one

---

## Example Workflow for Updating a Constraint

Suppose this exists:

```sql id="7zjlwm"
CHECK (age > 13)
```

but now the requirement becomes:

```sql
age > 18
```

PostgreSQL workflow:

### Step 1 — Drop Old Constraint

```sql id="s7ymod"
ALTER TABLE customers
DROP CONSTRAINT customer_age_check;
```

---

### Step 2 — Add New Constraint

```sql id="frwrt4"
ALTER TABLE customers
ADD CONSTRAINT customer_age_check
CHECK (age > 18);
```

---

## PostgreSQL-Specific Notes

### PostgreSQL Uses ALTER More Extensively

Compared to MySQL:

* PostgreSQL relies heavily on:

  ```sql
  ALTER TABLE ... ALTER COLUMN
  ```
* instead of:

  ```sql
  MODIFY COLUMN
  ```

This makes PostgreSQL syntax:

* more explicit
* more standard-compliant
* more modular
