# SQL Joins

## What are SQL Joins?

In SQL, a **Join** is a technique used to combine data from two or more tables based on a related column between them.

Joins are used when the information required for a query is distributed across multiple tables rather than stored in a single table. By combining these tables, SQL creates a **virtual result set** containing all the necessary information without permanently merging the tables.

A join generally consists of three components:

* Two (or more) tables to be combined.
* A common column (or relationship) between those tables.
* A join condition that specifies how the tables should be matched.

For example, consider the following tables:

**Customers**

| Customer_ID | Name    |
| ----------- | ------- |
| 101         | Alice   |
| 102         | Bob     |
| 103         | Charlie |

**Orders**

| Order_ID | Customer_ID | Product |
| -------- | ----------- | ------- |
| 1        | 101         | Laptop  |
| 2        | 103         | Phone   |
| 3        | 101         | Mouse   |

Using an SQL Join, we can retrieve the customer's name along with the products they ordered.

| Name    | Product |
| ------- | ------- |
| Alice   | Laptop  |
| Alice   | Mouse   |
| Charlie | Phone   |

Without joins, retrieving this information would require multiple separate queries.

---

### Why is Data Stored in Multiple Tables?

A common question is:

> Why not store all the information in one large table?

The answer lies in **Database Normalization**, which is the process of organizing data to reduce redundancy and improve data integrity.

Suppose an online shopping website stores all information in a single table.

| Order_ID | Customer_Name | Customer_Address | Product  | Price |
| -------- | ------------- | ---------------- | -------- | ----- |
| 1        | Alice         | Delhi            | Laptop   | 60000 |
| 2        | Alice         | Delhi            | Mouse    | 800   |
| 3        | Alice         | Delhi            | Keyboard | 1500  |

Notice that **Alice's name and address are repeated in every order.**

This creates several problems:

* **Data Redundancy:** The same information is stored repeatedly, wasting storage.
* **Update Anomaly:** If Alice changes her address, every row must be updated.
* **Insertion Anomaly:** Customer information cannot be added until an order exists.
* **Deletion Anomaly:** Deleting Alice's last order also removes her customer information.

Instead, databases separate information into multiple related tables.

**Customers**

| Customer_ID | Name  | Address |
| ----------- | ----- | ------- |
| 101         | Alice | Delhi   |

**Orders**

| Order_ID | Customer_ID | Product  | Price |
| -------- | ----------- | -------- | ----- |
| 1        | 101         | Laptop   | 60000 |
| 2        | 101         | Mouse    | 800   |
| 3        | 101         | Keyboard | 1500  |

The tables are connected using **Customer_ID**, allowing SQL Joins to combine them whenever necessary.

### Benefits of Using Multiple Tables

* Eliminates duplicate data.
* Improves storage efficiency.
* Makes updates easier and more consistent.
* Preserves data integrity.
* Makes databases scalable and easier to maintain.

## Types of SQL Joins

SQL provides six major types of joins:

1. Inner Join
2. Left Join (Left Outer Join)
3. Right Join (Right Outer Join)
4. Full Outer Join
5. Cross Join
6. Self Join

Each join returns a different set of rows depending on how matching and non-matching records are handled.

---

## Cross Join (Cartesian Product)

A **Cross Join**, also known as a **Cartesian Product**, returns **every possible combination of rows** from the two tables.

Unlike other joins, a Cross Join **does not require a matching condition**.

If:

* Table A contains **m** rows
* Table B contains **n** rows

Then the resulting table will contain:

> **m × n rows**

### Example

**Students**

| Student |
| ------- |
| Alice   |
| Bob     |

**Courses**

| Course  |
| ------- |
| SQL     |
| Python  |
| Tableau |

Cross Join Result:

| Student | Course  |
| ------- | ------- |
| Alice   | SQL     |
| Alice   | Python  |
| Alice   | Tableau |
| Bob     | SQL     |
| Bob     | Python  |
| Bob     | Tableau |

Since there are:

* 2 students
* 3 courses

The result contains:

> 2 × 3 = **6 rows**

### Common Use Cases

* Generating all possible combinations.
* Creating test data.
* Product configuration systems.
* Scheduling every employee for every available shift.
* Creating all color-size combinations in an inventory.

### Important Note

Cross Joins can generate extremely large result sets. Therefore, they should be used carefully with large tables, as they may significantly impact query performance.

### SQL Example

```sql
SELECT *
FROM users1 t1
CROSS JOIN groups t2;
```

---

## Inner Join

An **Inner Join** returns **only those rows that have matching values in both tables**.

Rows without a matching value in either table are excluded from the result.

This is the most commonly used SQL join.

### Example

**Students**

| Student_ID | Name    |
| ---------- | ------- |
| 1          | Alice   |
| 2          | Bob     |
| 3          | Charlie |

**Enrollments**

| Student_ID | Course |
| ---------- | ------ |
| 1          | SQL    |
| 3          | Python |

Inner Join Result:

| Name    | Course |
| ------- | ------ |
| Alice   | SQL    |
| Charlie | Python |

Notice that **Bob** does not appear because there is no matching enrollment record.

### Common Use Cases

* Displaying customers along with their orders.
* Showing employees with their departments.
* Listing students enrolled in courses.
* Combining sales records with product information.

### SQL Example

```sql
SELECT *
FROM users1 t1
INNER JOIN membership t2
ON t1.user_id = t2.user_id;
```

### Verifying the Join

```sql
SELECT DISTINCT(t1.user_id)
FROM users1 t1
INNER JOIN membership t2
ON t1.user_id = t2.user_id;
```

This query helps verify which `user_id`s successfully matched during the join.

---

## Left Join (Left Outer Join)

A **Left Join** returns:

* All rows from the **left table**
* Matching rows from the **right table**

If no matching row exists in the right table, SQL fills the right-side columns with **NULL** values.

### Example

**Customers**

| Customer_ID | Name    |
| ----------- | ------- |
| 101         | Alice   |
| 102         | Bob     |
| 103         | Charlie |

**Orders**

| Customer_ID | Product |
| ----------- | ------- |
| 101         | Laptop  |
| 103         | Phone   |

Left Join Result:

| Name    | Product |
| ------- | ------- |
| Alice   | Laptop  |
| Bob     | NULL    |
| Charlie | Phone   |

Although Bob has not placed any orders, he still appears because every row from the left table is preserved.

### Common Use Cases

* Finding customers who have never placed an order.
* Listing employees regardless of whether they belong to a department.
* Showing all products even if they have never been sold.

### SQL Example

```sql
SELECT *
FROM users1 t1
LEFT JOIN membership t2
ON t1.user_id = t2.user_id;
```

---

## Right Join (Right Outer Join)

A **Right Join** is the exact opposite of a Left Join.

It returns:

* All rows from the **right table**
* Matching rows from the **left table**

If no matching row exists in the left table, SQL fills the left-side columns with **NULL** values.

### Example

Suppose the membership table contains a user that does not exist in the users table.

**Users**

| User_ID | Name  |
| ------- | ----- |
| 1       | Alice |
| 2       | Bob   |

**Membership**

| User_ID | Group     |
| ------- | --------- |
| 1       | Admin     |
| 3       | Moderator |

Right Join Result:

| Name  | Group     |
| ----- | --------- |
| Alice | Admin     |
| NULL  | Moderator |

The membership for User_ID = 3 is still included because every row from the right table is preserved.

### Common Use Cases

* Displaying every membership record even if user details are missing.
* Finding orphan records in relational databases.

### SQL Example

```sql
SELECT *
FROM users1 t1
RIGHT JOIN membership t2
ON t1.user_id = t2.user_id;
```

---

## Full Outer Join

A **Full Outer Join** returns:

* All matching rows from both tables.
* All unmatched rows from the left table.
* All unmatched rows from the right table.

Whenever a match is missing, SQL fills the missing columns with **NULL** values.

It can be thought of as the combination of a **Left Join** and a **Right Join**.

### Example

**Customers**

| Customer_ID | Name  |
| ----------- | ----- |
| 101         | Alice |
| 102         | Bob   |

**Orders**

| Customer_ID | Product |
| ----------- | ------- |
| 101         | Laptop  |
| 103         | Phone   |

Full Outer Join Result:

| Name  | Product |
| ----- | ------- |
| Alice | Laptop  |
| Bob   | NULL    |
| NULL  | Phone   |

The result includes:

* Matching records.
* Left-only records.
* Right-only records.

### Common Use Cases

* Comparing two datasets.
* Data migration.
* Data reconciliation.
* Identifying missing records on either side.

### SQL Example

```sql
SELECT *
FROM users1 t1
FULL OUTER JOIN membership t2
ON t1.user_id = t2.user_id;
```

---

## Self Join

A **Self Join** is a join in which a table is joined with itself.

Instead of joining two different tables, SQL treats the same table as two separate instances by assigning different aliases.

Self Joins are useful whenever relationships exist **within the same table**.

### Example

Consider an employee table where each employee stores the ID of their manager.

| Employee_ID | Employee | Manager_ID |
| ----------- | -------- | ---------- |
| 1           | Alice    | NULL       |
| 2           | Bob      | 1          |
| 3           | Charlie  | 1          |
| 4           | David    | 2          |

Using a Self Join, we can retrieve each employee along with their manager's name.

| Employee | Manager |
| -------- | ------- |
| Bob      | Alice   |
| Charlie  | Alice   |
| David    | Bob     |

Without a Self Join, retrieving this information from the same table would not be possible.

### Common Use Cases

* Employee-manager relationships.
* Organizational hierarchy.
* Family tree relationships.
* Finding employees working under the same manager.
* Comparing salaries of employees within the same department.
* Finding customers who share the same billing address.

### SQL Example

```sql
SELECT *
FROM users1 t1
JOIN users1 t2
ON t1.emergency_contact = t2.user_id;
```

Here, `users1` is treated as two logical tables:

* `t1` represents each user.
* `t2` represents the user's emergency contact.

The join matches `emergency_contact` with `user_id`, allowing user details to be retrieved alongside the corresponding emergency contact information.

## SQL Set Operations

**Set Operations** are SQL operators that combine the results of two or more `SELECT` statements into a single result set. They are particularly useful when working with tables that have similar structures or when comparing datasets.

Unlike **Joins**, which combine tables **horizontally** (adding columns), **Set Operations** combine result sets **vertically** (adding rows).

### Conditions for Using Set Operations

Before using any set operation, the participating `SELECT` statements must satisfy the following conditions:

* Both queries must return the **same number of columns**.
* The corresponding columns must have **compatible data types**.
* The columns should represent similar kinds of data.
* The column names in the final output are taken from the **first `SELECT` statement**.

---

### Types of SQL Set Operations

SQL provides four primary set operations:

1. **UNION**
2. **UNION ALL**
3. **INTERSECT**
4. **EXCEPT** (called `MINUS` in Oracle)

---

## UNION

The **UNION** operator combines the results of two or more `SELECT` statements into a single result set while **automatically removing duplicate rows**.

Think of it as finding the **unique union** of two datasets.

### Example

Suppose we have two tables.

**person1**

| Name    |
| ------- |
| Alice   |
| Bob     |
| Charlie |

**person2**

| Name    |
| ------- |
| Bob     |
| Charlie |
| David   |

Query:

```sql
SELECT Name FROM person1
UNION
SELECT Name FROM person2;
```

Result:

| Name    |
| ------- |
| Alice   |
| Bob     |
| Charlie |
| David   |

Notice that **Bob** and **Charlie** appear only once because `UNION` removes duplicates.

### Common Use Cases

* Combining customer lists from multiple branches.
* Merging employee records from different offices.
* Creating a list of unique values from multiple tables.

---

## UNION ALL

The **UNION ALL** operator is similar to `UNION`, except that it **does not remove duplicate rows**.

Since SQL does not need to check for duplicates, `UNION ALL` is generally **faster** than `UNION`.

### Example

Using the same tables:

**person1**

| Name    |
| ------- |
| Alice   |
| Bob     |
| Charlie |

**person2**

| Name    |
| ------- |
| Bob     |
| Charlie |
| David   |

Query:

```sql
SELECT Name FROM person1
UNION ALL
SELECT Name FROM person2;
```

Result:

| Name    |
| ------- |
| Alice   |
| Bob     |
| Charlie |
| Bob     |
| Charlie |
| David   |

Unlike `UNION`, duplicate rows are preserved.

### Common Use Cases

* Combining transaction logs.
* Merging sales records.
* Appending datasets where duplicates are meaningful.

---

## INTERSECT

The **INTERSECT** operator returns **only the rows that are common to both result sets**.

Only records that exist in **both** queries are included.

### Example

**person1**

| Name    |
| ------- |
| Alice   |
| Bob     |
| Charlie |

**person2**

| Name    |
| ------- |
| Bob     |
| Charlie |
| David   |

Query:

```sql
SELECT Name FROM person1
INTERSECT
SELECT Name FROM person2;
```

Result:

| Name    |
| ------- |
| Bob     |
| Charlie |

Only the names present in **both tables** are returned.

### Common Use Cases

* Finding customers who purchased from two different stores.
* Identifying students enrolled in multiple courses.
* Comparing common records between two datasets.

---

## EXCEPT (MINUS)

The **EXCEPT** operator (called **MINUS** in Oracle) returns the rows that exist in the **first result set but not in the second**.

It performs a **set difference** operation and returns only distinct rows.

### Example

**person1**

| Name    |
| ------- |
| Alice   |
| Bob     |
| Charlie |

**person2**

| Name    |
| ------- |
| Bob     |
| Charlie |
| David   |

Query:

```sql
SELECT Name FROM person1
EXCEPT
SELECT Name FROM person2;
```

Result:

| Name  |
| ----- |
| Alice |

Since **Alice** exists only in `person1`, she is returned.

Similarly,

```sql
SELECT Name FROM person2
EXCEPT
SELECT Name FROM person1;
```

would return:

| Name  |
| ----- |
| David |

### Common Use Cases

* Finding customers who have not placed any orders.
* Identifying products not sold during a particular period.
* Detecting records that exist in one dataset but not another.

---

## Quick Comparison of Set Operations

| Operator           | Removes Duplicates | Returns                                                         |
| ------------------ | ------------------ | --------------------------------------------------------------- |
| **UNION**          | Yes                | All unique rows from both result sets                           |
| **UNION ALL**      | No                 | All rows from both result sets, including duplicates            |
| **INTERSECT**      | Yes                | Only the rows common to both result sets                        |
| **EXCEPT (MINUS)** | Yes                | Rows present in the first result set but absent from the second |

---

## SQL Examples

```sql
-- UNION
SELECT *
FROM person1
UNION
SELECT *
FROM person2;

-- UNION ALL
SELECT *
FROM person1
UNION ALL
SELECT *
FROM person2;

-- INTERSECT
SELECT *
FROM person1
INTERSECT
SELECT *
FROM person2;

-- EXCEPT
SELECT *
FROM person1
EXCEPT
SELECT *
FROM person2;
```

### Difference Between SQL Joins and Set Operations

| SQL Joins                                                  | SQL Set Operations                                                                                |
| ---------------------------------------------------------- | ------------------------------------------------------------------------------------------------- |
| Combine tables horizontally by adding columns.             | Combine result sets vertically by adding rows.                                                    |
| Require a relationship (join condition) between tables.    | Require compatible `SELECT` statements with the same number of columns and compatible data types. |
| Used to retrieve related information from multiple tables. | Used to merge, compare, or filter complete result sets.                                           |
| Examples: `INNER JOIN`, `LEFT JOIN`, `RIGHT JOIN`.         | Examples: `UNION`, `UNION ALL`, `INTERSECT`, `EXCEPT`.                                            |

**In simple terms:**

* **Joins** answer the question: *"How can I combine related information from different tables?"*
* **Set Operations** answer the question: *"How can I combine or compare the results of multiple queries?"*

## Joining Multiple Tables

In real-world databases, the information required for a query is often spread across **more than two tables**. In such cases, SQL allows us to join multiple tables together in a single query.

Each additional table is joined using its relationship with one of the previously joined tables.

This technique is commonly used in normalized databases, where each table stores a specific type of information.

### Why Do We Need Multiple Joins?

Consider an e-commerce database with the following tables:

* **Users** → Stores customer information.
* **Orders** → Stores order details and references the customer using `user_id`.
* **Order_Details** → Stores the products purchased in each order and references the order using `order_id`.

Suppose we want to answer the following question:

> **"Which customer purchased which products?"**

No single table contains all this information.

The relationships are:

```
Users
   │
   │ user_id
   ▼
Orders
   │
   │ order_id
   ▼
Order_Details
```

Since:

* `Users` and `Order_Details` **do not share a common column**, they cannot be joined directly.
* The `Orders` table acts as an **intermediate (bridge) table**, connecting the two.

Therefore, we perform the join in two steps:

1. Join **Order_Details** with **Orders** using `order_id`.
2. Join the resulting table with **Users** using `user_id`.

### SQL Example

```sql
SELECT *
FROM order_details_2 t1
JOIN orders t2
ON t1.order_id = t2.order_id
JOIN users t3
ON t2.user_id = t3.user_id;
```

### How the Query Works

1. `order_details_2` is joined with `orders` using `order_id`.
2. The resulting table now contains both `order_id` and `user_id`.
3. This intermediate result is then joined with `users` using `user_id`.
4. The final result contains information from all three tables.

### Key Takeaway

When two tables cannot be joined directly because they do not share a common key, look for an **intermediate table** that relates to both. By chaining multiple joins together, SQL can retrieve information distributed across an entire relational database.

## Filtering Columns After a Join

When performing joins, it is **not necessary to retrieve every column** from the participating tables.

Using `SELECT *` is useful during exploration, but in production queries it is generally better to select only the columns that are actually needed.

Selecting specific columns offers several advantages:

* Improves query readability.
* Reduces unnecessary data transfer.
* Makes the result easier to understand.
* Can improve query performance, especially for large tables.

### Example 1

Retrieve the order details along with the customer's name.

```sql
SELECT t1.order_id,
       t1.amount,
       t1.profit,
       t3.name
FROM order_details_2 t1
JOIN orders t2
ON t1.order_id = t2.order_id
JOIN users t3
ON t2.user_id = t3.user_id;
```

Result:

| Order ID | Amount | Profit | Customer Name |
| -------- | ------ | ------ | ------------- |
| 101      | 1200   | 250    | Alice         |
| 102      | 800    | 120    | Bob           |

Only the required columns are displayed instead of every column from all three tables.

### Example 2

Retrieve each order along with the customer's name and city.

```sql
SELECT t1.order_id,
       t2.name,
       t2.city
FROM orders t1
JOIN users t2
ON t1.user_id = t2.user_id;
```

Result:

| Order ID | Name  | City  |
| -------- | ----- | ----- |
| 101      | Alice | Delhi |
| 102      | Bob   | Pune  |

## Filtering Rows After a Join

After joining tables, the `WHERE` clause can be used to filter the rows returned by the query.

The join first combines the tables, and then the `WHERE` clause filters the resulting dataset.

### Example 1: Find All Orders Placed in Pune

```sql
SELECT *
FROM orders t1
JOIN users t2
ON t1.user_id = t2.user_id
WHERE t2.city = 'Pune';
```

**Explanation**

1. `orders` is joined with `users`.
2. The `WHERE` clause filters only those rows where the customer's city is **Pune**.

---

### Example 2: Find All Orders Under the "Chairs" Category

```sql
SELECT *
FROM order_details_2 t1
JOIN category t2
ON t1.category_id = t2.category_id
WHERE t2.vertical = 'Chairs';
```

**Explanation**

The query retrieves only those orders whose category belongs to the **Chairs** product vertical.

## Practice Questions Using Joins

The following examples demonstrate how joins can be combined with aggregate functions, grouping, sorting, and filtering to answer real-world business questions.

---

### 1. Find All Profitable Orders

```sql
SELECT t1.order_id,
       SUM(profit)
FROM orders t1
JOIN order_details_2 t2
ON t1.order_id = t2.order_id
GROUP BY t1.order_id
HAVING SUM(profit) > 0;
```

**Explanation**

* Join each order with its order details.
* Calculate the total profit for every order.
* Return only those orders whose total profit is greater than zero.

---

### 2. Find the Customer Who Has Placed the Most Orders

```sql
SELECT t2.user_id,
       COUNT(*),
       t2.name,
       t2.city,
       t2.state
FROM orders t1
JOIN users t2
ON t1.user_id = t2.user_id
GROUP BY t2.user_id, name, state, city
ORDER BY COUNT(*) DESC
LIMIT 1;
```

**Explanation**

* Join orders with customers.
* Count the number of orders placed by each customer.
* Sort the customers in descending order of order count.
* Return the customer with the highest number of orders.

---

### 3. Find the Most Profitable Category

```sql
SELECT vertical,
       SUM(profit)
FROM order_details_2 t1
JOIN category t2
ON t1.category_id = t2.category_id
GROUP BY t2.vertical
ORDER BY SUM(profit) DESC
LIMIT 1;
```

**Explanation**

* Join order details with product categories.
* Calculate the total profit generated by each category.
* Return the category with the highest profit.

---

### 4. Find the Most Profitable State

```sql
SELECT state,
       SUM(profit)
FROM orders t1
JOIN order_details_2 t2
ON t1.order_id = t2.order_id
JOIN users t3
ON t1.user_id = t3.user_id
GROUP BY state
ORDER BY SUM(profit) DESC
LIMIT 1;
```

**Explanation**

This query demonstrates joining **three tables**:

* `Orders`
* `Order_Details`
* `Users`

The total profit is calculated for each state, and the state with the highest overall profit is returned.

---

### 5. Find All Categories with Profit Greater Than 5000

```sql
SELECT t1.category_id,
       category,
       SUM(profit)
FROM category t1
JOIN order_details_2 t2
ON t1.category_id = t2.category_id
GROUP BY t1.category_id, category
HAVING SUM(profit) > 5000;
```

**Explanation**

* Join categories with order details.
* Calculate the total profit generated by each category.
* Return only those categories whose cumulative profit exceeds **5000**.

---

## Summary

By combining joins with clauses such as `SELECT`, `WHERE`, `GROUP BY`, `HAVING`, `ORDER BY`, and aggregate functions, SQL can answer complex analytical questions involving data spread across multiple related tables. This combination forms the foundation of many real-world SQL queries used in reporting, business intelligence, and data analytics.
