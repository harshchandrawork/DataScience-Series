# SQL Case Study: Zomato Data Analysis

## Project Overview
This case study demonstrates the practical application of Structured Query Language (SQL) in solving real-world business problems within the food delivery sector.

## Dataset & Methodology
- Data Source: The analysis is performed on a synthesized, simulated dataset modeled after Zomato.
- Authenticity: Please note that while the dataset is entirely fictional and created for demonstration purposes, it accurately mirrors the schema, complexities, and data structures found in production-level applications.
- Objective: To showcase how relational databases can be queried to extract actionable business insights, track key performance indicators (KPIs), and analyze user behavior.
