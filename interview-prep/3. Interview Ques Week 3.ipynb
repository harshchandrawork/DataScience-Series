## ğŸ”¹ **How Indexing Works in Sets? Understanding Hashing**  

Python sets **do not support traditional indexing** like lists or tuples because they rely on **hashing** to store elements. This makes sets **unordered and unindexed**, meaning elements are not stored in a predictable order.

---

### **ğŸ”¹ Why Do Sets Use Hashing?**  
ğŸ”¹ Unlike lists and tuples, which use **sequential memory storage**, sets rely on **hashing** for:  
âœ” **Faster lookups** (`O(1)` time complexity for searching).  
âœ” **Efficient insertion and deletion** (`O(1)` on average).  
âœ” **No duplicate elements** (hashing ensures uniqueness).  

> âš  **Since sets use hashing, their order may change every time a program runs.**

---

### **ğŸ”¹ Understanding Hashing**
ğŸ”¹ Hashing is a process that **maps data to a fixed-size memory location** using a **hash function**.  
âœ” A **hash function** takes an input (element) and produces a **unique index (hash value)**.  
âœ” This hash value determines where the element is stored in **hash tables**.  

#### **ğŸ”¹ Example: Simple Hashing Function**  
Letâ€™s consider a **hash function**:
\[
\text{hash(value)} = value \mod 5
\]
This function maps an element to an index between **0 and 4**.

##### **ğŸ”¹ How elements are stored using hashing?**
| Element | Hash Calculation (`value % 5`) | Stored at Index |
|---------|------------------------|----------------|
| 12 | `12 % 5 = 2` | Index `2` |
| 25 | `25 % 5 = 0` | Index `0` |
| 19 | `19 % 5 = 4` | Index `4` |
| 32 | `32 % 5 = 2` | **Collision at Index `2`** |

---

### **ğŸ”¹ Collision Handling in Sets**
When two elements **produce the same hash value**, a **collision** occurs.  
Python uses **chaining (linked lists)** and **open addressing (probing)** to handle this.

#### **ğŸ”¹ 1ï¸âƒ£ Chaining (Linked List Approach)**
When a collision occurs, elements are stored **as linked lists** at the same index.

âœ… **Example:**  
\[
\begin{array}{|c|c|}
\hline
\text{Index} & \text{Stored Values} \\
\hline
0 & 25 \\
1 & - \\
2 & \mathbf{12 \rightarrow 32} \text{ (Chained)} \\
3 & - \\
4 & 19 \\
\hline
\end{array}
\]

ğŸ”¹ **Advantage:** Can store multiple elements at the same index.  
ğŸ”¹ **Disadvantage:** **Slower lookups** (`O(k)`, where `k` is the length of the chain).  

---

#### **ğŸ”¹ 2ï¸âƒ£ Open Addressing (Probing)**
Instead of chaining, probing **finds an alternative empty slot** to store the element.

##### **ğŸ”¹ Example of Linear Probing**
If `32` collides at **index 2**, try **index 3, then 4, etc.**, until an empty slot is found.

| Attempt | Index |
|---------|------|
| `hash(32) = 2` (Collision) | âŒ 2 (Occupied) |
| Try next index (Linear Probing) | âœ… 3 (Empty) |

âœ” **New Table After Linear Probing:**
\[
\begin{array}{|c|c|}
\hline
\text{Index} & \text{Stored Values} \\
\hline
0 & 25 \\
1 & - \\
2 & 12 \\
3 & 32 \text{ (Probed)} \\
4 & 19 \\
\hline
\end{array}
\]

ğŸ”¹ **Advantage:** Faster retrieval in some cases.  
ğŸ”¹ **Disadvantage:** Can lead to **clustering** (elements grouped together).  

> **Other Probing Methods:** Quadratic Probing, Double Hashing.

---

### **ğŸ”¹ Heterogeneous Data in Sets**
Python **supports heterogeneous data types** in sets, meaning a set can contain **integers, strings, floats, etc.**.

âœ” **Example:**  
```python
my_set = {10, "apple", 5.5, (1, 2)}
```
ğŸ”¹ Python ensures **hash functions work across different data types** by **converting** each element to a **unique hashable form**.

âœ” **Hashing in action:**
```python
print(hash(10))        # Hash value for an integer
print(hash("apple"))   # Hash value for a string
print(hash(5.5))       # Hash value for a float
print(hash((1, 2)))    # Hash value for a tuple
```

> âš  **Lists, dictionaries, and sets are unhashable** because they are **mutable**.

---

### **ğŸ”¹ Why Use Hashing Instead of Lists/Tuples?**
| **Feature** | **List** | **Tuple** | **Set (Hashing)** |
|------------|--------|--------|----------------|
| **Storage Order** | Sequential | Sequential | Random (Based on hash function) |
| **Duplicates Allowed?** | âœ… Yes | âœ… Yes | âŒ No |
| **Lookup Time Complexity** | `O(n)` | `O(n)` | `O(1)` (Best Case) |
| **Insertion Complexity** | `O(1)` (Append) | âŒ Immutable | `O(1)` |
| **Mutable?** | âœ… Yes | âŒ No | âœ… Yes |

âœ” **Hashing is used for fast lookup and unique storage**.  
âœ” **Lists/Tuples are better for ordered data**.

---

### **ğŸ”¹ Key Takeaways**
âœ… **Sets use hashing**, making them **unordered and unindexed**.  
âœ… **Hashing provides fast lookups** (`O(1)`) but requires **collision handling**.  
âœ… **Collisions are resolved using Chaining (Linked Lists) or Probing (Open Addressing)**.  
âœ… **Sets can store heterogeneous data** by ensuring all elements are hashable.  
âœ… **Lists and tuples store data sequentially**, while **sets use hashing** for efficiency.  

## ğŸ”¹ **Why Dictionary Keys Cannot Be of Mutable Data Types?**  

In Python, **dictionary keys must be immutable** because they are **hashed** using a **hash function** to determine their storage location. If a mutable object were allowed as a key, it could lead to **unexpected behavior and errors**.  

---

### **ğŸ”¹ The Role of Hashing in Dictionary Keys**  
Python dictionaries use **hash tables**, which rely on a **hash function** to store key-value pairs efficiently.  

âœ” **How it works:**  
- The key is passed through a **hash function** to compute a **unique hash value**.  
- This **hash value** determines where the key-value pair is stored in memory.  
- When retrieving a value, Python re-hashes the key to find its location quickly (`O(1)` lookup time).  

> ğŸ”¹ **Immutable objects (e.g., strings, numbers, tuples) are hashable** and produce a **consistent hash value**.  

---

### **ğŸ”¹ What Happens If a Mutable Object is Used as a Key?**  
Mutable objects (like **lists, sets, or dictionaries**) **can change after creation**. If a dictionary key were mutable:  

1ï¸âƒ£ **Hash Value Instability:**  
   - Since hashing is based on the keyâ€™s content, changing a mutable key **would change its hash value**.  
   - The dictionary would **lose track** of where the value was stored, leading to retrieval failures.  

2ï¸âƒ£ **Data Corruption Risk:**  
   - A dictionary relies on **fixed hash values** for keys. If a key changes, Python **wouldnâ€™t be able to locate the value** stored under the original hash.  

3ï¸âƒ£ **Duplicate Key Issue:**  
   - Dictionaries enforce **unique keys**.  
   - If a key could be modified, multiple different-looking keys might **hash to the same value**, breaking dictionary integrity.  

---

### **ğŸ”¹ Example: Using a Mutable Key (Fails)**  
```python
my_dict = { [1, 2, 3]: "value" }  # âŒ TypeError: unhashable type: 'list'
```
ğŸ”¹ **Why does this fail?**  
- Lists are **mutable**, meaning their content can change.  
- Python prevents this to ensure **dictionary integrity**.  

---

### **ğŸ”¹ Example: Using an Immutable Key (Works)**  
```python
my_dict = { (1, 2, 3): "value" }  # âœ… Works
print(my_dict[(1, 2, 3)])  # Output: value
```
âœ” **Tuples are immutable**, so their hash value **remains the same**.

> **âš  Exception:** If a **tuple contains mutable elements**, it becomes **unhashable**:  
```python
my_dict = { (1, 2, [3, 4]): "value" }  # âŒ TypeError
```
âœ” **Reason:** The list `[3, 4]` inside the tuple is **mutable**.

---

### **ğŸ”¹ Ensuring Uniqueness and Efficient Lookup**  
âœ” Dictionaries use **hashing to prevent duplicate keys**.  
âœ” If mutable objects were allowed, different-looking keys **might hash to the same location** after modification.  

> âœ… **Immutable keys ensure fast lookups and data integrity.**  

---

#### **ğŸ”¹ Key Takeaways**
âœ… **Dictionary keys must be immutable** because they are passed to a **hash function**.  
âœ… **Mutable objects (lists, sets, dictionaries) cannot be keys** as they can change after creation.  
âœ… **Changing a mutable key would break hash consistency**, making dictionary lookups unreliable.  
âœ… **Tuples work as keys**, but only if they contain **only immutable elements**.

## ğŸ”¹ **`enumerate()` in Python**  

### **ğŸ”¹ What is `enumerate()`?**  
`enumerate()` is a built-in Python function that **associates a counter with an iterable**, allowing you to **iterate over both the index and the element** at the same time.  

âœ” **Instead of using a separate counter variable**, `enumerate()` automatically provides an **index** while looping through a list, tuple, string, or any other iterable.

---

### **ğŸ”¹ Syntax of `enumerate()`**
```python
enumerate(iterable, start=0)
```
âœ” **Parameters:**  
- `iterable`: Any sequence (list, tuple, string, etc.) that needs to be iterated over.  
- `start`: (Optional) The starting value of the counter (default is `0`).  

âœ” **Returns:**  
- An `enumerate` object (which is an **iterator** yielding `(index, element)` pairs).  

---

### **ğŸ”¹ Basic Example**
```python
fruits = ["apple", "banana", "cherry"]

for index, fruit in enumerate(fruits):
    print(index, fruit)
```
âœ” **Output:**
```
0 apple
1 banana
2 cherry
```
ğŸ”¹ **Here, `enumerate()` keeps track of the index automatically**, removing the need for a manual counter.

---

### **ğŸ”¹ Using `enumerate()` with a Custom Start Value**
```python
fruits = ["apple", "banana", "cherry"]

for index, fruit in enumerate(fruits, start=1):
    print(index, fruit)
```
âœ” **Output:**
```
1 apple
2 banana
3 cherry
```
ğŸ”¹ **Setting `start=1` makes the counter begin from 1 instead of 0.**  

---

### **ğŸ”¹ Practical Use Cases of `enumerate()`**
#### âœ… **1. Replacing `range(len(iterable))` (More Readable)**
```python
# Without enumerate
for i in range(len(fruits)):
    print(i, fruits[i])

# With enumerate (More Pythonic)
for index, fruit in enumerate(fruits):
    print(index, fruit)
```
âœ” **Using `enumerate()` makes the code cleaner and more readable**.

---

#### âœ… **2. Using `enumerate()` with Conditional Statements**
```python
numbers = [10, 20, 30, 40]

for index, num in enumerate(numbers):
    if num == 30:
        print(f"Found {num} at index {index}")
```
âœ” **Output:**
```
Found 30 at index 2
```
ğŸ”¹ **Useful for finding elements and their positions in a list.**  

---

#### âœ… **3. Creating a Dictionary from an Iterable**
```python
fruits = ["apple", "banana", "cherry"]

fruit_dict = dict(enumerate(fruits, start=1))
print(fruit_dict)
```
âœ” **Output:**
```python
{1: 'apple', 2: 'banana', 3: 'cherry'}
```
ğŸ”¹ **Quickly converts an iterable into a dictionary where keys are indexes.**  

---

#### âœ… **4. Using `enumerate()` with List Comprehension**
```python
squares = [num**2 for index, num in enumerate(range(5))]
print(squares)
```
âœ” **Output:**
```
[0, 1, 4, 9, 16]
```
ğŸ”¹ **Combining `enumerate()` with list comprehension for concise iteration.**  

---

### **ğŸ”¹ Key Takeaways**
âœ” `enumerate()` is used to iterate over an iterable while keeping track of the index.  
âœ” It removes the need to manually maintain a counter with `range(len(iterable))`.  
âœ” It returns an `enumerate` object that can be converted into a `list` or `dictionary`.  
âœ” `enumerate()` improves readability and makes the code more Pythonic.  

## ğŸ”¹ **Destructor in Python**  

### **ğŸ”¹ What is a Destructor?**  
A **destructor** is a special method in Python that is called **when an object is about to be destroyed or deleted from memory**. It is the **exact opposite of a constructor (`__init__()`)**, which is used to initialize an object.  

âœ” **A destructor is useful for cleaning up resources** such as:  
- **Closing a file** that was opened.  
- **Releasing database connections.**  
- **Freeing up memory allocated to an object.**  
- **Logging the deletion of an object.**  

---

### **ğŸ”¹ How to Define a Destructor?**  
A destructor is defined using **`__del__()`** method in a class.

#### **ğŸ”¹ Basic Example**
```python
class Example:
    def __init__(self):
        print("Constructor: Object is created.")
    
    def __del__(self):
        print("Destructor: Object is deleted.")

# Creating an object
obj = Example()

# Deleting the object manually
del obj
```
âœ” **Output:**  
```
Constructor: Object is created.  
Destructor: Object is deleted.
```
ğŸ”¹ **Here, the `__del__()` method is automatically called when `del obj` is executed.**  

---

### **ğŸ”¹ Destructor vs Constructor**
| Feature          | Constructor (`__init__()`) | Destructor (`__del__()`) |
|-----------------|--------------------------|--------------------------|
| **Purpose**     | Initializes an object    | Cleans up before object deletion |
| **Called When?**| Object is created        | Object is deleted from memory |
| **Common Use**  | Setting up resources (e.g., opening files, connecting to databases) | Releasing resources (e.g., closing files, disconnecting from databases) |

---

### **ğŸ”¹ When is a Destructor Called?**
âœ” A destructor is **automatically called when an object is deleted** using `del` or when all references to the object are removed.

#### âœ… **Example: Destructor Called Automatically**
```python
class Example:
    def __del__(self):
        print("Destructor is called.")

def create_object():
    obj = Example()

# Object is created inside function
create_object()

# As function ends, obj goes out of scope, so destructor is called
```
âœ” **Output:**  
```
Destructor is called.
```
ğŸ”¹ **Here, the destructor is triggered automatically as `obj` goes out of scope.**  

---

### **ğŸ”¹ Destructor Use Case: Releasing Resources**
âœ” Destructors **ensure proper cleanup of resources**, preventing memory leaks.

#### âœ… **Example: Closing a File in a Destructor**
```python
class FileHandler:
    def __init__(self, filename):
        self.file = open(filename, 'w')
        print("File opened.")
    
    def __del__(self):
        self.file.close()
        print("File closed.")

# Creating an instance
file_obj = FileHandler("test.txt")

# Deleting the object, which triggers the destructor
del file_obj
```
âœ” **Output:**  
```
File opened.
File closed.
```
ğŸ”¹ **Here, `__del__()` ensures that the file is closed before the object is deleted.**  

---

### **ğŸ”¹ Destructor and Database Connection**
âœ” If an object connects to a database in `__init__()`, then `__del__()` can **ensure that the connection is closed** before deletion.

#### âœ… **Example: Closing a Database Connection**
```python
class Database:
    def __init__(self):
        print("Connecting to database...")
    
    def __del__(self):
        print("Closing database connection...")

db = Database()  # Constructor runs
del db           # Destructor runs
```
âœ” **Output:**  
```
Connecting to database...
Closing database connection...
```
ğŸ”¹ **Ensures that the database connection is closed when the object is deleted.**  

---

### **ğŸ”¹ When Will a Destructor NOT Be Called?**
âœ” **A destructor will not be called until all references to an object are deleted.**  
âœ” **If multiple variables refer to the same object, Python will only delete the object when the last reference is removed.**

#### âœ… **Example: Destructor Delayed Due to Multiple References**
```python
class Example:
    def __del__(self):
        print("Destructor called.")

obj1 = Example()
obj2 = obj1  # obj2 also refers to the same object

del obj1  # Destructor NOT called, because obj2 still refers to it
del obj2  # Now, destructor is called
```
âœ” **Output:**  
```
Destructor called.
```
ğŸ”¹ **Destructor is not called when `obj1` is deleted because `obj2` still references the same object.**  

---

### **ğŸ”¹ Are Destructors Useful in Data Science?**
âœ” **In Data Science & Machine Learning, destructors are not widely used.**  
âœ” Most libraries (like NumPy, Pandas, TensorFlow) **automatically manage memory**.  
âœ” **Garbage collection (GC) in Python handles most memory cleanup automatically.**  

However, destructors **can still be useful for:**
- Closing database connections (e.g., PostgreSQL, MySQL).
- Freeing GPU memory in deep learning (e.g., TensorFlow `sess.close()`).
- Cleaning up temporary files.

---

### **ğŸ”¹ Key Takeaways**
âœ” **A destructor is a special method (`__del__()`) used to clean up before an object is deleted.**  
âœ” It **opposes** a constructor (`__init__()`), which initializes an object.  
âœ” It is **called automatically when an object is deleted** or goes out of scope.  
âœ” It **ensures resources like files and database connections are properly closed**.  
âœ” **Python's garbage collector handles most cleanup, making destructors less necessary in Data Science.**

#### âœ… **Example: Using `issubclass()`**

#### ğŸ”¹ **Key Uses of `issubclass()`**
âœ” Useful **when working with inheritance-based class hierarchies**.  
âœ” Helps in **debugging and class-type validation**.  
âœ” **Rarely used in Data Science** but useful in **OOP-based applications**.

## ğŸ”· **Underscores in Python (`_var`, `var_`, `__var`, `__var__`, `_`)**

Python uses underscores in variable and method names for **specific purposes**, ranging from **naming conventions** to **name mangling** for encapsulation.

---

### **1ï¸âƒ£ Single Leading Underscore (`_var`)**
ğŸ“Œ **Used for indicating a "private" variable (convention only, not enforced).**  
- Itâ€™s a **soft indication** that the variable/method is **intended for internal use**.  
- However, Python **does not prevent** access to it.

#### âœ… **Example**
```python
class Test:
    def __init__(self):
        self.foo = 11       # Public variable
        self._bar = 23      # "Private" variable (by convention)

t = Test()
print(t.foo)   # âœ… Prints: 11
print(t._bar)  # âœ… Prints: 23 (accessible, but not recommended)
```
ğŸ’¡ **Key Takeaway**: The single underscore **does not** prevent access but signals that it **should not be accessed directly**.

---

### **2ï¸âƒ£ Single Trailing Underscore (`var_`)**
ğŸ“Œ **Used to avoid naming conflicts with Python keywords.**  
- If a variable name conflicts with a Python keyword (e.g., `class`, `def`), add an underscore (`_`) at the end.

#### âœ… **Example**
```python
def make_object(name, class_):  # 'class' is a keyword, so we use 'class_'
    pass
```
ğŸ’¡ **Key Takeaway**: A single trailing underscore is used **only to avoid conflicts with reserved keywords**.

---

### **3ï¸âƒ£ Double Leading Underscore (`__var`)**
ğŸ“Œ **Triggers name mangling to prevent accidental name conflicts in subclasses.**  
- **Python automatically renames** variables starting with double underscores.
- This helps **avoid unintended overriding** in subclasses.

#### âœ… **Example**
```python
class Test:
    def __init__(self):
        self.foo = 11
        self._bar = 23
        self.__baz = 42  # Double underscore variable

t = Test()
print(dir(t))  # Check available attributes
```
âœ” **Output of `dir(t)`** (simplified):
```
['_Test__baz', '_bar', 'foo']
```
Notice that `__baz` is now `_Test__baz` due to **name mangling**.

ğŸ’¡ **Key Takeaway**: Name mangling prevents **accidental modification in subclasses**.

---

### **4ï¸âƒ£ Double Leading & Trailing Underscore (`__var__`)**
ğŸ“Œ **Used for Python's built-in methods and magic methods (Dunder methods).**  
- These are special methods that Python automatically calls in certain situations.

#### âœ… **Examples**
| Dunder Method | Purpose |
|--------------|---------|
| `__init__` | Constructor (called during object creation) |
| `__str__`  | Defines string representation (`str(obj)`) |
| `__repr__` | Used by `repr(obj)` for debugging representation |
| `__call__` | Makes an object callable like a function |
| `__add__` | Defines behavior for `+` operator (`obj1 + obj2`) |

```python
class Test:
    def __init__(self, value):
        self.value = value

    def __str__(self):
        return f"Value: {self.value}"

t = Test(10)
print(str(t))  # âœ… Calls __str__ method automatically
```
âœ” **Output**:
```
Value: 10
```

ğŸ’¡ **Key Takeaway**: Avoid defining your own `__var__` methods unless you're overriding a built-in **magic method**.

---

### **5ï¸âƒ£ Single Underscore (`_`)**
ğŸ“Œ **Used in two ways:**
1. **To indicate an insignificant variable** (placeholder).
2. **Holds the result of the last expression in an interactive Python session (REPL).**

#### âœ… **Example 1: Ignoring Values**
```python
for _, value in enumerate(["A", "B", "C"]):
    print(value)  # âœ… Only uses 'value', ignores '_'
```

## ğŸ”® **Magic Methods: `__repr__` vs `__str__` in Python**  

Python provides **magic methods** (also called **dunder methods**) to customize object behavior. Among them, **`__repr__` and `__str__`** define how objects are represented as strings.  

| **Method** | **Purpose** | **Usage** |
|-----------|------------|-----------|
| `__repr__` | **For developers** (Unambiguous, debugging, logging) | `repr(obj)` or just typing `obj` in an interactive session |
| `__str__` | **For users** (Readable, user-friendly output) | `str(obj)` or `print(obj)` |

---

### **1ï¸âƒ£ `__repr__` (Developer-Oriented)**
ğŸ“Œ **Purpose:**  
- Provides an **unambiguous, detailed, and technical representation** of an object.  
- Used for **debugging and logging** to reconstruct the object in code.  

ğŸ“Œ **Rule of Thumb:**  
> "If you type `obj` in an interactive Python session, it should return a valid Python expression to recreate the object."

#### âœ… **Example**
```python
class Car:
    def __init__(self, brand, model, year):
        self.brand = brand
        self.model = model
        self.year = year

    def __repr__(self):
        return f"Car('{self.brand}', '{self.model}', {self.year})"

car1 = Car("Toyota", "Corolla", 2022)

print(repr(car1))  # OR just type car1 in an interactive session
```
âœ” **Output:**
```
Car('Toyota', 'Corolla', 2022)
```
ğŸ’¡ **Key Takeaway:** `repr(obj)` **should look like valid Python code** that can be used to reconstruct the object.

---

### **2ï¸âƒ£ `__str__` (User-Oriented)**
ğŸ“Œ **Purpose:**  
- Provides a **human-readable, user-friendly** representation of the object.  
- Used when printing output for **end-users**.  

ğŸ“Œ **Rule of Thumb:**  
> "`__str__` should be easy to read and understand."

#### âœ… **Example**
```python
class Car:
    def __init__(self, brand, model, year):
        self.brand = brand
        self.model = model
        self.year = year

    def __repr__(self):
        return f"Car('{self.brand}', '{self.model}', {self.year})"

    def __str__(self):
        return f"{self.year} {self.brand} {self.model}"

car1 = Car("Toyota", "Corolla", 2022)

print(str(car1))  # OR print(car1)
```
âœ” **Output:**
```
2022 Toyota Corolla
```
ğŸ’¡ **Key Takeaway:** `str(obj)` provides a **more readable and user-friendly** output.

---

### **3ï¸âƒ£ Difference in Interactive Mode**
In **Python's interactive shell**, when you type an object, **`__repr__` is called by default**:
```python
car1  # Calls __repr__()
```
âœ” **Output:**
```
Car('Toyota', 'Corolla', 2022)
```
However, when using `print(car1)`, **`__str__` is called**:
```python
print(car1)  # Calls __str__()
```
âœ” **Output:**
```
2022 Toyota Corolla
```

---

### **4ï¸âƒ£ What If `__str__` Is Not Defined?**
If `__str__` is **not defined**, Python will fall back to `__repr__` when `print(obj)` is used.

#### âœ… **Example**
```python
class Car:
    def __init__(self, brand, model, year):
        self.brand = brand
        self.model = model
        self.year = year

    def __repr__(self):
        return f"Car('{self.brand}', '{self.model}', {self.year})"

car1 = Car("Toyota", "Corolla", 2022)
print(car1)  # No __str__, so it uses __repr__
```
âœ” **Output:**
```
Car('Toyota', 'Corolla', 2022)
```

---

### **5ï¸âƒ£ Summary Table**
| Feature | `__repr__` | `__str__` |
|---------|-----------|-----------|
| Purpose | For developers (debugging, logging) | For users (readable output) |
| Output Style | Detailed, **unambiguous** (valid Python expression) | Human-friendly, **readable** |
| When Called? | `repr(obj)` or just `obj` in interactive mode | `str(obj)` or `print(obj)` |
| Should Return? | String that reconstructs the object | String meant for display |

---

### **ğŸš€ Key Takeaways**
âœ” **`__repr__` is for developers**, **`__str__` is for users**.  
âœ” **If `__str__` is missing, Python falls back to `__repr__`**.  
âœ” **Always implement `__repr__`** (best practice).  
âœ” **Use `__repr__` for debugging**, **`__str__` for displaying results to users**.

## **How Can Objects Be Stored in Sets Even Though They Are Mutable?**  

### **ğŸ”¹ Understanding the Problem**  
Python **sets** store unique objects, and they rely on **hashing** to quickly check for duplicates. However, sets only allow **hashable** objects, not necessarily **immutable** ones. This leads to an interesting question:  

> **If sets only accept hashable objects, and most mutable objects (like lists and dictionaries) are unhashable, how do some mutable objects still end up inside sets?**  

---

### **ğŸ”¸ 1. Python Doesn't Test for Mutability, It Tests for Hashability**  
Python **doesn't explicitly check if an object is mutable** before storing it in a set. Instead, it checks if the object is **hashable** by calling its `__hash__()` method.  

ğŸ“Œ **Key Rule**  
âœ” **A hashable object must:**
- Implement `__hash__()` method, which returns a unique integer for the object.
- Implement `__eq__()` to ensure equality comparison.
- Have a hash value that does not change over time (i.e., its attributes must be immutable).  

âœ” **Example of Unhashable Object (Mutable List)**
```python
my_set = { [1, 2, 3] }  # âŒ TypeError: unhashable type: 'list'
```
âœ… **Example of Hashable Object (Immutable Tuple)**
```python
my_set = { (1, 2, 3) }  # âœ… Works fine
```
> Lists are mutable, so they **donâ€™t have a `__hash__()` method** and canâ€™t be stored in sets. Tuples, on the other hand, are immutable and can be hashed.

---

### **ğŸ”¸ 2. What is `__hash__()` and How Does It Work?**
The `__hash__()` method is used by Python to generate a unique identifier (integer) for objects stored in hash-based collections like **sets** and **dictionaries**.

ğŸ“Œ **Example of Hashable Object (Custom Class with `__hash__`)**
```python
class Car:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model
    
    def __hash__(self):
        return hash((self.brand, self.model))  # Hashing immutable tuple

car1 = Car("Toyota", "Corolla")
car2 = Car("Toyota", "Corolla")

print(hash(car1))  # Example: 872648392
print(hash(car2))  # Example: 872648392
```
âœ” **Why does this work?**  
- `__hash__()` returns an integer computed using the **immutable tuple `(brand, model)`**.
- Since tuples are immutable, their hash values **remain constant**.

---

### **ğŸ”¸ 3. Why Does `hash()` Return an Integer?**
When you call `hash(obj)`, Python generates a **unique integer** for the object based on its contents.  

ğŸ“Œ **Example: Hashing Different Data Types**
```python
print(hash(10))         # âœ… Works: Hashes an integer
print(hash("Python"))   # âœ… Works: Hashes a string
print(hash((1, 2, 3)))  # âœ… Works: Hashes a tuple
```
âœ” The hash function produces **fixed-length integers** that allow **quick lookups** in sets and dictionaries.

> **Key Rule:** If two objects are **equal (`__eq__` returns `True`)**, their hash values **must also be equal**.

---

### **ğŸ”¸ 4. Why Can Some Mutable Objects Be Stored in Sets?**
ğŸ“Œ **Custom Mutable Objects Can Be Hashable!**  
Even though **mutable objects like class instances** can change over time, they **can still be hashable** if they:  
1. **Implement `__hash__()` correctly** and ensure the hash value doesn't change.  
2. **Override `__eq__()`** so objects with the same attributes are considered equal.  

âœ… **Example: A Hashable Mutable Object**
```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __hash__(self):
        return hash(self.name)  # Only hashing immutable name

    def __eq__(self, other):
        return self.name == other.name

p1 = Person("Alice", 25)
p2 = Person("Alice", 30)

print(hash(p1))  # Same hash for both because only `name` is used
print(hash(p2))

my_set = {p1, p2}  # âœ… Works fine because they have the same hash!
print(my_set)  # Only one entry because they're considered equal
```
âœ” Since **only the name is hashed**, even though `age` changes, the object is **still hashable**.

---

### **ğŸ”¸ 5. What Happens if `__hash__()` is Not Defined?**
If an object **does not implement `__hash__()`**, Python checks if itâ€™s **mutable**.  

- **Immutable objects** (like `int`, `str`, `tuple`) **have a default `__hash__()`** method.
- **Mutable objects** (like `list`, `dict`) **are unhashable** by default.

ğŸ“Œ **Example: Making a List Hashable (Bad Practice)**
```python
class CustomList:
    def __init__(self, data):
        self.data = data

    def __hash__(self):
        return hash(tuple(self.data))  # Convert list to immutable tuple

c1 = CustomList([1, 2, 3])
print(hash(c1))  # âœ… Works, but risky if list changes later
```
ğŸš¨ **Warning!** If `self.data` changes, the objectâ€™s hash becomes invalid!

---

### **ğŸ”¸ 6. Why Not Just Store Objects in Lists Instead of Sets?**
If lists can store all types of objects, why do we need sets with hashing?

âœ” **Efficiency**:  
- **Sets use hashing**, so checking for duplicates is **O(1)** (very fast).  
- **Lists use iteration**, so checking for duplicates is **O(n)** (slower).  

âœ” **Example: Searching in a Set vs. List**
```python
big_list = [i for i in range(100000)]
big_set = set(big_list)

print(99999 in big_list)  # âŒ Slow: O(n) search
print(99999 in big_set)   # âœ… Fast: O(1) search
```
ğŸš€ **Sets are much faster than lists for lookups!**

| Feature | **Sets** | **Lists** |
|---------|---------|-----------|
| **Storage** | Uses **hashing** | Uses **indexing** |
| **Performance** | Fast lookups (`O(1)`) | Slow lookups (`O(n)`) |
| **Duplicates** | Not allowed | Allowed |
| **Order** | Unordered | Ordered |
| **Requires Hashing?** | âœ… Yes | âŒ No |

### **ğŸ”¹ Final Key Takeaways**
âœ” **Objects in sets must be hashable, not necessarily immutable.**  
âœ” Python **does not check for mutability**, only for the presence of `__hash__()`.  
âœ” **`hash()` function returns an integer**, allowing **quick lookups** in sets.  
âœ” **Mutable objects can be stored in sets if their hash value remains constant.**  
âœ” **Sets are faster than lists for lookups due to hashing.**
