```python
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
```

```python
a = np.linspace(-10, 9, 20, dtype=int)
b = np.linspace(-10, 9, 20, dtype=int)
a, b
```

```python
xx, yy = np.meshgrid(a, b)
```

```python
plt.scatter(xx, yy)
```

```python
def func(x, y):
    return x**2 + y**2

zz = func(xx, yy)
zz
```

```python
fig = px.scatter_3d()

fig.add_trace(go.Surface(x=xx,y=yy,z=zz))

fig.show()
```
